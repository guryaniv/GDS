```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
dir = '../input/'
```

```python
answers = pd.read_csv(dir + 'answers.csv')
questions = pd.read_csv(dir + 'questions.csv')
comments = pd.read_csv(dir + 'comments.csv')
emails = pd.read_csv(dir + 'emails.csv')
professionals = pd.read_csv(dir + 'professionals.csv')
matches = pd.read_csv(dir + 'matches.csv')

group_memberships = pd.read_csv(dir + 'group_memberships.csv')
groups = pd.read_csv(dir + 'groups.csv')

school_memberships = pd.read_csv(dir + 'school_memberships.csv')
students = pd.read_csv(dir + 'students.csv')
tag_users = pd.read_csv(dir + 'tag_users.csv')
tags = pd.read_csv(dir + 'tags.csv')
tag_questions = pd.read_csv(dir + 'tag_questions.csv')
```

```python
answers = pd.concat([answers.answers_id, answers.answers_author_id.drop_duplicates(keep = 'first'), answers.answers_question_id.drop_duplicates(keep = 'first'), answers.answers_date_added.drop_duplicates(keep = 'first'), answers.answers_body.drop_duplicates(keep = 'first')], axis = 1)
```

I'll tie up the same columns and merge them.
(professionals['professionals_id'] = comments['comments_author_id']), (emails['emails_recipient_id'] = professionals['professionals_id']).....
And I make it neat by removing columns of the same value.(by drop)

```python
professionals = professionals.merge(comments, left_on = 'professionals_id', right_on = 'comments_author_id')
emails = emails.merge(professionals, left_on = 'emails_recipient_id', right_on = 'professionals_id')
emails = emails.merge(matches, left_on = 'emails_id',right_on = 'matches_email_id')
emails.drop('matches_email_id', axis = 1, inplace = True)
```

same process...

```python
answers = answers.merge(emails, left_on = 'answers_author_id', right_on = 'emails_recipient_id')
answers.drop('emails_recipient_id', axis = 1, inplace=True)
```

```python
questions = questions.merge(answers, left_on = 'questions_id', right_on = 'answers_question_id')
questions.drop('answers_question_id', axis = 1, inplace = True)
```

```python
groups = groups.merge(group_memberships, left_on = 'groups_id', right_on = 'group_memberships_group_id')
groups.drop('group_memberships_group_id', axis = 1, inplace = True)
```

```python
tags = tags.merge(tag_users, left_on='tags_tag_id', right_on='tag_users_tag_id')
tags.drop('tag_users_tag_id', axis = 1, inplace = True)
tags = tags.merge(tag_questions, left_on = 'tags_tag_id', right_on = 'tag_questions_tag_id')
tags.drop('tag_questions_tag_id', axis = 1, inplace = True)
tags = tags.merge(school_memberships, left_on='tag_users_user_id', right_on = 'school_memberships_user_id')
tags.drop('school_memberships_user_id', axis = 1, inplace = True)
tags = tags.merge(groups, left_on='tag_users_user_id', right_on = 'group_memberships_user_id')
tags.drop('group_memberships_user_id', axis = 1, inplace = True)
```

```python
questions = pd.concat([questions, tags, students], axis = 1)
questions.drop(['tags_tag_id', 'students_id'], axis = 1, inplace = True)
questions.drop(['questions_id', 'questions_author_id', 'answers_id', 'answers_author_id', 'matches_question_id', 'tag_users_user_id', 'tag_questions_question_id', 'groups_id', 'comments_id', 'comments_author_id', 'comments_parent_content_id'], axis = 1, inplace = True)
```

```python
questions['questions'] = questions['questions_body'].astype(str) + questions['questions_date_added'].astype(str) + str('#') + questions['tags_tag_name'].astype(str)
questions.drop(['questions_body', 'questions_date_added', 'tags_tag_name'], axis = 1, inplace = True)
```

I tie the answers_date_added and answer to the answer, professionals_id and so on to an answer column.

```python
questions['answers'] = questions['answers_body'].astype(str) + questions['answers_date_added'] + str('professionals :') + questions['professionals_id'].astype(str) + questions['professionals_headline'].astype(str) + questions['professionals_location'].astype(str) + questions['professionals_industry'].astype(str) + questions['professionals_date_joined'].astype(str)
questions.drop(['answers_body', 'answers_date_added', 'professionals_date_joined', 'professionals_headline', 'professionals_industry', 'professionals_location', 'professionals_id'], axis = 1, inplace = True)
```

Comments are also merged into one in the same way as in reply.

```python
questions['comments'] = questions['comments_date_added'] + str(',') + questions['comments_body'] + str('( email  : ') + questions['emails_id'].astype(str) + questions['emails_date_sent'].astype(str) + questions['emails_frequency_level'].astype(str) + str(')')
questions.drop(['comments_date_added', 'comments_body', 'emails_id', 'emails_date_sent', 'emails_frequency_level'], axis = 1, inplace = True)
```
