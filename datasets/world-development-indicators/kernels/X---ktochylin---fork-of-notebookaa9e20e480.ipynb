```python
hist_indicator = 'Population ages 65 and above'
hist_country = 'UKR'
hist_country_usa = 'USA'
hist_country_afg = 'AFG'
hist_country_wld = 'WLD'
hist_country_tcd = 'TCD'

mask1 = data['IndicatorName'].str.contains(hist_indicator, case=False)
#print(mask1)
mask2 = data['CountryCode'].str.contains(hist_country)
mask3 = data['CountryCode'].str.contains(hist_country_usa)
mask_afg = data['CountryCode'].str.contains(hist_country_afg)
mask_wld = data['CountryCode'].str.contains(hist_country_wld)
mask_tcd = data['CountryCode'].str.contains(hist_country_tcd)

#print(mask2.any())
stage = data[mask1]
#stage = data[mask1 & mask2]
stage.head()
stage_ukraine = data[mask1 & mask2]
stage_ukraine.head()
stage_usa = data[mask1 & mask3]
stage_usa.head()
stage_afg = data[mask1 & mask_afg]
stage_afg.head()
stage_wld = data[mask1 & mask_wld]
stage_wld.head()
stage_tcd = data[mask1 & mask_tcd]
stage_tcd.head()

#stage.head()
#stage_src = stage[['CountryCode','CountryName']]
#stage_src['value_min'] = stage.groupby(['CountryCode','CountryName'])['Value'].min()
#stage_min = stage.groupby(['CountryCode','CountryName'])['Value'].min()
#stage_src = stage_max.append(stage_min)
#stage_src['value_max'] = stage.groupby(['CountryCode','CountryName'])['Value'].max()
```
