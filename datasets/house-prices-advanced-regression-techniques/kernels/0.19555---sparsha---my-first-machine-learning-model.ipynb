Determining the Sale Price of houses in the state of Iowa
The prediction is based on evaluating based on features like Lot Area, Overall Quality, Year Built and Total Rms Above Ground.

```python
#Import the necessary libraries
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
```

We are using Random Forest algorithm to determine the sale price of the houses.The basic definition of the function is as follows-
RandomForestRegressor(bootstrap=True, criterion='mse', max_depth=None,
           max_features='auto', max_leaf_nodes=None,
           min_impurity_decrease=0.0, min_impurity_split=None,
           min_samples_leaf=1, min_samples_split=2,
           min_weight_fraction_leaf=0.0, n_estimators=10, n_jobs=1,
           oob_score=False, random_state=None, verbose=0, warm_start=False)
There are parameters which allow you to change the performance of the Random Forest.
One of the best features of Random Forest models is that they generally work reasonably even without this tuning.

Provide the appriopriate path and read the data using pandas library.

```python
#Path to the train file
train_file_path = '../input/house-prices-advanced-regression-techniques/train.csv' 

# Reading the training data
data = pd.read_csv(train_file_path)
```

Declare the field to be predicted.

```python
#Field to be predicted
y=data.SalePrice
```

Among the 81 columns in the dataset, the below 4 fields are considered for determining the prediction.

```python
#Fields to be considered for prediction (features)
predictors_list=['LotArea', 'OverallQual', 'YearBuilt', 'TotRmsAbvGrd']
train_X=data[predictors_list]
```

Use the Random Forest function and fit the model using the training data and y as the expected output.

```python
#Using Random Forest and fit the model with the training data
forest_model = RandomForestRegressor()
forest_model.fit(train_X, y)
```

After successfully training the model, test it using the test dataset and forecast the predictions.

```python
#Obtain the test data for prediction
test_file_path = pd.read_csv('../input/house-prices-advanced-regression-techniques/test.csv')
test_X= test_file_path[predictors_list]
iowa_preds = forest_model.predict(test_X)
```

```python
#Submit the output
my_submission = pd.DataFrame({'Id': test_file_path.Id, 'SalePrice': iowa_preds})
my_submission.to_csv('submission.csv', index=False)
```
