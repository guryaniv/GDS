```python
y=data.SalePrice
X=data[["LotArea","YearBuilt","1stFlrSF","2ndFlrSF","FullBath","BedroomAbvGr","TotRmsAbvGrd"]]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0)
```

```python
iowamodel1 = DecisionTreeRegressor()
iowamodel1.fit(X_train,y_train)
pred = iowamodel1.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test,pred)
```

```python
def get_mae(max_nodes, X_train,X_test,y_train,y_test):
    iowamodel2 = DecisionTreeRegressor(max_leaf_nodes=max_nodes, random_state=0)
    iowamodel2.fit(X_train,y_train)
    pred = iowamodel2.predict(X_test)
    mae = mean_absolute_error(y_test,pred)
    return mae
```

```python
for max_nodes in [5,50,80,500,5000]:
    mae = get_mae(max_nodes, X_train,X_test,y_train,y_test)
    print(max_nodes,mae)
```

```python
from sklearn.ensemble import RandomForestRegressor

forestmodel = RandomForestRegressor(random_state=1)
forestmodel.fit(X_train,y_train)
pred = forestmodel.predict(X_test)
mean_absolute_error(y_test,pred)
```

```python
test = pd.read_csv('../input/house-prices-advanced-regression-techniques/test.csv')
test_X = test[["LotArea","YearBuilt","1stFlrSF","2ndFlrSF","FullBath","BedroomAbvGr","TotRmsAbvGrd"]]
test_predict = forestmodel.predict(test_X)
test_predict
```

```python
my_submission = pd.DataFrame({'Id': test.Id, 'SalePrice': test_predict})
my_submission.to_csv('submission.csv', index=False)
print("output written")
```
