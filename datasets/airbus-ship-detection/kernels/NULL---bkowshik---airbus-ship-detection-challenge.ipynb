```python
import csv
import os
import random

import numpy as np
import pandas as pd
from PIL import Image
```

```python
# What are the differet folders in the dataset?
os.listdir("../input")
```

```python
def from_run_length(run_length, image):
    """Convert from Run Length Eocoding to a PIL.Image object."""
    
    label = np.zeros(np.prod(image.size), dtype=np.int8)
    
    if run_length:
        
        # Split run lengths into start and length.
        parts = [int(item) for item in run_length.split(' ')]

        # Reshape into an array of kind [[start, length], ...]
        pairs = np.array(parts, dtype=int).reshape(-1, 2)

        # Update value from zero to one.
        for start, length in pairs:
            label[start:start+length] = 1
    
    # Convert from Numpy arrays to PNG image.
    label =  Image.fromarray(label.reshape(image.size).transpose(), mode='P')    
    palette = [0, 0, 0, 255, 0, 0]
    label.putpalette(palette)
    
    return label
```

```python
with open ('../input/train_ship_segmentations.csv') as fp:
    reader = csv.DictReader(fp)
    segments = {}
    for row in reader:
        segments[row['ImageId']] = row['EncodedPixels']

print('Number of segments in training: {}'.format(len(segments)))
```

```python
# What does the training dataset look like?
images = os.listdir("../input/train/")
print('Number of images for training: {}'.format(len(images)))

images[:5]
```

```python
# Number of images and number of segments should be the same.
assert len(images) == len(segments)

with_ships = [image for image in images if segments[image]]

print('Images with ships: {}'.format(len(with_ships)))
print('Images without ships: {} ({}%)'.format(len(images) - len(with_ships), round(100.0 * len(with_ships) / len(images), 2)))
```
