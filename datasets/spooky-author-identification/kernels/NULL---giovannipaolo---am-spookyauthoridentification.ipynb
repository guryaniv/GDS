```python
df = pd.read_csv('../input/train.csv')
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, log_loss
from sklearn.pipeline import make_pipeline
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from xgboost import XGBClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(df['text'],
                                                    df['author'],
                                                    stratify=df['author'],
                                                    random_state=42)
```

```python
%%time
clf1 = make_pipeline(TfidfVectorizer(max_features=2000),
                     SVC(),
                     memory='cache',
                    ).fit(X_train, y_train)
```

```python
%%time
clf2 = make_pipeline(TfidfVectorizer(max_features=2000),
                     GradientBoostingClassifier(n_estimators=200),
                     memory='cache',
                    ).fit(X_train, y_train)
```

```python
%%time
clf3 = make_pipeline(TfidfVectorizer(max_features=2000),
                     RandomForestClassifier(n_estimators=200),
                     memory='cache',
                    ).fit(X_train, y_train)
```

```python
%%time
clf4 = make_pipeline(TfidfVectorizer(max_features=2000),
                     XGBClassifier(n_estimators=200),
                     memory='cache',
                    ).fit(X_train, y_train)
```

```python
%%time
clf5 = make_pipeline(TfidfVectorizer(max_features=2000, stop_words='english', ngram_range=(1,3)),
                     XGBClassifier(n_estimators=200),
                     memory='cache',
                    ).fit(X_train, y_train)
```

```python
%%time
clf6 = make_pipeline(TfidfVectorizer(max_features=2000),
                     MLPClassifier(),
                     memory='cache',
                    ).fit(X_train, y_train)
```

```python
%%time
clf7 = make_pipeline(TfidfVectorizer(max_features=2000),
                     MLPClassifier(hidden_layer_sizes=(200,200)),
                     memory='cache',
                    ).fit(X_train, y_train)
```

```python
%%time
clf8 = make_pipeline(TfidfVectorizer(max_features=2000),
                     MLPClassifier(hidden_layer_sizes=(500,)),
                     memory='cache',
                    ).fit(X_train, y_train)
```

```python
def eval_clf(clf, X_test, y_test):
    print(classification_report(y_test, clf.predict(X_test)))
    print('Cross-entropy:', log_loss(y_test, clf.predict_proba(X_test)))
    print('Accuracy:', accuracy_score(y_test, clf.predict(X_test)))
    print('Confusion matrix:\n', confusion_matrix(y_test, clf.predict(X_test)))
```

```python
eval_clf(clf8, X_test, y_test)
```

```python
[eval_clf(x, X_test, y_test) for x in [clf2, clf3, clf4, clf5]]
```
