
# House Prices - Advanced Regression Techniques



```python
import pandas as pd

main_file_path = '../input/train.csv'
data = pd.read_csv(main_file_path)
print(data.describe())
```

```python
print(data.columns)
```

```python
y = data['SalePrice']
```

```python
predictor_cols = ['LotArea','Utilities','HouseStyle','OverallCond','YearBuilt','CentralAir','Electrical','KitchenQual','YrSold','SaleType','SaleCondition']
X = data[predictor_cols]
```

```python
import numpy as np
X.fillna(-999,inplace=True)
categorical_features_indices = np.where(X.dtypes != np.float)[0]
```

```python
from catboost import CatBoostRegressor

from sklearn.model_selection import train_test_split

X_train, X_validation, y_train, y_validation = train_test_split(X, y, train_size=0.75, random_state=42)

my_model = CatBoostRegressor(learning_rate=0.01, loss_function='RMSE', logging_level='Silent')
my_model.fit(
    X_train, y_train,
    cat_features=categorical_features_indices,
    eval_set=(X_validation, y_validation),
#     logging_level='Verbose',  # you can uncomment this for text output
    plot=True
);
```

```python
# Read the test data
test = pd.read_csv('../input/test.csv')
# Treat the test data in the same way as training data. In this case, pull same columns.
test_X = test[predictor_cols]
test_X.fillna(-999,inplace=True)
# Use the model to make predictions
predicted_prices = my_model.predict(test_X)
```

```python
my_submission = pd.DataFrame({'Id': test.Id, 'SalePrice': predicted_prices})
# you could use any filename. We choose submission here
my_submission.to_csv('submission_CB.csv', index=False)
```
