```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
import math
```

```python
data = pd.read_csv("../input/kc_house_data.csv")
data.head()
```

```python
print(data.shape)
print(data.dtypes)
print(data.describe())
```

**Create X and y based on selected features and label**

```python
X = data[['bedrooms','bathrooms','sqft_living','floors']]
y = data['price']
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.2, random_state=0)
print(Xtrain.shape)
print(Xtest.shape)
```

```python
print(type(Xtrain))
print(type(ytrain))
```

```python
Xtrain = np.asmatrix(Xtrain)
Xtest = np.asmatrix(Xtest)
ytrain = np.ravel(ytrain)
ytest = np.ravel(ytest)
print(type(Xtrain))
print(type(ytrain))
```

```python
plt.scatter(X['sqft_living'], y)
plt.show()
```

```python
X['sqft_living'].hist()
plt.show()
```

**Model train**

```python
model = LinearRegression()
model.fit(Xtrain, ytrain)
```

```python
print(pd.DataFrame(list(zip(X.columns, np.transpose(model.coef_)))))
```

```python
model.intercept_
```

```python
pred = model.predict(Xtrain)
((pred-ytrain)*(pred-ytrain)).sum() / len(ytrain)  # MSE
```

```python
predtest = model.predict(Xtest)
((predtest-ytest)*(predtest-ytest)).sum() / len(ytest)
```

```python
(abs(predtest-ytest)/ytest).sum() / len(ytest)
```
