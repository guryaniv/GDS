```python
def norm(x):
    if x < 0:
        return 0
    return int(x*1000) / 1000

norm = np.vectorize(norm)

results = pd.DataFrame({
    'test_id' : sub["atest_id"],
    'price' : norm(sub["price"])
})

results.to_csv('sample_submission.csv', index=False)
```
