```python
import pandas as pd

main_file_path = '../input/train.csv'
data = pd.read_csv(main_file_path)
print(data.describe())
data.head()
```

```python
print(data.columns)
```

```python
house_price_data = data.SalePrice
house_price_data.head()
```

```python
columns_of_interest = ['SaleType', 'SaleCondition']
two_column_data = data[columns_of_interest]
two_column_data.describe()
```

```python
target_Variable = 'SalePrice'
y = data[target_Variable]
y.head()
```

```python
predictors =['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X = data[predictors]
```

```python
from sklearn.tree import DecisionTreeRegressor

house_pred_model = DecisionTreeRegressor()

house_pred_model.fit(X,y)
```

```python
print(house_pred_model.predict(X.head()))
print(y.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
train_X, val_X, train_y, val_y = train_test_split(X,y,random_state = 0)

house_pred_model = DecisionTreeRegressor()

house_pred_model.fit(train_X, train_y)

val_pred = house_pred_model.predict(val_X);

print(mean_absolute_error(val_pred, val_y))
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor

def get_mae(max_leaf_nodes, predictors_train, predictors_val, target_train, target_val):
    model = DecisionTreeRegressor(max_leaf_nodes = max_leaf_nodes, random_state = 0)
    model.fit(predictors_train, target_train)
    return mean_absolute_error(model.predict(predictors_val),target_val)
```

```python
for max_leaf_nodes in [5,50,100,500,1000,2000,3000,4000,5000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Max Leaf Nodes: %d  \t\t Mean Absolute Error: %d" %(max_leaf_nodes,my_mae))
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
random_forest_model = RandomForestRegressor(n_estimators = 100, n_jobs = 1)

random_forest_model.fit(train_X, train_y)

pred_val_y = random_forest_model.predict(val_X)

random_forest_mae = mean_absolute_error(val_y, pred_val_y)
print(random_forest_mae)
```

```python
for n_tree in [10,50,100,500,1000]:
    for max_depth_tree in [5,50,100,250,500,1000,5000]:
        random_forest_model = RandomForestRegressor(n_estimators = n_tree, n_jobs = -1,max_depth = max_depth_tree)
        random_forest_model.fit(train_X, train_y)
        pred_val_y = random_forest_model.predict(val_X)
        random_forest_mae = mean_absolute_error(val_y, pred_val_y)
        print("n_estimators: %d \t max_depth: %d \t mean absolute error: %d"%(n_tree,max_depth_tree,random_forest_mae))
        
```
