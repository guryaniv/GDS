```python
# -*- coding: utf-8 -*-
import os
import json
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

from scipy import signal
from scipy.io import wavfile
from matplotlib import pyplot as plt
%matplotlib inline

# Input data files are available in the "../input/" directory.
from subprocess import check_output
print(check_output(["ls", "../input"]).decode("utf8"))

# files list from startpath
def list_files(startpath, depth, visible_files=False):
    for root, dirs, files in os.walk(startpath):
        level = root.replace(startpath, '').count(os.sep)
        
        if level > depth:
            break
        indent = ' ' * 4 * (level)
        print(f'{indent}{os.path.basename(root)}/ [{len(files)}]')
        subindent = ' ' * 4 * (level + 1)
        
        if visible_files:
            for f in files:
                print('{}{}'.format(subindent, f))
                
# show Wave tone plot & spectogram image                
def showWavePlot(file_path):
    sample_rate, samples = wavfile.read(file_path)
    frequencies, times, spectogram = signal.spectrogram(samples, sample_rate)
    
    # plot samples
    plt.plot(samples, color='b')
    plt.ylabel('Amplitude')
    plt.xlabel('Time (ms)')

    # image show spectogram
    plt.figure(figsize = (10,10))
    plt.axis("off")
    plt.title(f'{os.path.basename(file_path)}')
    plt.imshow(spectogram)                 
```

```python
train_audio_path = '../input/train/audio'

list_files('../input', 3)
```

```python
audio_file = os.path.join(train_audio_path, 'house/61e50f62_nohash_1.wav')
showWavePlot(audio_file)
```

```python
use_labels = ['yes', 'no', 'up', 'down', 'left', 'right', 'on', 'off', 'stop', 'go', 'silence']

_labels = os.listdir(train_audio_path)


train_list = []
for item in _labels:
    if item in use_labels:
        files_list = os.listdir(os.path.join(train_audio_path, item))
        items = {}
        items["label"] = item
        items["filename"] = []
        for filename in files_list:
            items["filename"].append(filename)            
        train_list.append(items)
        
#with open("./train_audio.json", "w", encoding="utf-8")as json_file:
#        json.dump(train_list, json_file, ensure_ascii=False, indent="\t")
#print(f"Train audio files {len(train_list)}")

for _item in train_list:
    file_path = os.path.join(train_audio_path, _item["label"], _item["filename"][0])
    print(file_path)
    #showWavePlot(file_path)
```
