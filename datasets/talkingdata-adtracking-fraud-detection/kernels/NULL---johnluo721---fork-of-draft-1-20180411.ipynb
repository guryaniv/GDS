```python
import pandas as pd
import numpy as np
train=pd.read_csv('../input/train_sample.csv', parse_dates=True )
#train=pd.read_csv("/Users/johnluo/Desktop/Python/Dataset/train_sample.csv")
print (train.dtypes)
nb_attributed=len(train.loc[train['is_attributed']==1])
print (nb_attributed)
percent=nb_attributed/len(train)*100
print ("percentage : "+ str(percent)+"%")
```

```python
#http://strftime.org/

from datetime import datetime

date=[]
time=[]
for i in train["click_time"]:
    a=datetime.strptime(i,'%Y-%m-%d %H:%M:%S').date()
    b=datetime.strptime(i,'%Y-%m-%d %H:%M:%S').time()
    date.append(a)
    time.append(b)

date=pd.DataFrame(date,columns=["run_date"])
time=pd.DataFrame(time,columns=["time"])
train2=pd.concat([train,date,time],axis=1)
train2["click"]=1
```

#### Data Visualisation

```python
import matplotlib.pyplot as plt
%matplotlib inline

days=list(train2.run_date.unique())
days.sort()
fig = plt.figure(figsize=(25,30),dpi=50)
#char1 = fig.add_subplot(4,1,1)

for day in range(0,4):
    ax = fig.add_subplot(4,1,day+1)
    click_day=train2[train2.run_date==days[day]].pivot_table(index="time", values="click",aggfunc=np.sum)
    ax.scatter(click_day.index, click_day.click,color="blue", label=days[day])
    ax.legend(loc="upper left")
    #plt.set_xticks()

plt.show()
```
