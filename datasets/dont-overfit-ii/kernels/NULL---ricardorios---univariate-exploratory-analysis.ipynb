First of all, we are going to explore the target variable (knows also as response variable).

```python
target.value_counts()
```

There are more 1's than 0's, this is a problem for a machine learning model and  is knows as imbalanced classes problem, you can read more about this in the following link:

We are going to explore the type of the independient variables (knows as predictors) of the pandas dataframe.

```python
print((df_train.dtypes).head())
```

```python
print((df_train.dtypes).tail())
```

It seems that all the input variables are float.  You can check this with the following code:

```python
types_columns = df_train.dtypes
counter = 0 

for x in types_columns:
    if x == "float64":
        counter += 1 
print(counter) # if counter is 300 then all the variables are float
```

We are going to calculate the min,max, mean, and standard deviation of each input variable, if there are variables that have a great range of variation then machine learning models can be affected by this. 

```python
for x in df_train.columns:
    print("Minimum {}, Maximum {}, Mean {}, Standard Deviation {}".format(df_train[x].min(), df_train[x].max(), df_train[x].mean(), df_train[x].std() ))
```

It seems that all explanatory variables have a range of variation between -3 and 3, moreover the mean is near to zero and the standard deviation is near to 1.  Due to this we are not going to apply any transformation like normalization to the data. You can read more about this in the following link:

[What is normalization](https://www.quora.com/What-is-normalization-in-machine-learning)
