### The variable visitStartTime is a timestamp in Unix epoch time format.  Python has built-in tools to help you deal with this. 

### From Wikipedia, the free encyclopedia

### Unix time (also known as POSIX time or UNIX Epoch time) is a system for describing a point in time, defined as an approximation of the number of seconds that have elapsed since 00:00:00 Coordinated Universal Time (UTC), Thursday, 1 January 1970.

### Here are some example values you have seen in the variable "visitStartTime"

```python
import pandas as pd
ts = pd.DataFrame([1472830385, 1472880147, 1472865386, 1472881213, 1472822600, 1472807194, 1472817242], columns=['visitStartTime'])
ts.head()
```

### Let's convert this to a pandas "datetime" object 

```python
ts['visitStartTime'] = pd.to_datetime(ts['visitStartTime'], unit='s')
ts
```

### Now you can extract any part of the datetime you need, such as 'year', 'month', 'day', 'hour', 'minute', 'second', etc.

```python
ts['year'] = ts['visitStartTime'].dt.year
ts['month'] = ts['visitStartTime'].dt.month
ts['day'] = ts['visitStartTime'].dt.day
ts['hour'] = ts['visitStartTime'].dt.hour
ts['minute'] = ts['visitStartTime'].dt.minute
ts['second'] = ts['visitStartTime'].dt.second
ts['weekday'] = ts['visitStartTime'].dt.weekday
ts['weekofyear'] = ts['visitStartTime'].dt.weekofyear
ts
```
