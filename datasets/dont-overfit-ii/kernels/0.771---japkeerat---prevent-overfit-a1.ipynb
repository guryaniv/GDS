```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import RFE
from sklearn.model_selection import GridSearchCV
import warnings

sbn.set(style = 'white')
warnings.filterwarnings('ignore')
```

```python
train_data = pd.read_csv('../input/train.csv')
```

```python
train_data.head()
```

```python
train_data.drop('id', axis = 1, inplace = True)
```

```python
features = train_data.iloc[:,1:].values
target = train_data.iloc[:,0].values
```

```python
scale = StandardScaler()
features = scale.fit_transform(features)
```

```python
reg = RandomForestRegressor(n_estimators = 750, criterion = 'mse', max_features = 'sqrt', random_state = 42)
reg.fit(features, target)

selector = RFE(reg, 25, 1)
selector.fit(features, target)
```

```python
test_data = pd.read_csv('../input/test.csv')
test_data.head()
```

```python
ids = np.asarray(test_data['id'].tolist())
test_data.drop('id', axis = 1, inplace = True)
```

```python
target_features = test_data.iloc[:,:].values
target_features = scale.transform(target_features)

y_predict = selector.predict(target_features)
```

```python
submission = pd.read_csv('../input/sample_submission.csv')
submission['target'] = y_predict
submission.to_csv('submission.csv', index = False)
```
