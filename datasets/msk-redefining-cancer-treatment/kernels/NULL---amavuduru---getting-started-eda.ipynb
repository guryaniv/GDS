#Reading Data and Getting Set Up

```python
train = pd.read_csv('../input/training_variants')
test = pd.read_csv('../input/test_variants')
trainx = pd.read_csv('../input/training_text', sep="\|\|", engine='python', header=None, skiprows=1, names=["ID","Text"])
testx = pd.read_csv('../input/test_text', sep="\|\|", engine='python', header=None, skiprows=1, names=["ID","Text"])
```

So it looks like each ID corresponds to a unique gene and variation along with the rules in the form of text explaining the mutation. Given this information, we can join the two dataframes by ID to create a complete training set.

```python
train_data = pd.concat([trainx, train.drop('ID', axis=1)], axis=1)
train_data.head()
```

# Starting EDA

First let's check out the distribution of classes from the training data

```python
plt.figure(figsize=(12,8))
sns.countplot(x="Class", data=train_data)
plt.ylabel('Frequency', fontsize=12)
plt.xlabel('Class Count', fontsize=12)
plt.xticks(rotation='vertical')
plt.title("Frequency of Classes", fontsize=15)
```

Based on this plot, we can see that the class distribution is not balanced at all! Most of the mutations belong to classes 1, 2, 4, or 7, and a very small fraction of the classes belong to either class 8 or 9.
