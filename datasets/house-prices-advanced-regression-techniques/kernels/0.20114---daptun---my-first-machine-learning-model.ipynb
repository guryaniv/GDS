```python
import pandas as pd

main_file_path = '../input/train.csv'
data = pd.read_csv(main_file_path)
#print(data.describe())
```

```python

#print(data.columns)
price_data=data.SalePrice
#print(price_data.head())
cols=['LotArea', 'YrSold']
extra_data=data[cols]
#extra_data.describe()
```

```python
#Choose prediction target
y=data.SalePrice
#y.describe()

#Choose predictors
preds=['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
x=data[preds]
#x.describe()

#define model
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor()
model.fit(x,y)

#make some predictions
print("For the houses")
print(x.head())
print("The predictions are")
print(model.predict(x.head()))
print(y.head())
```

```python
#Error analysis using data used to construct... bad
from sklearn.metrics import mean_absolute_error

predicted_prices=model.predict(x)
mean_absolute_error(y, predicted_prices)
```

```python
#split the data, then use half the data to predict the other half

from sklearn.model_selection import train_test_split

train_x, val_x, train_y, val_y = train_test_split(x,y,random_state=0)
model=DecisionTreeRegressor()
model.fit(train_x,train_y)
val_preds=model.predict(val_x)

print(mean_absolute_error(val_y,val_preds))
```

```python
#get creative and write a bloody function
def get_mae(max_leaf_nodes, predictors_train, predictors_val, targ_train, targ_val):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(predictors_train, targ_train)
    preds_val = model.predict(predictors_val)
    mae = mean_absolute_error(targ_val, preds_val)
    return(mae)

for max_leaf_nodes in [5, 20, 50, 100, 5000]:
    my_mae = get_mae(max_leaf_nodes, train_x, val_x, train_y, val_y)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
```

```python
#It's #MachineLearningTime
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

forest_model = RandomForestRegressor()
forest_model.fit(train_x, train_y)
predics = forest_model.predict(val_x)
print(mean_absolute_error(val_y, predics))
```

```python
#Submission test

test=pd.read_csv('../input/test.csv')
test_x=test[preds]
predicted_prices = forest_model.predict(test_x)
print(predicted_prices)

submission=pd.DataFrame({'Id': test.Id, 'SalePrice': predicted_prices})
submission.to_csv('submission.csv', index=False)
```
