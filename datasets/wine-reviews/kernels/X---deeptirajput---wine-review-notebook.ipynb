```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load in 

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import scipy as sc
import matplotlib.pyplot as plt

df = pd.read_csv("../input/winemag-data-130k-v2.csv", encoding="utf-8")

# now look at the varieties
df_varieties=df['variety'].unique()
print('total varieties of wine in data')
print(df_varieties.size)
print('total wine in data')
print(df.size)

df_varieties_grp=df.groupby(by='variety')
df_varieties_grp_des=df_varieties_grp.describe()
#cross check result
df_Avidagos=df[df['variety']=='Abouriou']


def get_stats(group):
    return {'count': group.count(), 'variety_name' : group.name}

df_with_count=df['variety'].groupby(df['variety']).apply(get_stats).unstack()
df_with_count=df_with_count.sort_values(['count'],ascending=False)
df_top_five=df_with_count.head(5)
#check the type of variable
variety_list=list(df_top_five['variety_name'])
variety_count=list(df_top_five['count'])
plt.bar(variety_list,variety_count,align='edge')
plt.xlabel('wine variety')
plt.ylabel('wine counts in data')
plt.title('Top five variety of wine assessed in data')
plt.show()

plt.title('Top five variety of wine assessed in data')
plt.pie(variety_count,labels=variety_list)
plt.show()

# check the points of each variety of wine by taster(in case many taster gave points 
#calculate the avg.)


def get_stats_points(group):
       return {'mean': group.mean(),'variety_name' : group.name}

#result 
#Alicante Bouschet 89.36
# cross checking result
df_variety_points=df[['variety','points']]
sample_df=df_variety_points[df_variety_points['variety']=='Alicante Bouschet']
print('here is coming mean from dataframe')
print(sample_df['points'].mean())
print('here is coming mean from function')
print(sample_df['points'].groupby(sample_df['variety']).apply(get_stats_points).unstack())

#calculating on all data
df_variety_points_calculated_mean=df_variety_points['points'].groupby(df_variety_points['variety']).apply(get_stats_points).unstack()
df_variety_points_calculated_mean=df_variety_points_calculated_mean.sort_values(['mean'],ascending=False)

#Top five highly rated wines

df_top_five_highly_rated_wines=df_variety_points_calculated_mean.head(5)
print('==============================printing wines=======================')
print(df_top_five_highly_rated_wines)
variety_names_plt=df_top_five_highly_rated_wines['variety_name']
points_plt=df_top_five_highly_rated_wines['mean']

#Plotting wine rating bars
plt.title('Highly rated wines')
plt.xlabel('Wine name')
plt.ylabel('wine rating')
plt.bar(variety_names_plt,points_plt,align='edge')
plt.show()

#Plotting wine rating 
plt.title('Highly rated wines')
plt.xlabel('Wine name')
plt.ylabel('wine rating')
plt.plot(variety_names_plt,points_plt,'g^')
plt.show()

#Plotting wine rating line
plt.title('Highly rated wines')
plt.xlabel('Wine name')
plt.ylabel('wine rating')
plt.plot(variety_names_plt,points_plt)
plt.show()

#Plotting wine rating pie chart
plt.pie(points_plt, labels=variety_names_plt,autopct='%1.2f%%')
plt.title('Highly rated wines')
plt.show()

#Plotting wine rating scatter
plt.title('Highly rated wines')
plt.xlabel('Wine name')
plt.ylabel('wine rating')
plt.scatter(points_plt,variety_names_plt,c='r',marker='X')
plt.show()

#top rated wine variety and country relation
country_df=df['country'].unique()
print(country_df.size)

#check top rated wine and countries, it is found
top_most_wine=variety_names_plt[0]
df_country_top_rated_wine=df[df['variety']==top_most_wine]


country_list_per_variety = {}

#countries for top most rated wines
print('====================top wine with countries list=========================')
for i in range(5):
     variety_name=variety_names_plt[i]
     df_country_top_rated_wine=df[df['variety']==variety_names_plt[i]]
     countries=df_country_top_rated_wine['country'].unique()
     country_list_per_variety[variety_name]=list(countries)
     #country_list[i].append(list(countries))

print(country_list_per_variety)
# country_list_per_variety is the dictionary of countries for each variety i.e country_list[0] contains list 
# of countries for 1st variety 'Terrantez'
   
plt.title('Highly rated wines')
plt.xlabel('Wine name')
plt.ylabel('wine country')
v=np.array(list(country_list_per_variety.keys()))
a=np.array(list(country_list_per_variety.values()))
plt.plot(v,a,'ro')
plt.show()




```
