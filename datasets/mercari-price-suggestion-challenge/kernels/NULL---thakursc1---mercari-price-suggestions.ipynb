```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load in 
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt 
from sklearn import preprocessing
# Input data files are available in the "../input/" directory.
# For example, running this (by clicking run or pressing Shift+Enter) will list the files in the input directory
from subprocess import check_output
print(check_output(["ls", "../input"]).decode("utf8"))
# Any results you write to the current directory are saved as output.
train = pd.read_table("../input/train.tsv",index_col = 0)
test = pd.read_table("../input/test_stg2.tsv",index_col  = 0)
```

```python
train = train[['name', 'item_condition_id', 'category_name', 'brand_name', 'shipping',
       'item_description', 'price']]
train.head(10)
```

```python
train.describe()
```

```python
# Missing percentage 
train.isnull().sum()
```

```python
#A lot of category_name and brand_name are NAN's for now we categorize them into other
# We are left with 0 missing values after the follwoing step
train['category_name'].fillna("other_cat",inplace=True)
train['brand_name'].fillna("unknown_brand",inplace=True)
train['category_name'] = train['category_name'].apply(lambda x:x.split("/"))
#train['category_name'].apply(lambda x:len(x)).value_counts()/train.shape[0]*100
```

Lets see how price changes with product quality 

```python
import seaborn as sns
sns.boxplot(x='item_condition_id', y='price',data = train)
```
