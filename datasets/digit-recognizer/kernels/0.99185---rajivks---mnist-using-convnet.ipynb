```python
#read training data including label
df_train = pd.read_csv('../input/train.csv')
```

```python
np_data = df_train.values
#np.random.shuffle(np_data)
```

```python
#get train_data and label
train_data = np_data[:,1:]
train_label = np_data[:, 0]
```

```python
import matplotlib.pyplot as plt
plt.hist(train_label[8000:], bins=10)
```

```python
train_data = train_data.reshape((train_data.shape[0], 28,28,1))
```

```python
#recale the train_data to have values between 0 and 1
train_data = train_data.astype('float32')/255
```

```python
train_data.shape
```

```python
print('train_label shape = ', train_label.shape)
```

```python
from keras.utils import to_categorical
```

```python
#convert label into 1-hot encoded vectors
train_label = to_categorical(train_label)
```

```python
print('train_label shape = ', train_label.shape)
```

```python
#use first 8000 samples for validation
val_data = train_data[0:6000]
val_label = train_label[0:6000]
```

```python
partial_train_data = train_data[6000:]
partial_train_label = train_label[6000:]
```

```python
partial_train_data.shape
```

```python
#create nn model
from keras import models
from keras import layers

model = models.Sequential()
model.add(layers.Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)))       
model.add(layers.MaxPooling2D((2,2)))
model.add(layers.Conv2D(64, (3,3), activation='relu'))
model.add(layers.Dropout(0.3))
model.add(layers.MaxPooling2D((2,2)))
model.add(layers.Conv2D(64, (3,3), activation='relu'))
model.add(layers.Dropout(0.5))
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
#model.add(layers.Dense(32, activation='relu'))

model.add(layers.Dense(10, activation='softmax'))
```

```python
model.summary()
```

```python
from keras import optimizers
opt = optimizers.rmsprop(lr=0.0001)
```

```python
model.compile(loss='categorical_crossentropy', optimizer = 'rmsprop', metrics=['acc'])
```

```python
history = model.fit(partial_train_data, partial_train_label, epochs=20, batch_size=64,
                   validation_data=(val_data, val_label))
```

```python
history_stats = history.history
history_stats.keys()
```

```python
import matplotlib.pyplot as plt
epochs = 20
plt.plot(range(1,epochs + 1), history_stats['loss'], 'bo', label='Loss')
plt.plot(range(1,epochs + 1), history_stats['val_loss'], 'b', label='Validation Loss')
plt.legend()
plt.show()
```

```python
plt.clf()
plt.plot(range(1,epochs + 1), history_stats['acc'], 'bo', label='Accuracy')
plt.plot(range(1,epochs + 1), history_stats['val_acc'], 'b', label='Validation Accuracy')
plt.legend()
plt.show()
```
