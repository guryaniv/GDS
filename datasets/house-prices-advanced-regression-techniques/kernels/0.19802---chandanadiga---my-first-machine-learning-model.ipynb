# My First Machine Learning Model
Built in algorithms used are DecisionTreeRegressor & RandomForestRegressor.

```python
import pandas as pd
# from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

main_file_path_train = '../input/house-prices-advanced-regression-techniques/train.csv' # this is the path to the Iowa data that you will use
data_train = pd.read_csv(main_file_path_train)

interested_columns = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
train_X = data_train[interested_columns]
train_y = data_train.SalePrice

# data_model = DecisionTreeRegressor(max_leaf_nodes=500, random_state=0)
data_model = RandomForestRegressor()
data_model.fit(train_X, train_y)

main_file_path_test = '../input/house-prices-advanced-regression-techniques/test.csv' # this is the path to the Iowa data that you will use
data_test = pd.read_csv(main_file_path_test)
interested_columns_test_all = ['Id', 'LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
test_X_with_id = data_test[interested_columns_test_all]

print("Initial len: " + str(len(test_X_with_id)))
test_X_with_id = test_X_with_id.dropna()
print("After dropping NAN, len: " + str(len(test_X_with_id)))

test_X = test_X_with_id[interested_columns]

predicted_home_prices = data_model.predict(test_X)

my_submission = pd.DataFrame({'Id': test_X_with_id.Id, 'SalePrice': predicted_home_prices})
my_submission.to_csv('submission.csv', index=False)
```
