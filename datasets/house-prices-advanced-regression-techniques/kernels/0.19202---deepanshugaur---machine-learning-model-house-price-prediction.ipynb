```python
two_features=['LotArea','YearBuilt']
two_features_data=data[two_features]
print(two_features_data.describe())
```

```python
y=data.SalePrice 
price_predictors=['LotArea','YearBuilt','1stFlrSF','2ndFlrSF','FullBath','BedroomAbvGr','TotRmsAbvGrd']
X=data[price_predictors]

from sklearn.tree import DecisionTreeRegressor
price_model= DecisionTreeRegressor()
price_model.fit(X,y)
```

```python
print("Predicitons for 1st 5 houses are as follows : ")
print(X.head())
print("The price prediction of houses are as follows :")
print(price_model.predict(X.head()))
```

```python
from sklearn.metrics import mean_absolute_error
predicted_house_prices = price_model.predict(X)
mean_absolute_error(y, predicted_house_prices)
```

```python
from sklearn.model_selection import train_test_split

train_X, val_X, train_y, val_y = train_test_split(X, y,random_state = 0)
price_model = DecisionTreeRegressor()

price_model.fit(train_X, train_y)


val_predictions = price_model.predict(val_X)
print(mean_absolute_error(val_y, val_predictions))
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor

def get_mae(max_leaf_nodes, predictors_train, predictors_val, targ_train, targ_val):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(predictors_train, targ_train)
    preds_val = model.predict(predictors_val)
    mae = mean_absolute_error(targ_val, preds_val)
    return(mae)

for max_leaf_nodes in [5, 50,100, 500, 5000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

forest_model = RandomForestRegressor()
forest_model.fit(train_X, train_y)
melb_preds = forest_model.predict(val_X)
print(mean_absolute_error(val_y, melb_preds))
```

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor


train = pd.read_csv('../input/train.csv')


train_y = train.SalePrice
predictor_cols = ['LotArea', 'OverallQual', 'YearBuilt', 'TotRmsAbvGrd']


train_X = train[predictor_cols]

my_price_model = RandomForestRegressor()
my_price_model.fit(train_X, train_y)
```

```python
test = pd.read_csv('../input/test.csv')

test_X = test[predictor_cols]

predicted_prices = my_price_model.predict(test_X)

print(predicted_prices)
```

```python
my_price_submission = pd.DataFrame({'Id': test.Id, 'SalePrice': predicted_prices})

my_price_submission.to_csv('price_submission.csv', index=False)
```
