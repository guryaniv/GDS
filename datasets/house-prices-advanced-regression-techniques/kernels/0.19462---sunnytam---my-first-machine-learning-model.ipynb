```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

main_file_path = '../input/train.csv'
data = pd.read_csv(main_file_path)

y = data.SalePrice
predictors = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X = data[predictors]
train_X, val_X, train_y, val_y = train_test_split(X, y,random_state = 0)
price_model = DecisionTreeRegressor()
price_model.fit(train_X, train_y)

predictions = price_model.predict(val_X)
print('The MAE on prediction')
print(mean_absolute_error(val_y, predictions));
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

def get_mae(max_leaf_nodes, predictors_train, predictors_val, actual_train, actual_val):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(predictors_train, predictors_val)
    preds_val = model.predict(actual_train)
    mae = mean_absolute_error(actual_val, preds_val)
    return(mae)


main_file_path = '../input/train.csv'
data = pd.read_csv(main_file_path)

y = data.SalePrice
predictors = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X = data[predictors]
train_X, val_X, train_y, val_y = train_test_split(X, y,random_state = 0)

for leaf_of_nodes in [5, 50, 500, 5000, 50000, 500000, 99999999]:
    mae = get_mae(leaf_of_nodes, train_X, train_y, val_X, val_y)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(leaf_of_nodes, mae))
```
