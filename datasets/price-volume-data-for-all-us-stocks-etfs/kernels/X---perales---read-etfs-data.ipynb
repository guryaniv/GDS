### Import some libraries

```python
import pandas as pd
import zipfile
import re
import glob
```

```python
path = '../input/Data.zip' #Data.zip folder
zf = zipfile.ZipFile(path) 
```

```python
folder = 'ETFs' 
```

I have reading problems with Stocks folder. I realize that they appear when the file which I am trying to read is empty.

#### Now I am going to list the filenames of the kind of data I want to put into a DataFrame

```python

filelist = []
for i in zf.namelist():
    '''
    Take only the files which belong to the folder previously stored.
    '''
    if re.match(folder+'*', i):
        filelist.append(i)
```

#### Here, I take the tickers' name in order to present them in the final Pandas DataFrame

```python
#Take the name of the first ticker
filelist[0].split('/')[1].split('.')[0]
```

```python
#Loop which list the tickers' name from the filelist
tickers = []

for i in filelist:
    tickers.append(i.split('/')[1].split('.')[0]) #We want the tickers name which is after the / symbol.
    
```

### Read all the .txt files and concatenate them into a dataframe taking the date as the index column

```python
list_ = []

df = pd.DataFrame()

for i in filelist:
    
    df_a = pd.read_csv(zf.open(i),index_col=0)
    list_.append(df_a)
df = pd.concat(list_, axis=1, keys= tickers)
```
