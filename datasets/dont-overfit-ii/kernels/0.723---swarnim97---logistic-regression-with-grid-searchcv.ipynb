```python
train = pd.read_csv('../input/train.csv')
```

```python
train.head()
```

```python
train['target'].value_counts()
```

```python
train.info()
```

```python
x = train.iloc[:, 2:].values
```

```python
y = train.iloc[:, 1].values
```

```python
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.1, random_state=2019)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression(random_state=2019)
```

```python
clf.fit(X_train, Y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn import metrics
```

```python
metrics.f1_score(Y_test, y_pred)
```

```python
metrics.accuracy_score(Y_test, y_pred)
```

```python
metrics.confusion_matrix(Y_test, y_pred)
```

```python
metrics.roc_auc_score(Y_test, y_pred)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
classifier = LogisticRegression(random_state=2019)
```

```python
param_grid = {
    'class_weight' : ['balanced', None], 
    'penalty' : ['l2','l1'],  
    'C' : [0.001, 0.01, 0.1, 1, 2, 3, 4, 5, 10, 100, 1000],
    'solver' : ['liblinear', 'saga'],
}
```

```python
grid_search = GridSearchCV(classifier, param_grid, cv=5, scoring='accuracy')
```

```python
grid_search.fit(X_train, Y_train)
```

```python
grid_search.best_params_
```

```python
classifier = LogisticRegression(C=0.1, class_weight=None, penalty='l1', solver='liblinear', random_state=2019)
```

```python
classifier.fit(X_train, Y_train)
```

```python
y_pred1 = classifier.predict(X_test)
```

```python
metrics.f1_score(Y_test, y_pred1)
```

```python
metrics.accuracy_score(Y_test, y_pred1)
```

```python
metrics.roc_auc_score(Y_test, y_pred1)
```
