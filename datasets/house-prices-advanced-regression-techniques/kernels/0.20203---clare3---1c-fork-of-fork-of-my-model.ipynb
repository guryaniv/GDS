```python
# This is the version for submitting!

import pandas as pd
# don't need to split it any more
# or to find the errors
from sklearn.ensemble import RandomForestRegressor

main_file_path='../input/train.csv'
data=pd.read_csv(main_file_path)
y_train=data.SalePrice
predictors=['LotArea','YearBuilt','1stFlrSF','2ndFlrSF','FullBath','BedroomAbvGr','TotRmsAbvGrd']
X_train=data[predictors]
forest_model=RandomForestRegressor(max_leaf_nodes=90)
# make it have 90 because that seemed like one of the best numbers from both of my previous notebooks
```

```python
# fit it using the whole of the training dataset
forest_model.fit(X_train,y_train)
# now open the test data!!!
test=pd.read_csv('../input/test.csv')
X_test=test[predictors]
Y=forest_model.predict(X_test)
print(Y)
```

```python
clare_submission=pd.DataFrame({'Id':test.Id,'SalePrice':Y})
clare_submission.to_csv('submission.csv',index=False)
```
