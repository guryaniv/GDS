```python
data=pd.read_csv("../input/creditcard.csv")
```

```python
data.info()
```

```python
data.corr()
```

```python
f,ax=plt.subplots(figsize=(50,50))
sb.heatmap(data.corr(),annot=True,linewidth=2,fmt=".2f",ax=ax)
plt.show()
```

```python
data.Time.plot(kind = 'line', color = 'g',label = 'Time',linewidth=1,alpha = 0.5,grid = True,linestyle = '-')
data.Amount.plot(color = 'r',label = 'Amount',linewidth=1, alpha = 0.5,grid = True,linestyle = '-.')
plt.legend()                      # legend = puts label into plot
plt.xlabel('x axis')              # label = name of label
plt.ylabel('y axis')
plt.title('Time and Amount Variables Line Plot')            # title = title of plot
plt.show()
```

```python
plt.scatter(data.Time,data.Amount,alpha=0.3,color="b")
plt.legend()
plt.xlabel("Time")
plt.ylabel("Amount")
plt.show()
```

```python
data.Time.plot(kind="hist",color="purple",bins=65,figsize=(20,20))
plt.show()
```
