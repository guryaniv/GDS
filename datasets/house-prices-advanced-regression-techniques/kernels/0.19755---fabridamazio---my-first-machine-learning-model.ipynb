```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

# Train data
main_file_path = '../input/train.csv'
data = pd.read_csv(main_file_path)
# prediction target and predictors
y = data.SalePrice
predictor_cols = ['LotArea','YearBuilt','1stFlrSF','2ndFlrSF','FullBath','BedroomAbvGr','TotRmsAbvGrd']
X = data[predictor_cols]
# model creating and fitting
model = RandomForestRegressor()
model.fit(X, y)


# Test Data
test = pd.read_csv('../input/test.csv')
test_X = test[predictor_cols]
predicted_prices = model.predict(test_X)

# Submission
my_submission = pd.DataFrame({'Id': test.Id, 'SalePrice': predicted_prices})
my_submission.to_csv('submission.csv', index=False)
```

```python
# predictions
predictions = model.predict(X)
mae = mean_absolute_error(y, predictions)
print("Mean absolute error: ", mae)
train_X, val_X, train_y, val_y = train_test_split(X, y,random_state = 0)

def get_mae(max_leaf_nodes, predictors_train, predictors_val, targ_train, targ_val):
    model = RandomForestRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(predictors_train, targ_train)
    preds_val = model.predict(predictors_val)
    mae = mean_absolute_error(targ_val, preds_val)
    return(mae)


for max_leaf_nodes in [5, 30, 35, 40, 45, 50, 100, 500, 5000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
```
