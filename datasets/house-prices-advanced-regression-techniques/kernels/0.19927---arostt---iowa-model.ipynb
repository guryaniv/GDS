Used a Random Forest to predict housing prices based on certain attributes.


```python
import pandas as pd

main_file_path = '../input/train.csv'
data = pd.read_csv(main_file_path)
print(data.describe())
```

```python
data.columns # print out all the column titles
```

```python
housing_price_data = data.SalePrice
housing_price_data.head()
```

```python
two_columns = ['MoSold','YrSold']
two_column_data = data[two_columns]
two_column_data.describe()
```

```python
# Time to start building the model
y = data.SalePrice
predictors = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X = data[predictors]
```

```python
# Make a simple decision tree
from sklearn.tree import DecisionTreeRegressor

#Define model
iowa_model = DecisionTreeRegressor()

from sklearn.model_selection import train_test_split

train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=0)

# Fit model
iowa_model.fit(train_X,train_y)

# Predictions on val data
val_predictions = iowa_model.predict(val_X)

from sklearn.metrics import mean_absolute_error

# Show error on predictions
print(mean_absolute_error(val_y, val_predictions))
```

```python
def get_mae(max_leaf_nodes, predictors_train, predictors_val, targ_train, targ_val):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(predictors_train, targ_train)
    preds_val = model.predict(predictors_val)
    mae = mean_absolute_error(targ_val, preds_val)
    return(mae)

# Loop thru and compare different max leaf node levels with a decision tree
for max_leaf_nodes in [10, 20, 40, 80, 160, 320, 640, 1280, 2560, 5120]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
```

```python
# Time to implement a random forest to see how well that  works to make predictions
from sklearn.ensemble import RandomForestRegressor
forest_model = RandomForestRegressor()
forest_model.fit(train_X, train_y)
iowa_rf_pred = forest_model.predict(val_X)
print(mean_absolute_error(val_y, iowa_rf_pred))
```

```python
# Time to write actual model to submit
train = pd.read_csv("../input/train.csv")

train_y = train.SalePrice

train_X = train[predictors]

rf_model = RandomForestRegressor()

rf_model.fit(train_X, train_y)
```

```python
# Read in and then predict test files
test = pd.read_csv("../input/test.csv")
test_X = test[predictors]
predicted_prices = rf_model.predict(test_X)
print(predicted_prices)
```

```python
# Prepare test predictions for submission
my_submission = pd.DataFrame({'Id': test.Id, 'SalePrice': predicted_prices})
# you could use any filename. We choose submission here
my_submission.to_csv('submission.csv', index=False)
```
