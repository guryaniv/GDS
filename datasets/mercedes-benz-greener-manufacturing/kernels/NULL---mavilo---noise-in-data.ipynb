```python
df_train = pd.read_csv('../input/train.csv')
```

Group rows according to values in columns ( skip ID and y)

```python
columns=df_train.drop(["ID","y"],axis=1).columns.tolist()
sameentries=df_train.groupby(columns).apply(lambda x: list(x.index)).tolist()
```

Select only those combinations of values which have more than one row and calculate their y-spread

```python
ywidth=list()
for i in sameentries:
    if len(i)>1:
        ywidth.append(df_train.iloc[i]['y'].max()-df_train.iloc[i]['y'].min())
```

mean value of spread

```python
np.mean(ywidth)
```

Histogram of spreads

```python
hits=plt.hist(ywidth,bins=np.mgrid[0:50:2],color='red')
```
