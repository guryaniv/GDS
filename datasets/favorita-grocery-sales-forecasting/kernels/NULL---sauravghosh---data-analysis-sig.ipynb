**Standard Functions**

```python
def data_quality(df):
    print ("Data Types & Data Quality Report : ")
    print(" ")
    print ("1.Data Types >>>>>>>>>>>>>>>>>>")
    print(df.dtypes)
    col_names = df.dtypes.index
    dtype_list = np.array(df.dtypes)
    categorical_dq = []
    continious_dq = []
    
    for i in range(len(dtype_list)) :
        if dtype_list[i] == 'O' or dtype_list[i] == "bool":            
            categorical_dq.append(df[col_names[i]].describe())
            
        elif dtype_list[i] != 'O' :
            continious_dq.append(df[col_names[i]].describe())
    print(" ")    
    print ("2.Categorical_Values Data Quality>>>>>>>>")
    print (pd.DataFrame(categorical_dq))
    print(" ")
    print ("3.Continious_Values Data Quality>>>>>>>>>")
    print (pd.DataFrame(continious_dq))
        
```

```python
def distinct_values(df,col_list):
    for c in col_list :
        print ("Column Name : ", c)
        print ("Total Row Count : ", len(df[c].value_counts()))
        dataset = pd.DataFrame(df[c].value_counts())
        dataset.columns = ["Count"]
        dataset["%"] =  dataset["Count"]/float(dataset["Count"].sum())
        print (dataset)
    
```

```python
def get_week_month_year(df,col_name):
    df[col_name] = pd.to_datetime(df[col_name])
    df["week"] = df[col_name].dt.week
    df["month"] = df[col_name].dt.month
    df["year"] = df[col_name].dt.year
    return df
```
