```python
import pandas as pd

main_file_path = '../input/train.csv'
data = pd.read_csv(main_file_path)
print (data.describe()) 
```

```python
print(data.columns)
```

```python
data_price = data.SalePrice
print (data_price.head()) 
```

```python
column_interest = ['GarageArea', 'PoolArea' ]
two_columns_of_data = data[column_interest]
two_columns_of_data.describe()
```

Création du modèle 


```python
y = data.SalePrice
data_predictors = ['LotArea','YearBuilt','1stFlrSF','2ndFlrSF','FullBath','BedroomAbvGr','TotRmsAbvGrd']
X = data[data_predictors]
print (X)
```

```python
from sklearn.tree import DecisionTreeRegressor 
# definir le modéle 
data_model = DecisionTreeRegressor()
# Appliqué le modéle 
data_model.fit(X,y)
```

```python
print("Making predictions for the following 5 houses:") # utilisation du modele 
print(X.head())
print("The predictions are")
print(data_model.predict(X.head())) # predire le prix selon notre modele
```

**Mean Absolute Error  (MAE metric)**
calcul de la marge d'erreur

```python
from sklearn.metrics import mean_absolute_error
predicted_home_prices = data_model.predict(X)
mean_absolute_error(y, predicted_home_prices) # Calculer la marge d'erreur
```

```python
# utiliser la data validation sur nos données
from sklearn.model_selection import train_test_split 
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
val_y.describe()
melbourne_model = DecisionTreeRegressor()
melbourne_model.fit(train_X, train_y)
val_predictions = melbourne_model.predict(val_X)
print (mean_absolute_error(val_y, val_predictions))
print(X.head()) 
print( melbourne_model.predict(X.head()) )
```

**Overfitting (l'élagage) and Underfitting using validation data**


```python
# Calcul de mae par rapport a la taille de l'arbre de décision
from sklearn.metrics import mean_absolute_error 
from sklearn.tree import DecisionTreeRegressor

def get_mae(max_leaf_nodes,predictors_train, predictors_val, targ_train, targ_val ) : 
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state = 0)
    model.fit(predictors_train, targ_train)
    preds_val = model.predict(predictors_val)
    mae = mean_absolute_error(targ_val, preds_val)
    return(mae)
for max_leaf_nodes in [5,50,500,5000] : 
        my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
        print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
        
```

**Random Forest**

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error 

forest_model = RandomForestRegressor()
forest_model.fit(train_X, train_y) # learn from the model
prediction = forest_model.predict(val_X)
print (prediction)
print( mean_absolute_error(val_y, prediction))
```

```python
test = pd.read_csv('../input/test.csv')
predictor_cols =['LotArea','YearBuilt','1stFlrSF','2ndFlrSF','FullBath','BedroomAbvGr','TotRmsAbvGrd']
test_X = test[predictor_cols]
predicted_prices = forest_model.predict(test_X)
print (predicted_prices)
submission = pd.DataFrame({'Id': test.Id, 'SalePrice': predicted_prices })
submission.to_csv('submission.csv', index = False)
```
