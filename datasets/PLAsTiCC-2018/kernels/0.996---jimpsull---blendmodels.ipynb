```python
dfparam = pd.read_csv("../input/ourpathtowherewearenewparams/subm_0.610175_2018-12-07-23-29.csv")
dforig = pd.read_csv("../input/ourpathtowhereweare/subm_0.619987_2018-12-06-19-58.csv")
dfsix = pd.read_csv("../input/sixdeepourpathtowherewearenewparams/subm_0.613687_2018-12-07-04-17.csv")
dffour = pd.read_csv("../input/depthfourourpathtowherewearenewparams/subm_0.615926_2018-12-07-16-59.csv")

dfxgb= pd.read_csv("../input/xgb-from-ideas-from-kernels-and-discussion/single_subm_0.671072_XGB_2018-11-21-14-24.csv")
dfsm = pd.read_csv("../input/smote-on-lgbm-w-ideas-from-kernels-and-discussion/single_subm_0.751366_2018-12-10-14-48.csv")
dfnn = pd.read_csv("../input/trypredictingallatoncenn71/subm_0.639462_2018-12-10-23-23.csv")
```

```python
df_blend = .3*dfparam + .075*(dforig+dfsix+dffour) + .2 * (dfxgb + dfnn)
df_blend['object_id']=dfsm['object_id']
df_blend.to_csv('blend_submission.csv',index=False)
```
