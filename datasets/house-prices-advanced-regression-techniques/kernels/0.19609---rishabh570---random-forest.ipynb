```python
import pandas as pd
```

** Importing training data **

```python
# Read the training data
train_data = pd.read_csv('../input/train.csv')
```

** Selecting from training data**

```python
predictors = ['LotArea', 'YearBuilt', 'PoolArea', 'LotFrontage', 'EnclosedPorch', 'OverallQual']
train_X = train_data[predictors]
train_y = train_data.SalePrice
print(train_X.head())
print('\ntraining data SalePrice : \n')
print(train_y)
```

** Handling missing values in data**

```python
from sklearn.preprocessing import Imputer
my_imputer = Imputer()
imp_train_X = my_imputer.fit_transform(train_X)
```

** DecisionTreeRegressor model**

```python
from sklearn.ensemble import RandomForestRegressor

forest_housing_model = RandomForestRegressor()
forest_housing_model.fit(imp_train_X, train_y)
```

```python
 # Make predictions on training data
predicted_prices = forest_housing_model.predict(imp_train_X)
print('Predicted prices on training data :\n')
print(predicted_prices)
```

** Finding error in prediction **

```python
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(train_y, predicted_prices))
```

** Testing data comes into play! **

```python
# Read the test data
test_data = pd.read_csv('../input/test.csv')
```

```python
# Treat the test data in the same way as training data. In this case, pull same columns.
test_X = test_data[predictors]
```

** Handle the missing value case here also ! **

```python
imp_test_X = my_imputer.fit_transform(test_X)
```

** Do prediction on test data **

```python
print('Predicted test prices are:\n')
predicted_test_prices = forest_housing_model.predict(imp_test_X)
print(predicted_test_prices)
```

```python
first_submission = pd.DataFrame({'Id': test_data.Id, 'SalePrice': predicted_test_prices})
first_submission.to_csv('submission.csv', index=False)
```
