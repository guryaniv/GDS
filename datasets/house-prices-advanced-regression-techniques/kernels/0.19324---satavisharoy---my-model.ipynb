```python
import pandas as pd
main_file_path = '../input/train.csv'
data = pd.read_csv(main_file_path)
print(data.columns)

y = data.SalePrice
predictors = ['LotArea', 'OverallQual', 'YearBuilt', 'TotRmsAbvGrd']
X = data[predictors]
from sklearn.tree import DecisionTreeRegressor
data_model = DecisionTreeRegressor()
data_model.fit(X,y)
```

```python
print("Making predictions for the following 5 houses:")
print(X.head())
print("The predictions are")
print(data_model.predict(X.head()))
```

```python
from sklearn.metrics import mean_absolute_error
predicted_home_prices = data_model.predict(X)
mean_absolute_error(y, predicted_home_prices)
```

```python
from sklearn.model_selection import train_test_split
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state= 0)
data_model = DecisionTreeRegressor()
data_model.fit(train_X, train_y)
val_predictions = data_model.predict(val_X)
print(mean_absolute_error(val_y, val_predictions))
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor

def get_mae(max_leaf_nodes, predictors_train, predictors_val, targ_train, targ_val):
    data_model = DecisionTreeRegressor(max_leaf_nodes = max_leaf_nodes, random_state = 0)
    data_model.fit(predictors_train, targ_train)
    val_predictions = data_model.predict(predictors_val)
    mae = mean_absolute_error(targ_val, val_predictions)
    return(mae)
```

```python
for max_leaf_nodes in [5, 50, 500, 5000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Max leaf nodes: %d \t\t Mean Absolute Error: %d" %(max_leaf_nodes, my_mae))
    
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

forest_model = RandomForestRegressor()
forest_model.fit(train_X, train_y)
iowa_preds = forest_model.predict(val_X)
print(mean_absolute_error(val_y, iowa_preds))
```

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor

#Read the data
train = pd.read_csv('../input/train.csv')

#pull data into target (y) and predictors (X)
train_y = train.SalePrice
predictor_cols =['LotArea', 'OverallQual', 'YearBuilt', 'TotRmsAbvGrd']

#Create training predictors data
train_X = train[predictor_cols]

iowa_model = RandomForestRegressor()
iowa_model.fit(train_X, train_y)
```

```python
#Read the test data
test = pd.read_csv('../input/test.csv')
#Treat the test data same way as training data, pull the same columns.
test_X = test[predictor_cols]

#Use the model to make predictions
predicted_prices = iowa_model.predict(test_X)
#Take a look at predicted prices to ensure they look sensible
print(predicted_prices)
```

```python
my_submission = pd.DataFrame({'Id': test.Id, 'SalePrice': predicted_prices})
# you could use any filename. We choose submission here
my_submission.to_csv('submission.csv', index=False)
```
