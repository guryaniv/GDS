```python
#choosing the Prediction target
y = data.SalePrice
#Choosing Predictors
lowa_predictors = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF',
                  'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X = data[lowa_predictors]

#Build model
from sklearn.tree import DecisionTreeRegressor
#Define model
lowa_model = DecisionTreeRegressor()
#fit model
lowa_model.fit(X, y)

#output
print('Making predictions for the fowllowing 5 houses')
print(X.head())
print('The predictions are')
print(lowa_model.predict(X.head()))
```

```python
#build MAE
from sklearn.metrics import mean_absolute_error
predicted_home_price = lowa_model.predict(X)
mean_absolute_error(y, predicted_home_price)
```

Split Data

```python
from sklearn.model_selection import train_test_split

#set data
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
#define model
lowa_model = DecisionTreeRegressor()
#fit model
lowa_model.fit(train_X, train_y)
#get predicted prices on validation data
val_predictions = lowa_model.predict(val_X)
print(mean_absolute_error(val_y, val_predictions))
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor

def get_mea(max_leaf_nodes, predictors_train, predictors_val, targ_train, targ_val):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(predictors_train, targ_train)
    preds_val = model.predict(predictors_val)
    mae = mean_absolute_error(targ_val, preds_val)
    return(mae)

#compare MAE with different values of max_leaf_nodes
for max_leaf_nodes in [5, 50, 500, 5000]:
    my_mae = get_mea(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Max leaf nodes: %d \t\t Mean Absolute Error: %d" %(max_leaf_nodes, my_mae))
```

**Random Forest.**

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

forest_model = RandomForestRegressor()
forest_model.fit(train_X, train_y)
lowa_forest_preds = forest_model.predict(val_X)
print(mean_absolute_error(val_y, lowa_forest_preds))
```

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor

# Read the data
train = pd.read_csv('../input/train.csv')

# pull data into target (y) and predictors (X)
train_y = train.SalePrice
predictor_cols = ['LotArea', 'OverallQual', 'YearBuilt', 'TotRmsAbvGrd']

# Create training predictors data
train_X = train[predictor_cols]

my_model = RandomForestRegressor()
my_model.fit(train_X, train_y)
```

```python
# Read the test data
test = pd.read_csv('../input/test.csv')
# Treat the test data in the same way as training data. In this case, pull same columns.
test_X = test[predictor_cols]
# Use the model to make predictions
predicted_prices = my_model.predict(test_X)
# We will look at the predicted prices to ensure we have something sensible.
print(predicted_prices)
```

```python
my_submission = pd.DataFrame({'Id': test.Id, 'SalePrice': predicted_prices})
# you could use any filename. We choose submission here
my_submission.to_csv('submission.csv', index=False)
```
