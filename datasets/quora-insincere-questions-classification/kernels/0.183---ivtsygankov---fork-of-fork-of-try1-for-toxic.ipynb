```python
import numpy as np
import pandas as pd
```

```python
trainset = pd.read_csv(r"../input/train.csv")

import re
datatr=[]
keystr=[]
for i in range(0, len(trainset)):
    m=trainset.question_text[i]
    m=m.lower()
    datatr.append(re.split("[^a-z]", m))
    keystr.append(trainset.target[i])

for i in range(0, len(datatr)):
    for j in range (len(datatr[i])-1, -1, -1):
        if datatr[i][j]==' ' or datatr[i][j]=='':
            del datatr[i][j]            
```

```python
testset = pd.read_csv(r"../input/test.csv")

datate=[]
idte=[]
for i in range(0, len(testset)):
    m=testset.question_text[i]
    m=m.lower()
    datate.append(re.split("[^a-z]", m))
    idte.append(testset.qid[i])

for i in range(0, len(datate)):
    for j in range (len(datate[i])-1, -1, -1):
        if datate[i][j]==' ' or datate[i][j]=='':
            del datate[i][j]  
```

```python
import nltk
```

```python
lemma = nltk.WordNetLemmatizer()
for i in range(0, len(datatr)):
    datatr[i] = [lemma.lemmatize(each) for each in datatr[i]]
```

```python
for i in range(0, len(datate)):
    datate[i] = [lemma.lemmatize(each) for each in datate[i]]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
max_features = 20
count_vectorizer = CountVectorizer(stop_words = "english",max_features = max_features) 
datatr1 = count_vectorizer.fit_transform(trainset.question_text).toarray()
datate1 = count_vectorizer.fit_transform(testset.question_text).toarray()
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(datatr1,keystr)
```

```python
y_pred = nb.predict(datate1)
```

```python
answer1='qid,prediction\n'
for i in range(0,len(idte)):
    answer1=answer1+str(idte[i])+','+str(int(y_pred[i]))+'\n'

f = open('submission.csv', 'w')
f.write(answer1)
f.close()
```
