```python
from sklearn.tree import DecisionTreeRegressor
#Choosing Prediction Target
y = lowa_sale_price
#Choosing Predictors
predictors = ['LotArea','YearBuilt','1stFlrSF','2ndFlrSF','FullBath','BedroomAbvGr','TotRmsAbvGrd']
X = data[predictors]
#Define Model
Prediction_Model = DecisionTreeRegressor()
#Fit Model
Prediction_Model.fit(X, y)
```

```python
#Time to make predictions
Prediction_Model.predict(X.head())
```

```python
y.head()
```

```python
#Mean Absolute Error
from sklearn.metrics import mean_absolute_error
#Predict again for all
prediction = Prediction_Model.predict(X)
#Check Error
mean_absolute_error(y, prediction)
```

```python
from sklearn.model_selection import train_test_split
#get the train and test values of X and y
train_X, test_X, train_y, test_y = train_test_split(X, y, random_state=0)
#Define model
model = DecisionTreeRegressor()
#Fit Model
model.fit(train_X,train_y)

#Predict For Test value
predictions = model.predict(test_X)

#Get The Mean Absolute Error
from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(test_y, predictions))
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor

def get_mae(max_leaf_nodes, predictors_train, predictors_val, targ_train, targ_val):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(predictors_train, targ_train)
    preds_val = model.predict(predictors_val)
    mae = mean_absolute_error(targ_val, preds_val)
    return(mae)
```

```python
for max_leaf_nodes in [25, 40, 50, 75, 110, 150, 300, 500, 1500, 5000]:
    my_mae = get_mae(max_leaf_nodes, train_X, test_X, train_y, test_y)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

forest_model = RandomForestRegressor()
forest_model.fit(train_X, train_y)
melb_preds = forest_model.predict(test_X)
print(mean_absolute_error(test_y, melb_preds))
```

```python
#Start to submit
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
```

```python
#Get the training set
train = pd.read_csv('../input/train.csv')
```

```python
train_y = train.SalePrice
```

```python
predictor_cols = ['LotArea', 'OverallQual', 'YearBuilt', 'TotRmsAbvGrd']
train_X = train[predictor_cols]
```

```python
#Define and fit model
model = RandomForestRegressor()
model.fit(train_X,train_y)
```

```python
# Read the test data
test = pd.read_csv('../input/test.csv')
# Treat the test data in the same way as training data. In this case, pull same columns.
test_X = test[predictor_cols]
# Use the model to make predictions
predicted_prices = model.predict(test_X)
# We will look at the predicted prices to ensure we have something sensible.
print(predicted_prices)
```

```python
#Prepare dat to submit
my_submission = pd.DataFrame({'Id': test.Id, 'SalePrice':predicted_prices})
#Final Edit
my_submission.to_csv('submission.csv', index = False)
```
