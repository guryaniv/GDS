```python
import pandas as pd

main_file_path = '../input/train.csv'
test_file_path = '../input/test.csv'
data = pd.read_csv(main_file_path)
test_data = pd.read_csv(test_file_path)
"""Get a simple describe of whole data"""
# print(data.describe())
"""Get list of columns"""
# print(data.columns)
"""Select data from multiple columns"""
pref_columns = ["LotArea", "YearBuilt", "1stFlrSF", "2ndFlrSF",
               "FullBath", "BedroomAbvGr", "TotRmsAbvGrd"]
# print(data[pref_columns])
"""Describe selected columns"""
pref_data = data[pref_columns]
# print(pref_data.describe())

###############################################
############ BUILDING MODELS ##################
###############################################

"""Select target and variable columns from dataset"""

y = data.SalePrice
X = data[pref_columns]
from sklearn.tree import DecisionTreeRegressor

# Define
house_price_model = DecisionTreeRegressor()

# Fit
house_price_model.fit(X, y)

# Predict
# print("Predict sales parice for the following houeses:")
# print(X.head())
# print("Predictions are:")
# print(house_price_model.predict(X.head()))


# Evaluate : (using MAE method)
from sklearn.metrics import mean_absolute_error

predicted_prices = house_price_model.predict(X)
mae = mean_absolute_error(y, predicted_prices)
# print("Mean Absolute Error is: {}".format(mae))

"""Spllit data to the train and validate, randomly to get 
right validation method"""

from sklearn.model_selection import train_test_split

# Set set random_state will return same result on each call
train_x, val_x, train_y, val_y = train_test_split(X, y, random_state=0)

# Train our model with half bounch of intial data
split_house_price_model = DecisionTreeRegressor()
split_house_price_model.fit(train_x, train_y)

# Use our tained model to calculate second bounch of data
split_predicted_prices = split_house_price_model.predict(val_x)
split_mea = mean_absolute_error(val_y, split_predicted_prices)
# print("Trained Mean Absolute Error is: {}".format(split_mea))


"""Find overfitting or underfitting spot of your model"""
# OVERFITTING: Depth of tree is high, contain's many leafs, 
# each leaf has few samples and leading to accurate less prediction
# UNDERFITTING: Depth of tree is low, Not enough data selected for 
# models, each leaf contain many samples but again leading to 
# accurate less prediction

def get_mae(max_leaf_nodes, predictors_train, predictors_val, targ_train, targ_val):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(predictors_train, targ_train)
    predicts = model.predict(predictors_val)
    return mean_absolute_error(predicts, targ_val)

# for i in [5, 30, 50, 100, 500, 5000]:
#     mae = get_mae(i, train_x, val_x, train_y, val_y)
#     print("Prediction for depth: {}  is: {}".format(i, mae))
# According to the responses, 50 is the best

"""Random forest prediction to by pass overft and underft problem"""
from sklearn.ensemble import RandomForestRegressor

random_forest = RandomForestRegressor()
random_forest.fit(train_x, train_y)
rf_predict = random_forest.predict(val_x)
rf_mea = mean_absolute_error(rf_predict, val_y)
# print("Random forest Mean Absolute Error: {}".format(rf_mea))
# print(data.describe())
test_predict = random_forest.predict(test_data[pref_columns])
my_submission = pd.DataFrame({"Id": test_data.Id, "SalePrice": test_predict})
my_submission.to_csv('home_price_data_submission_test.csv', index=False)
```
