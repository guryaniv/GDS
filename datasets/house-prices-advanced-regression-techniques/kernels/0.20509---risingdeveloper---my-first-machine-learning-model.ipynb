A simple ML model that test between a tree regressor and a forest tree regressor algorithm. In the end , I used a Forest 
tree regressor model as it has a lower MAE than the later. I made predictions for the house 'SalePrice' from the IOWA city. 




```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.ensemble import RandomForestRegressor

#Training data
main_file_path = '../input/train.csv' # this is the path to the Iowa data that you will use
#Test data
test = pd.read_csv('../input/test.csv')

data = pd.read_csv(main_file_path)

#Prediction target: Output value (y)
y = data['SalePrice']

#Predictors/ extracted important features : stored in X
predictor_list = ['LotArea', 'YearBuilt','1stFlrSF','2ndFlrSF','FullBath','BedroomAbvGr', 'TotRmsAbvGrd']
X = data[predictor_list]

# Split Predictors and target data into two sets: Train and Validation set
train_X, val_X, train_y, val_y = train_test_split(X,y, random_state = 0)

#predict the price using Random Forest Regressor
forest_model = RandomForestRegressor()
forest_model.fit(train_X, train_y)

#Pull out the predictor list from the test data
test_X = test[predictor_list]

predicted_prices = forest_model.predict(test_X)
print(predicted_prices)

#submission
my_submission = pd.DataFrame({'Id': test.Id, 'SalePrice': predicted_prices})
my_submission.to_csv('rising_dev_rf_model.csv', index=False)


#some functions created during training to test two models
def get_mae(max_leaf_nodes, predictors_train, predictors_val, targ_train, targ_val):
    '''A function below to calculate the MAE for different values of maximum tree nodes
        so we can pick the optimun max_node number so as to prevent underfitting and overfiting.'''
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(predictors_train, targ_train)
    predicted_val = model.predict(predictors_val)
    mae = mean_absolute_error(targ_val, predicted_val)
    return(mae) 


def get_mae_of_random_forest_model(predictors_train, predictors_val, targ_train, targ_val):
    forest_model = RandomForestRegressor()
    forest_model.fit(predictors_train, targ_train)
    rf_predicted_value = forest_model.predict(predictors_val)
    rf_mae = mean_absolute_error(targ_val, rf_predicted_value)
    return (rf_mae)


# #Using the tree regressor with a maximum nodes of 50 after testing specific values and picking 50 as the optimum value
# my_mae = get_mae(50,train_X, val_X, train_y, val_y)
# print("Using Tree regressor with nodes: %d \t\t Mean Absolute Error: %d" %(max_leaf_nodes, my_mae))

# #Using the random forest model on the same set of data
# rf_mae = get_mae_of_random_forest_model(train_X, val_X, train_y, val_y)
# print("Using Random Forest regressor \t\t Mean Absolute Error: %d" %(rf_mae))
       







```
