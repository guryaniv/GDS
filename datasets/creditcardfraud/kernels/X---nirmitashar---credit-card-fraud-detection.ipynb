Import dependencies

```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load in 
import keras.backend as K
from keras import utils
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from keras.models import Sequential
from keras.layers import Dense,Dropout,Flatten
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from imblearn.over_sampling import SMOTE
np.random.seed(7)
n_features = 30
```

Read Data and define features and labels

```python
df = pd.read_csv("../input/creditcard.csv")
print(df.shape)
label = df["Class"] 
features = df.drop("Class",1)
features.head()
```

Seperate train and test data, then oversample the train data to make up for the imbalance in class

```python
x_train,x_test,y_train,y_test = train_test_split(features,label,test_size = 0.3)
resampler = SMOTE()
x_train,y_train = resampler.fit_sample(x_train,y_train)
```

Use random forest classifier and check accuracy

```python
clf = RandomForestClassifier()
clf.fit(x_train,y_train)
clf.score(x_test,y_test)
```

```python
rf_prediction = clf.predict(features[df["Class"]==1])
def prediction_accuracy(predictions,Class):
    amount_wrong = 0
    for prediction in predictions:
        if prediction != Class:
            amount_wrong = amount_wrong + 1
    print(amount_wrong)
    print(len(predictions))
prediction_accuracy(rf_prediction,1)
```

Make sequential model

```python
model = Sequential()
model.add(Dense(30, input_dim = n_features, activation = "relu"))
model.add(Dense(20, activation = "sigmoid"))
model.add(Dense(1,activation = "sigmoid"))
model.compile(loss="binary_crossentropy",optimizer="rmsprop",metrics=["accuracy"])
```

Train model and check accuracy

```python
model.fit(x_train,y_train,batch_size=500,epochs = 100)
print(model.evaluate(x_test,y_test))
```

```python
Class = 1
nn_predictions = np.round(model.predict(features[df["Class"]==Class].values).reshape(-1))
#print(nn_predictions)
#label[df["Class"]==Class].values
prediction_accuracy(nn_predictions,Class)
model.save("credit_card.h5")
```
