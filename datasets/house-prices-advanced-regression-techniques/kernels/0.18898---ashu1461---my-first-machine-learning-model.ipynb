```python
import pandas as pd

from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

main_file_path = '../input/train.csv'
data = pd.read_csv(main_file_path)

def getPredictionModel(predictors, target):
    model = RandomForestRegressor()
    model.fit(predictors, target)
    return model
    
predictor_cols = ['LotArea', 'OverallQual', 'YearBuilt', 'TotRmsAbvGrd']

trainy = data.SalePrice
trainX = data[predictor_cols]

test = pd.read_csv('../input/test.csv')
testX = test[predictor_cols]

model = getPredictionModel(trainX, trainy)
predictedPrices = model.predict(testX)

submission = pd.DataFrame({'Id': test.Id, 'SalePrice': predictedPrices})
submission.to_csv('submission.csv', index=False)
```
