```python
import pandas as pd
#Import data as given
main_file_path = '../input/train.csv'
data = pd.read_csv(main_file_path)
print('hello world')
```

```python
print(data.describe())
```

```python
# Trying out selecting single column
Price_data = data.SalePrice
print(Price_data.head())
```

```python
Sale_columns = ['LotArea','YearBuilt','1stFlrSF','2ndFlrSF','FullBath','BedroomAbvGr','TotRmsAbvGrd']
data_without_missing = data.dropna(axis=1)
Sale_data = data_without_missing [Sale_columns]
#Why does the describe command only show one column?
print(Sale_data.describe())
```

```python
y = data_without_missing.SalePrice
Predictors = ['LotArea','YearBuilt','1stFlrSF','2ndFlrSF','FullBath','BedroomAbvGr','TotRmsAbvGrd']
X = data_without_missing [Predictors]
```

```python
from sklearn.tree import DecisionTreeRegressor

data_model =  DecisionTreeRegressor()
data_model.fit(X,y)
```

```python
print("The predictor is:")
print(X.head())
print("The prediction is:")
print(data_model.predict(X.head()))
```

```python
# Calculate error, in-sample:
from sklearn.metrics import mean_absolute_error
predicted_home_prices = data_model.predict(X)
mean_absolute_error(y, predicted_home_prices)
```

```python
#Calculate error, between-sets:
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
train_X, val_X, train_y, val_y = train_test_split(X, y,random_state = 0)
Split_model = DecisionTreeRegressor()
Split_model.fit(train_X,train_y)
val_predictions = Split_model.predict(val_X)
print(mean_absolute_error(val_y, val_predictions))
```

```python
#Define a function to get mean absolute error from 
def get_mae(max_leaf_nodes, predictors_train, predictors_val, targ_train, targ_val):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(predictors_train, targ_train)
    preds_val = model.predict(predictors_val)
    mae = mean_absolute_error(targ_val, preds_val)
    return(mae)
```

```python
# compare MAE with differing values of max_leaf_nodes
for max_leaf_nodes in [5, 50, 500, 5000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
```

```python
#Testing out random forest
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

forest_model = RandomForestRegressor()
forest_model.fit(train_X, train_y)
melb_preds = forest_model.predict(val_X)
print(mean_absolute_error(val_y, melb_preds))
```

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor

# Read the data
train = pd.read_csv('../input/train.csv')

# pull data into target (y) and predictors (X)
train_y = train.SalePrice
predictor_cols = ['LotArea', 'OverallQual', 'YearBuilt', 'TotRmsAbvGrd']

# Create training predictors data
train_X = train[predictor_cols]

my_model = RandomForestRegressor()
my_model.fit(train_X, train_y)
```

```python
# Read the test data
test = pd.read_csv('../input/test.csv')
# Treat the test data in the same way as training data. In this case, pull same columns.
test_X = test[predictor_cols]
# Use the model to make predictions
predicted_prices = my_model.predict(test_X)
# We will look at the predicted prices to ensure we have something sensible.
print(predicted_prices)
```

```python
my_submission = pd.DataFrame({'Id': test.Id, 'SalePrice': predicted_prices})
# you could use any filename. We choose submission here
my_submission.to_csv('submission.csv', index=False)
```
