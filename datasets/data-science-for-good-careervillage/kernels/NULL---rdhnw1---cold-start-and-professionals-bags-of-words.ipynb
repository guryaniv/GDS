## Cold Start
When professionals join, we don’t know very much about them. This is a cold start problem and can result in inappropriate questions being directed to them with the result that they become discouraged and disengage.

Cold start problems can be addressed by engagement processes, user interface design and working with meta data.

Here are some recommended strategies, most of which can be implemented with relatively little disruption:

* It would be good if we made it easier for professionals to add tags and perhaps categorise them: eg, sector, skills, experience, education. 

* When professionals join they should be encouraged to go onto the website and answer a few questions. We can use the information from the answer to improve the recommender.

* When a question is presented, the professional only has two options: answer or ignore. When the professional ignores the question, we are losing valuable information. The professional will have spent time reading the question and should be given the opportunity to provide feedback. Not only is this useful to CV, it engages the professional and is good for motivation.  Questions are likely to be ignored because they are either off-topic, too generic or out of the professionals’ experience. Initially, we could have 2 new buttons: “off-topic” and “not for me”. “Off topic” questions could be re-directed for the moderator to review. When a question is marked “not for me”, a supplementary question could be asked: was the question “too generic” or “not in your field”. “Too generic” should go to the moderator for review. 

* In all case (off-topic, too generic and not in your field), the question should be marked so that the professional doesn’t see it again. This is really important as currently revisits to the CV site are discouraging because professionals are faced with a list of questions that they have previously ignored.


## Recommender for Engaged Professionals

We can easily map professionals to questions answered. There are 51k answers and so we can map 51k times for 24k questions and 28k Professionals.

We could train a model with these question/professional pairs. Given a new question the model would predict the professionals who have answered the most similar questions in the past. This is not exactly what is being asked for which is “the professionals most likely to answer” but it is a pretty good proxy. 

There are problems with this approach: about a third of the answers are provided by professionals who have answered 5 or less questions. For a neural network, this is too small when you consider we need training, validation and test sets. There is also the problem of introducing new professionals who have no professional/answer links.

It would be better to have a system which not only improves with a professional’s engagement but works on day one when we have little information about the professional.

The preferred approach is to produce a bag of words for each profession which includes everything we know about the professional: industry, job, tags followed as well as the title, body and tags of all questions they have answered.

Then when a question is asked, a bag of words containing title, body and tags is produced. This bag of words is then compared with all the professionals’ bags of words and the best matches give the list of professionals  to contact. The reason for lumping everything together (title, body and tags) is that students use the interface in different ways. Some ask the question in the title, some put the tags in the body and others don’t bother with tags.

The comparison can be done by finding the cosine between the question and the professionals. Some examples using td can be found [below](#tfidf)

The cosine threshold could be set higher for newer professionals to improve the likelihood of a good match.

When a question is answered the question information (title, body and text) is added to the professionals bag of words. When “not in your field” is selected, this information can be added into a negative bag of words. This would be done comparing each word in the question with the professionals bag of words. When there is a mismatch the word is added to the negative bag of words. No words are added to the professionals bag of words when the professional answers “not in your field”.

More work can be done on experimenting with data cleaning to see if things like stemming, bespoke stops words, no stop words, lemmings and other sematic analysis could improve the match. Also different methods of document comparison can be explored like LDA that others are considering.

Further fine tuning is possible to prioritise professionals with a good word match using quality of previous answers, location and activity data.

It is also possible to produce a bag of words for all the questions and compare with new questions. Students could then get immediate feedback and could be asked if their question was answered or if they still want professionals to be asked. In the end, this option could become a chat bot. The executive would have to decide if this was a preferred approach by balancing the benefit of immediacy against the lack of an individual human interaction.

Re-engagement strategies could be developed to alert professional that have not answered for a while to a question where the cosine match is high.

On relaunch it would be good to contact all the professionals that have shown interest in the past and deliver some good matches.

The key insight that makes this approach useful is the building of a bag of words for the professional so that it can be used to compare with new questions.

### Here is what I have found out so far. Challenge and collaboration is very welcome and should be fun! Please share....

1. Interest in answering questions drops off quickly but  there is weak evidience that those that stay continue to contribute. [Over 25% of answers](#answers_first_day) are provided in the first day of involvement in Career Village. As we need to focus on the early stages of the relationship with the professional, we should not rely on using data from the professionals' answers to guide the targetting.

2. Professionals can access questions from the CV website and so are not restricted to answering emails. Care is therfore needed when using inferred links between answers and professinals to build a recomender.

3. Tags are a working part of the existing CV system with students being familar with the idea and are happy to use them.

4. However there are problems with tags: not all students use them and many are so generic that they are not useful for targetting. Some form of augmentation is needed and that is the essence of the job.

5. Although we shouldn't completely ignore the requirements of the professionals still answering questions after, say 30 days, intial focus should be on these first 30 days. Most membership organisations have "on-boarding" programmes and so it would be useful to think about how an automatic recommender system could help. I have joined CV as a professional and can share some experiences. There are three types of questions (the examples below are not real to retain confidentiality)
            off topic...."how should I fix my internet"
            difficult to answer or understand..."what is the scope of technology in marketing"
            just right..."what kind of things does a web designer do"
As a professional your choices are either to answer or ignore the questions. Ignoring a question leaves a bad feeling and too many of those leads, possibly, to the professional moving away from CV. Ideally professinals new to CV should only get "just right" questions. Can we build a recommender that is that sophisticated or is some mix of human and machine more practical?

6. Are professionals leaving because of inadequate feedback. We all need to know that our efforts, especially voluntary ones, are appreciated. We have asked CV for more information on this.

7. For the professional that stays with CV, there are a number of useful techniques that people have identified: identifying a professional that identified a similar question, identifying a professional with a similar profile to a professsional who answered a similar question.

# Example of using tfidf to compare a question with the professionals' bags of words<a id='tfidf'></a>

[Go to summary](#summary)  

## Looking for professionals with animation experience

First an example of using tfidf...

```python
"""Example of tfidf"""

import nltk, string
from sklearn.feature_extraction.text import TfidfVectorizer

#nltk.download('punkt') # if necessary...


#stemmer = nltk.stem.porter.PorterStemmer()
remove_punctuation_map = dict((ord(char), None) for char in string.punctuation)

#def stem_tokens(tokens):
#    return [stemmer.stem(item) for item in tokens]

'''remove punctuation, lowercase, stem'''
def normalize(text):
    return nltk.word_tokenize(text.lower().translate(remove_punctuation_map))

vectorizer = TfidfVectorizer(tokenizer=normalize, stop_words='english')

def cosine_sim(text1, text2):
    tfidf = vectorizer.fit_transform([text1, text2])
    return ((tfidf * tfidf.T).A)[0,1]


print (cosine_sim('a little bird', 'a little bird'))
print (cosine_sim('a little bird', 'a little bird chirps'))
print (cosine_sim('a little bird', 'a big dog barks'))
```

The example above shows that the cosine  near 1 means that there is a good match between two sets of words. A number near 0 means that there is not a match.

In the example below, all the professionals' bags of words are compared with the question bag of words with an index of 12900. We know from the above that this question is about animation.

I have printed out 2 professional bag of words, one of which has (index = 8) aimation experience and the other (index = 16) doesn't. I print out the cosine for each professional and you can see where there are matches...

```python


import nltk, string
from sklearn.feature_extraction.text import TfidfVectorizer


remove_punctuation_map = dict((ord(char), None) for char in string.punctuation)

'''remove punctuation, lowercase, stem'''
def normalize(text):
    return nltk.word_tokenize(text.lower().translate(remove_punctuation_map))

vectorizer = TfidfVectorizer(tokenizer=normalize, stop_words='english')

def cosine_sim(text1, text2):
    tfidf = vectorizer.fit_transform([text1, text2])
    return ((tfidf * tfidf.T).A)[0,1]

print ('Find professionals to answer this question ')
print ("questions_bow.loc [12900]['bow'])", questions_bow.loc [12900]['bow'])
print ()
print ()

print ('Example of profession with animation experience, cosine = .54')
print("df_p_bow.loc[8,'bow']",df_p_bow.loc[8]['bow'])
print ()

print ('Example of profession without animation experience, cosine = 0.06')
print("df_p_bow.loc[3,'bow']",df_p_bow.loc[3]['bow'])


df_p_bow_len = len(df_p_bow)
p_rank = pd.DataFrame( columns=['id', 'cos'])
#p_rank.loc[0] = 24324,.25

i = 0
while i < df_p_bow_len:
    cos = cosine_sim(df_p_bow.loc[i]['bow'], questions_bow.loc [12900]['bow'])
    p_rank.loc[i] = df_p_bow.loc[i]['professionals_id'],cos

    i += 1
#p_rank = p_rank.sort_values('cos', ascending = False )
p_rank.head(20)
```
