### Fare

```python
train.Fare=train.Fare.fillna(train.Fare.mean())
test.Fare=test.Fare.fillna(train.Fare.mean())
```

### Cabin

```python
train.Cabin=train.Cabin.fillna("unknow")
test.Cabin=test.Cabin.fillna("unknow")
```

### Embarked

```python
train.Embarked=train.Embarked.fillna(train.Embarked.mode()[0])
test.Embarked=test.Embarked.fillna(train.Embarked.mode()[0])
```

### Age
Considering title

```python
train['title']=train.Name.apply(lambda x: x.split('.')[0].split(',')[1].strip())
test['title']=test.Name.apply(lambda x: x.split('.')[0].split(',')[1].strip())
```

```python
newtitles={
    "Capt":       "Officer",
    "Col":        "Officer",
    "Major":      "Officer",
    "Jonkheer":   "Royalty",
    "Don":        "Royalty",
    "Sir" :       "Royalty",
    "Dr":         "Officer",
    "Rev":        "Officer",
    "the Countess":"Royalty",
    "Dona":       "Royalty",
    "Mme":        "Mrs",
    "Mlle":       "Miss",
    "Ms":         "Mrs",
    "Mr" :        "Mr",
    "Mrs" :       "Mrs",
    "Miss" :      "Miss",
    "Master" :    "Master",
    "Lady" :      "Royalty"}
```

```python
train['title']=train.title.map(newtitles)
test['title']=test.title.map(newtitles)
```

```python
train.groupby(['title','Sex']).Age.mean()
```

```python
def newage (cols):
    title=cols[0]
    Sex=cols[1]
    Age=cols[2]
    if pd.isnull(Age):
        if title=='Master' and Sex=="male":
            return 4.57
        elif title=='Miss' and Sex=='female':
            return 21.8
        elif title=='Mr' and Sex=='male': 
            return 32.37
        elif title=='Mrs' and Sex=='female':
            return 35.72
        elif title=='Officer' and Sex=='female':
            return 49
        elif title=='Officer' and Sex=='male':
            return 46.56
        elif title=='Royalty' and Sex=='female':
            return 40.50
        else:
            return 42.33
    else:
        return Age 
```

```python
train.Age=train[['title','Sex','Age']].apply(newage, axis=1)
test.Age=test[['title','Sex','Age']].apply(newage, axis=1)
```

### 1.4 - Prepare for model

```python
#droping features I won't use in model
#train.drop(['PassengerId','Name','Ticket','SibSp','Parch','Ticket','Cabin']
train.drop(['PassengerId','Name','Ticket','SibSp','Parch','Ticket','Cabin'],axis=1,inplace=True)
test.drop(['PassengerId','Name','Ticket','SibSp','Parch','Ticket','Cabin'],axis=1,inplace=True)
```

```python
titanic=pd.concat([train, test], sort=False)
```

```python
titanic=pd.get_dummies(titanic)
```

```python
train=titanic[:len_train]
test=titanic[len_train:]
```

```python
# Lets change type of target
train.Survived=train.Survived.astype('int')
train.Survived.dtype
```

```python
xtrain=train.drop("Survived",axis=1)
ytrain=train['Survived']
xtest=test.drop("Survived", axis=1)
```

# 2 - Model

```python
seed=1
np.random.seed(seed)
```

```python
# sequential model allows to define model layers-to-layers (for most problems)
model=Sequential()

#fully connected layers (dense class)
#input layer with 20 neurons, input dimensions, initialization method and activation function
model.add(Dense(20,input_dim=17, init='uniform', activation='relu'))

#hidden layer with 8 neurons, initialization method and activation function
model.add(Dense(8, init='uniform', activation='relu'))

#hidden layer with 5 neurons, initialization method and activation function
model.add(Dense(5, init='uniform', activation='relu'))

#output layer with initialization method and activation fuction sigmoid for output between 0 an 1
model.add(Dense(1,init='uniform', activation='sigmoid'))
```

```python
#compiling specifying loss function (logarithmic loss) and optimizer (gradient descent)
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(xtrain, ytrain, epochs=100, batch_size=10, verbose=1)
```

# 3 - Submission

```python
pred=model.predict(xtest)
```

```python
preds=[round(x[0]) for x in pred]
```

```python
output=pd.DataFrame({'PassengerId':test2['PassengerId'],'Survived':preds})
```

```python
output.Survived=output.Survived.astype('int')
```

```python
output.to_csv('submission.csv', index=False)
```
