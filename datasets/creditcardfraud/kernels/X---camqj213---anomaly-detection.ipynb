```python
corr = data.corr()
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
```

```python
X = np.array(data)
np.random.shuffle(X)

y = X[:,30]
X = X[:,0:30]
print(y)
print(X)
```

```python
standard_scaler = StandardScaler()
scaled_X = standard_scaler.fit_transform(X)
print(scaled_X)
```

```python
n_samples = len(scaled_X)
partition = int(.9 * n_samples)
X_train = scaled_X[:partition]
y_train = y[:partition]
X_test = scaled_X[partition:]
y_test = y[partition:]
```

```python
logistic = linear_model.LogisticRegression()
print('LogisticRegression score: %f'
      % logistic.fit(X_train, y_train).score(X_test, y_test))
```

```python
predicted = logistic.predict(X_test)
print(predicted)
print(len(predicted), len(y_test))
true_positive, false_positive, true_negative, false_negative = 0, 0, 0, 0
for p, i in zip(predicted, y_test):
    if p == 0 and i == 0:
        true_negative += 1
    elif p == 0 and i == 1:
        false_negative += 1
    elif p == 1 and i == 1:
        true_positive += 1
    elif p == 1 and i == 0:
        false_positive += 1
print(true_positive, false_positive)
print(false_negative, true_negative)
```

The confusion matrix shows a very bad F1 score. All transactions are recognized as normal transactions.
