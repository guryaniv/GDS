# Introduction

### First machine learning model from Kaggle
We are going to explore the differences in terms of results between decision tree regressor and random forest regressor.

```python
import pandas as pd

# Load data 
main_file_path = '../input/house-prices-advanced-regression-techniques/' # this is the path to the Iowa data that you will use

train_data = pd.read_csv(main_file_path + 'train.csv')
test_data = pd.read_csv(main_file_path + 'test.csv')
```

### Preparing data

```python
train_y = train_data.SalePrice # Prediction target
predictors = ['LotArea',
'YearBuilt',
'1stFlrSF',
'2ndFlrSF',
'FullBath',
'BedroomAbvGr',
'TotRmsAbvGrd']

train_X = train_data[predictors] # subset of data composed only of predictor columns
```

### Train the models

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

# Training with default parameters for both the models

tree_model = DecisionTreeRegressor()
tree_model.fit(train_X, train_y)

forest_model = RandomForestRegressor()
forest_model.fit(train_X, train_y)
```

### Predicting results

```python
test_X = test_data[predictors]

tree_predictions = tree_model.predict(test_X)
forest_predictions = forest_model.predict(test_X)
```

### Submission

```python
my_submission = pd.DataFrame({'Id': test_data.Id, 'SalePrice': forest_predictions})
my_submission.to_csv('first_submission.csv', index=False)
```
