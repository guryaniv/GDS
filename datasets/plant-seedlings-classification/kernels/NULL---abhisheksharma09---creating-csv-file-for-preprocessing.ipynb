```python
import pandas as pd
import os
```

```python
PATH = "../input/train/"
```

```python
species = os.listdir(f"{PATH}")
```

```python
specie = []
filename = []
for file in species:
    for images in os.listdir(f"{PATH}{file}"):
        specie.append(file)
        image, _ = images.split('.')
        filename.append(image)
data = list(zip(specie, filename)) 
```

```python
df = pd.DataFrame(data, columns = ['specie', 'filename'])
```

```python
df.to_csv('train.csv', index = None)
```
