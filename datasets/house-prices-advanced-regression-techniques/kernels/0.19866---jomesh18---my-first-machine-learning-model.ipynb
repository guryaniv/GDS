```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split

def get_mae(max_leaf_nodes,train_X,val_X,train_y,val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state = 0)
    model.fit(train_X,train_y)
    predict_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, predict_val)
    return mae

train_X, val_X, train_y, val_y = train_test_split(X,y,random_state=0)

for max_leaf_nodes in [5,50,500,5000]:
    my_mae = get_mae(max_leaf_nodes,train_X,val_X,train_y,val_y)
    print('For max leaf node {} mean absolute error is {}'.format(max_leaf_nodes,my_mae))
    
```

```python
from sklearn.ensemble import RandomForestRegressor

forest_model = RandomForestRegressor()
forest_model.fit(train_X, train_y)
iowa_preds = forest_model.predict(val_X)
print(mean_absolute_error(val_y, iowa_preds))
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor

train_data = pd.read_csv('../input/house-prices-advanced-regression-techniques/train.csv')
#print(train_data.head())
train_y = train_data.SalePrice
train_X = train_data[['LotArea',
'YearBuilt',
'1stFlrSF',
'2ndFlrSF',
'FullBath',
'BedroomAbvGr',
'TotRmsAbvGrd']]

forest_model = RandomForestRegressor()
forest_model.fit(train_X, train_y)

test_data = pd.read_csv('../input/house-prices-advanced-regression-techniques/test.csv')
test_X = test_data[['LotArea',
'YearBuilt',
'1stFlrSF',
'2ndFlrSF',
'FullBath',
'BedroomAbvGr',
'TotRmsAbvGrd']]

my_pred = forest_model.predict(test_X)
my_submission = pd.DataFrame({'Id': test_data.Id, 'SalePrice': my_pred})
my_submission.to_csv('submission.csv', index = False)
```
