# Reading input files

```python
df = pd.read_csv("../input/train.csv")
```

```python
df['sum'] = df['toxic'] + df["severe_toxic"] + df['obscene'] + df['threat'] + df['insult'] + df['identity_hate']
```

```python
c_df = df[df['sum']!=0]
```

```python
c_df.drop(['id','comment_text','sum'],axis=1,inplace=True)
```

```python
c_df.corr()
```
