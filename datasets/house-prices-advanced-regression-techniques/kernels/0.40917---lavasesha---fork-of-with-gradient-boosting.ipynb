## Lets train the models using the Diffrent models from train split

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
X=TT_train1.drop('SalePrice',axis=1)
y=TT_train1[['SalePrice']]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=7)
lm=LinearRegression()
lm.fit(X_train,y_train)

model_score_test = lm.score(X_test, y_test)
model_Score_train= lm.score(X_train,y_train)

print("Model train score is ",model_Score_train)
print("Model test score is",model_score_test)
```

```python
predictions=lm.predict(X_test)
train_pred=lm.predict(X_train)
plt.scatter(y_test,predictions)
```

```python
print('Train RMSE:', np.sqrt(metrics.mean_squared_error(y_train, train_pred)))
print('Test RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
# Standardize the dataset
modelsN = []
modelsN.append(('LR', LinearRegression()))
modelsN.append(('KNN', KNeighborsRegressor(n_neighbors=2)))
modelsN.append(('CART', DecisionTreeRegressor(max_depth=3)))
modelsN.append(('AB', AdaBoostRegressor()))
modelsN.append(('RF',RandomForestRegressor()))
modelsN.append(('GB',GradientBoostingRegressor()))
```

```python
#without scaling the inputs into uniform scale
results = []
names = []
for name, model in modelsN:
    model.fit(X_train, y_train)
    results.append(model.score(X_test , y_test))
    names.append(name)
    predicted_train=model.predict(X_train)
    predicted_validation=model.predict(X_test)
    print('\nTrain score of {} is: {}'.format(name, model.score(X_train, y_train)))
    print('\nTest score of {} is: {}'.format(name, model.score(X_test , y_test)))
    print('MAE of {} is :{}'.format(name,metrics.mean_absolute_error(y_test, predicted_validation)))
    print('MSE of {} is {}'.format(name,metrics.mean_squared_error(y_test, predicted_validation)))
    print('RMSE of {} is {}'.format(name,np.sqrt(metrics.mean_squared_error(y_test, predicted_validation))))
```
