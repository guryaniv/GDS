```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt

# sampleSub = pd.read_csv("../input/sample_submission.csv")
train = pd.read_csv("../input/train.csv")
# test = pd.read_csv("../input/test.csv")

train.head(5)
```

```python
# fig 1: histogram - kills
kills = np.array( pd.Series(train['kills']).sort_values(ascending=False, kind='mergesort') )
# selecting players have kills of differnt range to compare them in a histogram
_10kills = kills[np.where(kills > 10)]
_20kills = kills[np.where(kills > 20)]
_30kills = kills[np.where(kills > 30)]

# make figure
fig_1 = plt.figure(num="Figure 1 - Kills vs. Players Histogram", figsize=(15, 9), dpi=90)

ax_0 = plt.subplot2grid((3, 3), (0, 0), rowspan=3, colspan=2 )
ax_1 = plt.subplot2grid((3, 3), (0, 2), rowspan=1, colspan=1 )
ax_2 = plt.subplot2grid((3, 3), (1, 2), rowspan=1, colspan=1 )
ax_3 = plt.subplot2grid((3, 3), (2, 2), rowspan=1, colspan=1 )


ax_0.set_title("Kills vs. Players")
ax_0.set_xlabel("Kills")
ax_0.set_ylabel("Killers(All)")
ax_0.grid()
ax_0.hist(kills, bins=kills[0], alpha=0.7, rwidth=0.9, color='k')

ax_1.set_title("Kills vs. Players")
ax_1.set_ylabel("Killers (kills >10)")
ax_1.hist(_10kills, bins=_10kills[0], alpha=0.7, rwidth=0.9, color='g')

ax_2.set_ylabel("Killers (kills >20)")
ax_2.hist(_20kills, bins=_20kills[0], alpha=0.7, rwidth=0.9, color='b')

ax_3.set_ylabel("Killers (kills >40)")
ax_3.set_xlabel("Kills")
ax_3.hist(_30kills, bins=_30kills[0], alpha=0.7, rwidth=0.9, color='r')

plt.show()
```
