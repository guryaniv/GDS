# Introduction of CNN 

## Load Dataset

```python
import pandas as pd
```

```python
train = pd.read_csv("../input/train.csv")
test = pd.read_csv("../input/test.csv")
```

```python
Y = train['label']
X = train.drop(labels=["label"], axis=1)
```

## Preprocess the data

```python
X = X / 255.0
test = test / 255.0
```

```python
from sklearn.model_selection import train_test_split
random_seed=0
X_train, X_val, Y_train, Y_val = train_test_split(X, Y, test_size=0.1, random_state=random_seed)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
import tensorflow as tf
from tensorflow import keras
from keras import layers
from keras import models
```

```python
plt.figure()
Xshow = X_train.values.reshape(-1,28,28,1)
plt.imshow(Xshow[0][:,:,0])
plt.colorbar()
plt.grid(False)
```

## Build the model

### Simple Convnet

```python
model = models.Sequential()
model.add(layers.SeparableConv2D(32,(3,3), activation="elu", input_shape=(28,28,1)))
model.add(layers.MaxPooling2D((2,2)))
model.add(layers.SeparableConv2D(64,(3,3), activation="elu"))
model.add(layers.MaxPooling2D((2,2)))
model.add(layers.SeparableConv2D(64,(3,3), activation="elu"))
```

```python
model.summary()
```

### Convnet + Classifier

```python
model.add(layers.Flatten())
model.add(layers.Dropout(0.5))
model.add(layers.Dense(64, activation="elu"))
model.add(layers.Dense(10, activation="softmax"))
```

```python
model.summary()
```

### Compile the model

```python
model.compile(optimizer='adam', 
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

### Train the model

```python
X_val = X_val.values.reshape(-1,28,28,1)
X_train = X_train.values.reshape(-1,28,28,1)
test = test.values.reshape(-1,28,28,1)
X = X.values.reshape(-1,28,28,1)
```

```python
from keras.utils.np_utils import to_categorical
Y_train = to_categorical(Y_train, num_classes = 10)
Y = to_categorical(Y, num_classes = 10)
```

```python
model.fit(X, Y, epochs=20)
```

```python
Y_val = to_categorical(Y_val, num_classes = 10)
```

```python
test_loss, test_acc = model.evaluate(X_val, Y_val)
print("Test Accuracy: {}".format(test_acc))
```

```python
predictions = model.predict(test)
predictions = np.argmax(predictions, axis=1)
```

```python
submission = pd.read_csv('../input/sample_submission.csv')
submission["Label"] = predictions
```
