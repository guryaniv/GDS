```python
import pandas as pd
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split

main_file_path = '../input/house-prices-advanced-regression-techniques/train.csv'
data = pd.read_csv(main_file_path)

y = data.SalePrice

analyzed_columns = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X = data[analyzed_columns]

train_X, test_X, train_y, test_y = train_test_split(X.values, y.values, test_size=0.25)

# print(train_X.describe())

xgb_model = XGBRegressor(n_estimators=1000, learning_rate=0.05)
xgb_model.fit(train_X, train_y, early_stopping_rounds=5, eval_set=[(test_X, test_y)], verbose=False)

# verify
verify_file_path = '../input/house-prices-advanced-regression-techniques/test.csv' 
verify_data = pd.read_csv(verify_file_path)


verify_X = verify_data[analyzed_columns]
predicted_prices = xgb_model.predict(verify_X.values)

print('')
print(predicted_prices)

my_submission = pd.DataFrame({'Id': verify_data.Id, 'SalePrice': predicted_prices})

# you could use any filename. We choose submission here
my_submission.to_csv('submission.csv', index=False)

```
