```python
#Default Arguments
def f(a,b=1,c=2):
    y =a+b+c
    return y 
print(f(5))
print(f(5,4,3))
```

```python
#Flexible Arguments
def f(*args):
    for i in args:
        print(i)
f(1)
    
```

```python
def f(**kwargs):
    for key, value in kwargs.items():
        print(key, " ", value)
        f(country = 'Spain', capital = 'Madrid', Popoulation = 1234567)
        print(key, " ", value)
```
