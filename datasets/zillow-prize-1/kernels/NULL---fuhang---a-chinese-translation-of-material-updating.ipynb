Considering the gaining popularity of kaggle competition in China, where a growing number of university student even high school student are participating, I think it would be better if there is a Chinese Translation of some of the material in the input data.  



So I use google translate (i knew some chinese partcipants are blocked by GFW) to translate some(may be not all) English material into Chinese one , and correct the translation according to my own understanding... 

Hoping that novices who can't read English well can understand the competition better.
