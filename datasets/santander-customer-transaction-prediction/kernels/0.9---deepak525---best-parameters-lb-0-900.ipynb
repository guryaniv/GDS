```python
%%time
train = pd.read_csv('../input/train.csv')
```

```python
test_df = pd.read_csv('../input/test.csv')
```

```python
X = train.drop(["ID_code", "target"], axis=1)
Y = train["target"]
X_test = test_df.drop(["ID_code"], axis=1)
```

```python
X_test.head()
```

```python
n_fold = 15
folds = StratifiedKFold(n_splits=n_fold, shuffle=True, random_state=42)
```

```python
params = {'num_leaves': 13,
         'min_data_in_leaf': 80,
          'min_sum_hessian_in_leaf': 10.0,
         'objective': 'binary',
          'boost_from_average': False,
         'max_depth': -1,
         'learning_rate': 0.0123,
         'boosting': 'gbdt',
         'bagging_freq': 5,
         'tree_learner': "serial",
         'bagging_fraction': 0.331,
         'feature_fraction': 0.045,
         #'reg_alpha': 1.738,
         #'reg_lambda': 4.99,
         'random_state': 42,
         'metric': 'auc',
         #'min_gain_to_split': 0.01077313523861969,
         #'min_child_weight': 19.428902804238373,
         'num_threads': 8}
```

```python
prediction = np.zeros(len(X_test))
oof = np.zeros(len(X))
for fold_n, (train_index, valid_index) in enumerate(folds.split(X,Y)):
    print('Fold', fold_n, 'started at', time.ctime())
    X_train, X_valid = X.iloc[train_index], X.iloc[valid_index]
    y_train, y_valid = Y.iloc[train_index], Y.iloc[valid_index]
    
    train_data = lgb.Dataset(X_train, label=y_train)
    valid_data = lgb.Dataset(X_valid, label=y_valid)
        
    model = lgb.train(params,train_data,num_boost_round=900000,
                    valid_sets = [train_data, valid_data],verbose_eval=3000,early_stopping_rounds = 3000)
    oof[valid_index] = model.predict(X.iloc[valid_index], num_iteration=model.best_iteration)
            
    #y_pred_valid = model.predict(X_valid)
    prediction += model.predict(X_test, num_iteration=model.best_iteration)/15
print("CV score: {:<8.5f}".format(roc_auc_score(Y, oof)))
```

```python
sub = pd.DataFrame({"ID_code": test_df.ID_code.values})
sub["target"] = prediction
sub.to_csv("submission.csv", index=False)
```
