### Let's measure performance of adding two numbers in python/cython.

```python
%load_ext Cython
import numpy as np
```

```python
def add(x, y):
    return x + y
```

```python
%timeit add(1, 2)
```

Let's try adding integers in cython.

```python
%%cython
cpdef int add_ints(int x, int y):
    return x + y 
```

```python
%timeit add_ints(1, 2)
```

Does faster than in python equivalent.
Adding two floats in cython:

```python
%%cython
cpdef float add_floats(float x, float y):
    return x + y 
```

```python
%timeit add_floats(1.0, 2.0)
```

Works a bit slower than dealing with integers and faster than pythonic version.
What if we pass not floats to *add_floats* but integers.

```python
%timeit add_floats(1, 2)
```

Looks like python wastes some time on conversion and it works worse than the original pythonic way.
