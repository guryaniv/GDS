```python
import dicom # read the dicom files
import os # do directory operations
import pandas as pd # nice for data analysis

data_dir = '../input/sample_images/'
patients = os.listdir(data_dir)
labels_df = pd.read_csv('../input/stage1_labels.csv',index_col=0)

 # test
#print(os.listdir(data_dir))
#for labels in labels_df:
 #   print(labels)
#for files in patients:
 #    print(files)

print(len(labels_df))
print(len(patients))
labels_df.head()
```

```python
for patient in patients[:1]:
    #print(patient)
    label = labels_df.get_value(patient, 'cancer')
    #print(label)
    path = data_dir + patient
    #print(path)
    slices = [dicom.read_file(path + '/' + s) for s in os.listdir(path)]
    slices.sort(key = lambda x: float(x.ImagePositionPatient[2]))
    #print(len(slices), label)
    #print(slices[0])
    print(len(slices), slices[0].pixel_array.shape)
    
```

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

for patient in patients[:1]:
    label = labels_df.get_value(patient, 'cancer')
    path = data_dir + patient
    slices = [dicom.read_file(path + '/' + s) for s in os.listdir(path)]
    slices.sort(key = lambda x: float(x.ImagePositionPatient[2]))
    plt.imshow(slices[0].pixel_array)
    plt.show()
    
```

```python
IMG_PX_SIZE = 150

for patient in patients[:1]:
    label = labels_df.get_value(patient, 'cancer')
    path = data_dir + patient
    slices = [dicom.read_file(path + '/' + s) for s in os.listdir(path)]
    slices.sort(key = lambda x: float(x.ImagePositionPatient[2]))
    
    fig = plt.figure()
    for num,each_slice in enumerate(slices[:10]):
        y = fig.add_subplot(3,4,num+1)
        new_image = cv2.resize(np.numpy(each_slice.pixel_array),(IMG_PX_SIZE,IMG_PX_SIZE))
        y.imshow(new_image)
    plt.show()
```
