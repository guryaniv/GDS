```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
X_train = pd.read_csv('../input/X_train.csv')
y_train = pd.read_csv('../input/y_train.csv')
sample_data = pd.read_csv('../input/sample_submission.csv')
X_test_data = pd.read_csv('../input/X_test.csv')
```

```python
X_train.head()
```

```python
#encoding 
le = LabelEncoder()
y_train['surface'] = le.fit_transform(y_train['surface'])
```

```python
X_train = X_train.groupby('series_id').apply(np.mean)
```

```python
clf = RandomForestClassifier()
clf.fit(X_train.drop(['measurement_number','series_id','row_id'],axis = 1),y_train['surface'][X_train['series_id']])
```

```python
X_test_data = X_test_data.groupby('series_id').apply(np.mean)
```

```python
pred = clf.predict(X_test_data.drop(['measurement_number','series_id','row_id'],axis = 1))
```

```python
pred = le.inverse_transform(pred)
```

```python

sample_data['surface'] = pred
```
