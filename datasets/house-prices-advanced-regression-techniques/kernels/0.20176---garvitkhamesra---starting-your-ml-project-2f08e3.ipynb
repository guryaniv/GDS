```python
# Setting up pandas and data
import pandas as pd
dataset_file_path = '../input/house-prices-advanced-regression-techniques/train.csv'
houses_data = pd.read_csv(dataset_file_path)
df = pd.DataFrame(houses_data)
```

```python
#print(df.columns)
house_price = df.SalePrice
print(house_price.head())
```

```python
data_of_two_columns = ['LotArea', 'LotFrontage']
df_for_two = df[data_of_two_columns]
print(df_for_two.describe())
```

```python
# Using pridictors and models to predict prices
y = df.SalePrice
pridictors = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X = df[pridictors]

from sklearn.tree import DecisionTreeRegressor
df_model = DecisionTreeRegressor()
df_model.fit(X, y)
prediction_price = df_model.predict(X)
print(prediction_price)
```

```python
# Model Validation
from sklearn.metrics import mean_absolute_error
predictions = df_model.predict(X)
print(mean_absolute_error(y, predictions))

from sklearn.model_selection import train_test_split
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
df_test_model = DecisionTreeRegressor()
df_test_model.fit(train_X, train_y)
val_pridictors = df_test_model.predict(val_X)
print(mean_absolute_error(val_y, val_pridictors))
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor

def get_mae(max_leaf_nodes, predictors_train, predictors_val, targ_train, targ_val):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(predictors_train, targ_train)
    preds_val = model.predict(predictors_val)
    mae = mean_absolute_error(targ_val, preds_val)
    return(mae)
```

```python
# compare MAE with differing values of max_leaf_nodes
for max_leaf_nodes in [5, 45, 500, 5000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

model_random_forest = RandomForestRegressor()
model_rf_fit = model_random_forest.fit(train_X, train_y)
model_rf_pred = model_random_forest.predict(val_X)
print(mean_absolute_error(val_y, model_rf_pred))
```
