```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import Imputer

main_file_path = '../input/train.csv'
train_house = pd.read_csv(main_file_path)
train_house.shape
```

```python
test_file_path = '../input/test.csv'
test_house = pd.read_csv (test_file_path)
test_house.shape
```

```python
#checking, finding and counting all columns with NANs or missing values
print(train_house.isnull().sum())
```

```python
train_house.info()
```

```python
#so we have 43 object, 3 floats and 35 integers
#take 1: i exclude ll the non-numerical features
data_house = [train_house, test_house]
for x in data_house:
    numeric_train_house = train_house.select_dtypes(exclude = ['object'])
    numeric_test_house = test_house.select_dtypes(exclude = ['object'])
print(numeric_train_house.head())
```

```python
print(train_house.isnull().sum())
#dropping olumns with missing values
cols_with_missing = [cols for cols in numeric_train_house.columns
                    if numeric_train_house[cols].isnull().any()]
reduce_train_house = numeric_train_house.drop(cols_with_missing, axis = 1)
reduce_test_house = numeric_test_house.drop(cols_with_missing, axis = 1)
```

```python
'''#filling missing values
from sklearn.preprocessing import Imputer
#Now i take care of the missing numerical values
my_imputer = Imputer()
data_house = [train_house, test_house]
for x in data_house:
    imputed_train_house = my_imputer.fit_transform(numeric_train_house)
    imputed_test_house = my_imputer.fit_transform(numeric_test_house)
print(imputed_train_house)'''
```

```python
#creating training predictors
train = reduce_train_house['SalePrice']
predictors_cols = ['LotArea', 'YearBuilt', 'YearRemodAdd','KitchenAbvGr', 'TotRmsAbvGrd','YrSold']
train_x = reduce_train_house[predictors_cols]
test_x = reduce_test_house[predictors_cols]
```

```python
#define your model
house_model = RandomForestRegressor()
#fit your model
house_model.fit(train_x, train)
```

```python
#using the model to make a prediction
saleprice_prediction = house_model.predict(test_x)
print(saleprice_prediction)
saleprice_prediction.shape
```

```python
house_pred_submission = pd.DataFrame({'Id': test_house.Id, 'SalePrice': saleprice_prediction})
house_pred_submission.to_csv('my_Saleprice_prediction.csv', index = False)
```
