```python
import pandas as pd
melbourne_file_path = '../input/train.csv'
melbourne_data = pd.read_csv(melbourne_file_path)
print(melbourne_data.columns)
```

```python
melbourne_sale_price_data = melbourne_data.SalePrice
print(melbourne_sale_price_data.describe())
```

```python
two_columns_of_interest =  ['YearBuilt','SalePrice']
two_columns_of_interest_data = melbourne_data[two_columns_of_interest]
two_columns_of_interest_data.describe()
```

```python
y = melbourne_data.SalePrice
melbourne_predictors = ['LotArea','YearBuilt','1stFlrSF','2ndFlrSF','FullBath','BedroomAbvGr','TotRmsAbvGrd']
X = melbourne_data[melbourne_predictors]
from sklearn.tree import DecisionTreeRegressor
melbourne_model = DecisionTreeRegressor()
melbourne_model.fit(X,y)
```

```python
print("Making predictions for the following 5 houses:")
print(X.head())
print("The predictions are")
print(melbourne_model.predict(X.head()))
```

```python
from sklearn.metrics import mean_absolute_error
predicted_home_prices = melbourne_model.predict(X)
print(mean_absolute_error(y,predicted_home_prices))
```

```python
from sklearn.model_selection import train_test_split
train_X,val_X,train_Y,val_Y = train_test_split(X,y,random_state=0)
melbourne_model.fit(train_X,train_Y)
predicted_data = melbourne_model.predict(val_X)
print(mean_absolute_error(predicted_data,val_Y))
```

```python
def get_mae(max_leaf_nodes, predictors_train, predictors_val, targ_train, targ_val):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(predictors_train, targ_train)
    preds_val = model.predict(predictors_val)
    mae = mean_absolute_error(targ_val, preds_val)
    return(mae)
for max_leaf_nodes in [5, 50, 500, 5000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_Y, val_Y)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
```

```python
from sklearn.ensemble import RandomForestRegressor
forest_model = RandomForestRegressor()
forest_model.fit(train_X,train_Y)
prediction = forest_model.predict(val_X)
print(mean_absolute_error(prediction,val_Y))
```
