```python
import numpy as np
import pandas as pd 
import os
import gc
print(os.listdir("../input"))
```

```python
train = pd.read_csv('../input/train.csv')
train.shape
```

```python
train.head()
```

```python
train = np.array(train)
X = np.array(train[:,3:-1])
Y = np.array(train[:,-1])
X.shape, Y.shape
```

```python
import gc
del train
gc.collect()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
X = np.concatenate((np.ones((X.shape[0], 1)), X, X**2, X**3, X**4, X**5, X**6), axis=1)
X.shape
gc.collect()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.10)
del X, Y
gc.collect()
```

```python
model = LinearRegression(normalize=True, n_jobs=8)
```

```python
result = model.fit(X_train, Y_train)
```

```python
gc.collect()
```

```python
print("Train Score:",result.score(X_train, Y_train))
print("Test Score:",result.score(X_test, Y_test))
```
