```python
import pandas as pd

main_file_path = '../input/train.csv'
data = pd.read_csv(main_file_path)
data.describe().T
```

```python
data.columns
```

```python
melbourne_price_data = data.SalePrice
melbourne_price_data.head()
```

```python
two_columns_of_data = data[['LotArea','LotFrontage']]
two_columns_of_data.describe()
```

```python
predictors = ['LotArea','YearBuilt','1stFlrSF','2ndFlrSF','FullBath','BedroomAbvGr','TotRmsAbvGrd']
X = data[predictors]
Y = melbourne_price_data
```

```python
from sklearn.tree import DecisionTreeRegressor

iowa_model = DecisionTreeRegressor()
iowa_model.fit(X,Y)
```

```python
iowa_model.predict(X.head())
```

```python
from sklearn.cross_validation import train_test_split
X_train,X_test,y_train, y_test = train_test_split(X,Y,test_size = 0.3, random_state =0)
```

```python
iowa_model = DecisionTreeRegressor()
iowa_model.fit(X_train,y_train)
```

```python
val_prediction = iowa_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
validation_Error = mean_absolute_error(y_test, val_prediction)
print( validation_Error)
```

```python
def get_mae(max_leaf_nodes,predictor_train,targ_train,predictor_test,targ_test):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes,random_state=0)
    model.fit(predictor_train,targ_train)
    pred_val = model.predict(predictor_test)
    my_mae = mean_absolute_error(targ_test,pred_val)
    return my_mae
```

```python
for max_leaf_nodes in [5,50,250, 500, ]:
    my_mae = get_mae(max_leaf_nodes, X_train,y_train,X_test,y_test)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
```

```python
from sklearn.ensemble import RandomForestRegressor
forest_model = RandomForestRegressor()
forest_model.fit(X_train, y_train)
iowa_preds = forest_model.predict(X_test)
print(mean_absolute_error(y_test, iowa_preds))
```

```python
test_data = pd.read_csv('../input/test.csv')
test = test_data[predictors]
predicted_prices = forest_model.predict(test)
```
