Inspired by other kaggle kernels which are just about few lines to get submission file.  I'm trying out a kernel with less than 10 lines of code.  Thank you!

```python
import numpy as np, pandas as pd, lightgbm as lgb, warnings, os
from sklearn.model_selection import StratifiedKFold
param, nfolds, early_stopping_rounds = {'bagging_freq': 5,'learning_rate': 0.0083, 'bagging_fraction': 0.41,'num_leaves': 3,'metric':'auc','boost_from_average':'false','boost': 'gbdt','feature_fraction': 0.0405,'max_depth': -1, 'min_data_in_leaf': 80,'min_sum_hessian_in_leaf': 10.0, 'num_threads': 8,'tree_learner': 'serial','objective': 'binary','verbosity': -10}, 18, 3000
train_df, test_df = pd.read_csv('../input/train.csv', nrows= None), pd.read_csv('../input/test.csv', nrows= None)
features, folds, oof, predictions, test_df_IDs = [c for c in train_df.columns if c not in ['ID_code', 'target']], StratifiedKFold(n_splits=nfolds, shuffle=True, random_state=2319), np.zeros(len(train_df)), np.zeros(len(test_df)),test_df.ID_code.values
for fold_, (trn_idx, val_idx) in enumerate(folds.split(train_df.values, train_df['target'].values)):
    trn_data, val_data = lgb.Dataset(train_df.iloc[trn_idx][features], label=train_df['target'].iloc[trn_idx]), lgb.Dataset(train_df.iloc[val_idx][features], label=train_df['target'].iloc[val_idx])
    clf = lgb.train(param, trn_data, 1000000, valid_sets = [trn_data, val_data], verbose_eval=False, early_stopping_rounds = early_stopping_rounds)
    oof[val_idx], predictions  = clf.predict(train_df.iloc[val_idx][features], num_iteration=clf.best_iteration), predictions + clf.predict(test_df[features], num_iteration=clf.best_iteration) / folds.n_splits
pd.DataFrame({"ID_code": test_df_IDs, 'target':predictions}).to_csv("submission_10lines.csv", index=False), print("submitted file generated with 10 lines code")    
```
