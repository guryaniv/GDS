```python
from sklearn.model_selection import train_test_split

data_predictors = ['1stFlrSF','2ndFlrSF', 'FullBath', 'KitchenAbvGr', 'TotRmsAbvGrd', 'BedroomAbvGr', 'GarageArea', 'LotArea', 
                        'YearBuilt', 'GrLivArea', 'SalePrice']
   

data_to_train = data[data_predictors]

data_to_train = data_to_train.fillna(data.median())

y = data_to_train.SalePrice
x = data_to_train.drop('SalePrice' , axis=1)
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error
model1 = DecisionTreeRegressor(max_leaf_nodes=2, random_state=40)
model1.fit(X_train, y_train)
print("The predictions are")
predicted_price = model1.predict(X_test)
print(list(zip(predicted_price, y_test))[:20])
print(mean_absolute_error(y_test, predicted_price))
new_data_to_train = data_to_train.copy()
new_data_to_train["first_predict"] = model1.predict(data_to_train.drop('SalePrice' , axis=1))


```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
model2 = RandomForestRegressor( random_state=42)

y = new_data_to_train.SalePrice
x = new_data_to_train.drop('SalePrice' , axis=1)
X_train2, X_test2, y_train2, y_test2 = train_test_split(x, y, test_size=0.33, random_state=42)

model2.fit(X_train2, y_train2)
print("The predictions are")
predicted_prices = model2.predict(X_test2)
print(list(zip(predicted_price, y_test2))[:20])
print(mean_absolute_error(y_test2, predicted_prices))
```

```python
test = pd.read_csv('../input/test.csv')
# Treat the test data in the same way as training data. In this case, pull same columns.
predictor_cols = data_predictors.copy()
predictor_cols.remove('SalePrice')
test_X = test[predictor_cols]
test_X = test_X.fillna(test_X.median())

test_X["first_predict"] = model1.predict(test_X)

my_submission = pd.DataFrame({'Id': test.Id, 'SalePrice': model2.predict(test_X)})
my_submission.to_csv('submission.csv', index=False)
```
