# KM Path Algorithm

I was able developed my own technique of using K-Means Clustering as guide in finding an optimized path hence the name **KM Path**. The approach is to group nearest cities in small number of clusters using K-Means algorithm then find  the best path by checking all possible permuations using the the cluster centers. Then each cluster or group will be divided in to small number of clusters and an optimal path will be check for every sub-clusters in sequence using the path derived from main cluster sequence. This will be repeated until one is to one ratio has been achieved.
