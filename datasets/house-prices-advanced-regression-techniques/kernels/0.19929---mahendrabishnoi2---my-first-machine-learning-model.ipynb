```python
import pandas as pd

main_file_path = '../input/house-prices-advanced-regression-techniques/train.csv' # this is the path to the Iowa data that you will use
data = pd.read_csv(main_file_path)

# Run this code block with the control-enter keys on your keyboard. Or click the blue botton on the left
print('Some output from running this cell')
```

```python
print(data.describe())
```

```python
print(data.columns)
```

```python
price = data.SalePrice
print(price.head())
```

```python
columns_of_interest = ['LotArea', 'GrLivArea']
new_df = data[columns_of_interest]
print(new_df.describe())
```

```python
y = data.SalePrice
```

```python
list_of_columns = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X = data[list_of_columns]
```

```python
from sklearn.tree import DecisionTreeRegressor
#Define Model
model = DecisionTreeRegressor()
#fit the model
model.fit(X, y)
```

```python
print(X.head())
print('Predicting prices for above 5 houses')
model.predict(X.head())
```

```python
# predictons_on_training_data
pred_train = model.predict(X)
```

```python
# import mean_absolute_error function from sklearn library
from sklearn.metrics import mean_absolute_error

#calculate MAE on training data and print it in dollars
MAE = mean_absolute_error(y, pred_train)
print('{:.{}f}$'.format(MAE, 2))
```

```python
from sklearn.model_selection import train_test_split

train_x, val_x, train_y, val_y = train_test_split(X, y, random_state=0)
```

```python
#fit the model with training data
model.fit(train_x, train_y)
```

```python
#predict on cross validation data
pred_val = model.predict(val_x)
```

```python
#calculate MAE
MAE_val = mean_absolute_error(val_y, pred_val)
print('MAE on cross validation data: {:.{}f}$'.format(MAE_val, 2))
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor

def get_mae(max_leaf_nodes, predictors_train, predictors_val, targ_train, targ_val):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(predictors_train, targ_train)
    preds_val = model.predict(predictors_val)
    mae = mean_absolute_error(targ_val, preds_val)
    return(mae)
```

```python
# compare MAE with differing values of max_leaf_nodes
for max_leaf_nodes in [5, 50, 500, 5000]:
    my_mae = get_mae(max_leaf_nodes, train_x, val_x, train_y, val_y)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

forest_model = RandomForestRegressor()
forest_model.fit(train_x, train_y)
val_pred = forest_model.predict(val_x)
mae = mean_absolute_error(val_y, val_pred)
print(mae)
```
