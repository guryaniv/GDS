importing the required libraries

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

from tensorflow import keras  # tensorflow is our library, keras on top of it makes it simpler
from keras.optimizers import Adam
```

getting the Dataset

```python
train_data = pd.read_csv('../input/train.csv')
```

### EDA

```python
train_data.head()   # looking to our DataFrame
```

```python
train_data.info()    # we see that they are all integers
```

```python
train_data.describe()   # we see that it needs to be scaled (now it's between 0 and 255 )
```

Labels and Features are splitted

```python
train_y = train_data["label"] 

train_data.drop(["label"],axis=1,inplace=True)

train_x = train_data
```

Transformed to numpy arrays

```python
train_x = train_x.values.reshape(-1,28,28,1)
train_y = train_y.values
```

```python
from keras.utils.np_utils import to_categorical
train_y = to_categorical(train_y)
```

```python
train_y.shape
```

```python
train_x.shape
```

### Scaling

```python
train_x = train_x / 255.0
```

```python
train_x.shape
```

## Keras Model Starts Here

```python
from keras.models import Sequential
from keras.layers import Conv2D,MaxPool2D,AveragePooling2D,Dense,Dropout,Flatten
from keras.preprocessing.image import ImageDataGenerator
```

```python
model = Sequential()

model.add(Conv2D(32, kernel_size = (5,5),padding = 'same',activation ='relu', input_shape = (28,28,1)))
model.add(Conv2D(128,kernel_size=(5,5),padding="same",activation="relu"))
model.add(Conv2D(64,kernel_size=(5,5),padding="same",activation="relu"))
model.add(MaxPool2D(pool_size=(2,2),padding="same"))
model.add(Dropout(0.25))

model.add(Conv2D(64,kernel_size=(5,5),padding="same",activation="relu"))
model.add(Conv2D(32,kernel_size=(5,5),padding="same",activation="relu"))
model.add(MaxPool2D(pool_size=(2,2),padding="same"))
model.add(Dropout(0.25))

model.add(Conv2D(16,kernel_size=(2,2),padding="same",activation="relu"))
model.add(MaxPool2D(pool_size=(2,2),padding="same"))
model.add(Dropout(0.25))

model.add(Flatten())

model.add(Dense(64,activation = "relu"))
model.add(Dense(32,activation = "relu"))

model.add(Dense(10, activation = "softmax"))
```

```python
model.summary()
```

```python
model.compile(Adam(lr=0.0003),loss=keras.losses.categorical_crossentropy,metrics=['accuracy'])
```

```python
datagen = ImageDataGenerator(
       # rotation_range=0.5, 
       # zoom_range = 0.5, 
       # width_shift_range=0.5,  
       # height_shift_range=0.5
)

datagen.fit(train_x)
```

```python
model.fit_generator(datagen.flow(train_x,train_y, batch_size=20),steps_per_epoch=2000,epochs=30)
```

```python
test_x = pd.read_csv('../input/test.csv')

test_x.head(10)
```

```python
test_x = test_x.values.reshape(-1,28,28,1)
```

```python
test_x = test_x / 255.0
```

```python
predictions = model.predict(test_x)
```
