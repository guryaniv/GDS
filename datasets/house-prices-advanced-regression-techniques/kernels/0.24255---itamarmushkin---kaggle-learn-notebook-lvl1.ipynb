```python
import pandas as pd

main_file_path = '../input/train.csv'
data = pd.read_csv(main_file_path)

data.describe()
```

```python
print(data.columns)
data[['LotArea', 'LotFrontage']].describe()
```

```python
from sklearn.tree import DecisionTreeRegressor

y=data.SalePrice
predictors=['LotArea','YearBuilt','YrSold','TotRmsAbvGrd','BedroomAbvGr','PoolArea']
X=data[predictors]
model=DecisionTreeRegressor()
model.fit(X,y)
print('The first model used is a Decision Tree Regressor.')
print('The pice predictions for the first 5 houses are:')
print(model.predict(X.head()))
print('actual price values are:')
print(y.head().values)
```

```python
from sklearn.metrics import mean_absolute_error
naive_mae=mean_absolute_error(model.predict(X),y)
print('the mean absolute error of this model is %d USD.' %naive_mae)

from sklearn.model_selection import train_test_split
train_X, val_X, train_y, val_y = train_test_split(X, y,random_state = 0)
model.fit(train_X,train_y)
tt_mae=mean_absolute_error(model.predict(val_X),val_y)
print('with a train-test split, the mean absolute error is %d USD.' %tt_mae)
```

```python
def get_mae(max_leaf_nodes, predictors_train, predictors_val, targ_train, targ_val):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(predictors_train, targ_train)
    preds_val = model.predict(predictors_val)
    mae = mean_absolute_error(targ_val, preds_val)
    return(mae)

for max_leaf_nodes in [2, 5, 10, 50, 100, 500, 5000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
```

```python
from sklearn.ensemble import RandomForestRegressor
forest_model = RandomForestRegressor()
forest_model.fit(train_X, train_y)
forest_mae=mean_absolute_error(forest_model.predict(val_X),val_y)
print('with a Random Forest, the mean absolute error is %d USD.' %forest_mae)
```

```python
test = pd.read_csv('../input/test.csv')
test_X = test[predictors]
my_submission = pd.DataFrame({'Id': test.Id, 'SalePrice': forest_model.predict(test_X)})
my_submission.to_csv('submission.csv', index=False)
```
