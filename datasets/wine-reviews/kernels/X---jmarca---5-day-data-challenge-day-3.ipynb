```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load in 

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
from scipy import stats, integrate
from scipy.stats import ttest_ind
import matplotlib.pyplot as plt

# Input data files are available in the "../input/" directory.
# For example, running this (by clicking run or pressing Shift+Enter) will list the files in the input directory

from subprocess import check_output
print(check_output(["ls", "../input"]).decode("utf8"))

winedata = pd.read_csv("../input/winemag-data-130k-v2.csv")
print(winedata.describe(include='all'))
```

```python
# drop na values on price dimension
winedata_limited = winedata.dropna(subset=["points","variety","province"])
# limit to less than $100
cheaper_wines = winedata_limited["price"] <= 100
california_wines = winedata_limited[cheaper_wines]["province"] == "California"
pinot_noir = winedata_limited[cheaper_wines]["variety"] == "Pinot Noir"
not_california_wines = winedata_limited[cheaper_wines]["province"] != "California"
not_pinot_noir = winedata_limited[cheaper_wines]["variety"] != "Pinot Noir"
```

```python
sns.distplot(winedata_limited[cheaper_wines & california_wines]["points"], kde = False)
sns.distplot(winedata_limited[cheaper_wines & not_california_wines]["points"], kde = False)

ttest_ind(winedata_limited[cheaper_wines & california_wines]["points"], 
          winedata_limited[cheaper_wines & not_california_wines]["points"], 
          axis=0, equal_var=True)
```

So with a statistic of 17.19 and a p value of 3.24e-66, the two distributions are significant.  What about California Pinot Noir versus all other California wines?


```python
sns.distplot(winedata_limited[cheaper_wines & california_wines & pinot_noir]["points"], kde = False)
sns.distplot(winedata_limited[cheaper_wines & california_wines & not_pinot_noir]["points"], kde = False)

ttest_ind(winedata_limited[cheaper_wines & california_wines & pinot_noir]["points"], 
          winedata_limited[cheaper_wines & california_wines & not_pinot_noir]["points"], 
          axis=0, equal_var=True)

```

Again, the two population's assigned points are significantly different.  This time it is obvious from visual inspection of the plot, as the California pinot noir scores have a peak at about 92, while non-pinot California wines have a peak between 88 and 90 (but as always, that curious dip at 89).****
