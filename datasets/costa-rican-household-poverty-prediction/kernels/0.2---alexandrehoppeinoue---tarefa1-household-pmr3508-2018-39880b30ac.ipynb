# 2.0 - Realizando prediction

A partir da extensa base com apenas dados numéricos e sem valores NaN, podemos realizar a predição de dados a partir do knn facilmente

```python
#Atribuindo atributos e variáveis de classe da base de teste.
Xadult = cleanImputedAdultTrainData.iloc[:, 0:136]
XTestAdult = cleanImputedAdultTestData.iloc[:, :] #no NA
Yadult = cleanImputedAdultTrainData.iloc[:, 136]
```

```python
#Obtendo o classificador da biblioteca sklearn
n = 2;
knn = KNeighborsClassifier(n_neighbors=n)
 
```

```python
#faço validação cruzada
scores = cross_val_score(knn, Xadult, Yadult, cv = 10) #cv = 10 => ten-fold validation
scoresAux = numpy.copy(scores)
#loop for para encontrar o melhor valor de n, aproximadamente
while scores.mean() >= scoresAux.mean():
    scoresAux = numpy.copy(scores)
    n = n + 5
    knn = KNeighborsClassifier(n_neighbors=n)
    scores = cross_val_score(knn, Xadult, Yadult, cv = 10)
scores = numpy.copy(scoresAux)
n = n - 5
# Ajuste fino
while scores.mean() >= scoresAux.mean():
    scoresAux = numpy.copy(scores)
    n = n + 1
    knn = KNeighborsClassifier(n_neighbors=n)
    scores = cross_val_score(knn, Xadult, Yadult, cv = 10)
scores = numpy.copy(scoresAux)
n = n - 1
```

```python
#Valores de n e scores que aproximadamente maximizam a predição com base na validação cruzada
#não ocorre underfitting nem overfitting
print('n: ' + str(n))
print('Scores mean: ' + str(scores.mean()))
```
