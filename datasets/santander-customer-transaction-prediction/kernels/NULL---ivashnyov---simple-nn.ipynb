```python
import pandas as pd
from keras.models import Sequential
from keras import layers
from sklearn.model_selection import train_test_split
```

```python
train = pd.read_csv('../input/train.csv')
test = pd.read_csv('../input/test.csv')
```

```python
train.head()
```

```python
y = train['target']
id_code_train = train['ID_code']
train_ = train.drop(['target', 'ID_code'], axis=1)
id_code_test = test['ID_code']
test_ = test.drop(['ID_code'], axis=1)
```

```python
def save_prediction(prediction, file_name):
    pd.DataFrame({"ID_code":id_code_test,"target":prediction}).to_csv(file_name,index=False,header=True)
```

```python
x_train, x_test, y_train, y_test = train_test_split(train_, y, test_size=0.25, random_state=42)
```

```python
from keras.models import Sequential
from keras import layers

input_dim = train_.shape[1]
print('Input dimension =', input_dim)

model = Sequential()
model.add(layers.Dense(10, kernel_initializer = 'uniform', input_dim=input_dim, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='rmsprop', metrics=['accuracy'])
model.summary()
```

```python
history = model.fit(x_train, y_train, epochs=4,
                    validation_data=(x_test, y_test))
```

```python
prediction = model.predict(test_)
```

```python
save_prediction(prediction[:,0], 'result_nn.csv')
```
