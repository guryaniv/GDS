# Predict House Price

```python
import numpy as np 
import pandas as pd 
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split 
from sklearn import metrics 
import matplotlib.pyplot as plt 
import math 
```

1. Read Data  ../input/kc_house_data.csv

```python
data = pd.read_csv("../input/kc_house_data.csv")
data.head()
```

```python
data.dtypes
```

2. Select different featuress that may affect house prices

```python
X = data[['bedrooms','bathrooms','sqft_living','floors']]
Y = data['price']
```

3. Split dataset to training data and testing data

```python
xtrain, xtest, ytrain, ytest = train_test_split(X, Y, test_size = 1/3, random_state = 0)
```

```python
xtrain = np.asmatrix(xtrain)
xtest = np.asmatrix(xtest)
ytrain = np.ravel(ytrain)
ytest = np.ravel(ytest)
```

4. Relations ( Relation between living size and house price)

```python
plt.scatter(data['sqft_living'], Y)
```

Histogram of living size

```python
data['sqft_living'].hist(bins = 50)
```

5. Training model with linear regression

```python
model = LinearRegression()
model.fit(xtrain, ytrain)
```

```python
pd.DataFrame(list(zip(X.columns, np.transpose(model.coef_))))
```

```python
model.intercept_
```

( Predict price of a house with 3 bedrooms, 2 bathrooms, 2500 sqft, 2 floors)

6. Mean Square Error of the model on training dataset 

```python
pred_train = model.predict(xtrain)
mse_train = ((pred_train - ytrain)**2).sum()/len(ytrain)
mse_train
```

average relative error

```python
(abs(pred_train - ytrain)/ytrain).sum()/len(ytrain)
```

Mean Square Error of the model on testing dataset 

```python
predtest = model.predict(xtest)
((predtest-ytest)*(predtest-ytest)).sum() / len(ytest)
```

```python
(abs(predtest-ytest)/ytest).sum() / len(ytest)
```
