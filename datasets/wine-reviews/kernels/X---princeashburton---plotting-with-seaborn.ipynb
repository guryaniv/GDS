```python
reviews = pd.read_csv("../input/wine-reviews/winemag-data_first150k.csv", index_col=0)
import seaborn as sns
```

# Countplot 

With seaborn we do not have to call the value_counts, .countplot does it for us!

```python
sns.countplot(reviews['points'])
```

# KDE Plot 
"Kernel density estimate" will smooth outliers for us. 

```python
sns.kdeplot(reviews.query('price < 200').price)
```

KDE plots are better than line charts for interval data. However, it is not so great for ordinal categorical data 

```python
reviews[reviews['price'] < 200]['price'].value_counts().sort_index().plot.line()
```

We can also plot 2D data

### Bivariate KDE 

```python
sns.kdeplot(reviews[reviews['price'] < 200].loc[:,['price', 'points']].dropna().sample(5000))
```

### Displot

```python
sns.distplot(reviews['points'], bins=10,kde=False)
```

# Scatterplot and Hexplot

```python
sns.jointplot(x='price', y='points', data=reviews[reviews['price'] < 100])
```

```python
#Hexplot
sns.jointplot(x='price', y='points', data=reviews[reviews['price'] < 100], kind='hex',gridsize=20)
```

# Boxplot and violin plot

```python
df = reviews[reviews.variety.isin(reviews.variety.value_counts().head(5).index)]

sns.boxplot(
     x='variety',
      y='points',
     data=df)
```

boxplots will only work well if interval variables and nominal variables are large, they assume data us roughly normally distributed.

```python
sns.violinplot(x='variety',y='points',
              data =reviews[reviews.variety.isin(reviews.variety.value_counts()[:5].index)]
              )
```

# Which library do you choose? 
Seaborn works well with tidy-data (data normally in a record format).
Pandas is tied directly to the input data and expects your data to be in the right output shape. 

So pandas is great for EDA(Exploratory Data Analysis) while Seaborn will work great for complex visuals.
