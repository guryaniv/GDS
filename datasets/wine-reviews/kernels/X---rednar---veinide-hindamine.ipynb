**Veinide hindamine.** Andmed on saadud erinevatelt someljeedelt, kes hindasid veini 100 punkti skaalal. Kuid esitatud on aga ainult veinid, mis said vähemalt 80 punkti.

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt

df = pd.read_csv("../input/winemag-data_first150k.csv")
```

**Veini hinde ja hinna suhe.** Diagramm, mis vaatleb veini hinna ja punktide suhet.

```python
df.groupby("points")["price"].mean().plot(title="Hinna ja veini punktide suhe)");
```

Diagrammilt on selgelt näha, et mida kallimaks lähevad veinid, seda suuremaks lähevad nende punktid.

**Veinide hinna jaotus.** Järgnev histogramm vaatleb, milline oli kõige populaarsem hinne veinidele ning kui palju, milliseid hindeid jagati.

```python
df["points"].plot.hist(title="Veinide hinde jaotus", grid=True, rwidth=0.2);
```

Histogrammilt on näha, et kõige populaarsem hinne oli 87 ning, et selle said üle 35000 veini. Samuti on näha, et maksimaalseid punkte said vaid väga vähesed veinind.

**Veinide punktid hinna järgi.** Järgnevalt hajuvusdiagrammilt on näha täpsemalt, milliste hindadega veinind, kui palju punkte saavad.

```python
df.plot.scatter("points","price", marker="$ - $", alpha=0.1,s=67,color="purple", title="Veinide punktid hinna järgi");
```

Hajuvusdiagrammilt on näha, et maksimaalsed 100 punkti said väga vähesed veininid, kuid keskmiselt ei ole nad väga palju kallimad, kui veinid, mis on saanud natuke vähem punkte.
