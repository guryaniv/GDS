```python
# Load libraries
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

## Load and transform data

```python
# Load train data
df_train = pd.read_csv('../input/train.csv')
# Load test data
df_test = pd.read_csv('../input/test.csv')
# Convert texts to lists
texts_train = df_train.pop('comment_text').tolist()
texts_test = df_test.pop('comment_text').tolist()
# Add columns to test data
for i in df_train.columns[1:]:
    df_test[i] = 0
```

## LogisticRegression

```python
for i in df_train.columns[1:]:
    clf_pipeline = Pipeline([("vectorizer", CountVectorizer()), 
                             ("transformer", TfidfTransformer()),
                             ("classifier", LogisticRegression())])
    labels = df_train[i].tolist()
    model = clf_pipeline.fit(texts_train, labels)
    x = model.predict_proba(texts_test)
    y = []
    for j in range(0, len(x)): y.append(x[j][1])
    df_test[i] = y
```

## Create the submission file

```python
df_test.to_csv('submission.csv', index=False)
```
