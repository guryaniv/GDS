```python
def gen_cat_feature(df,X):
    test_feat_1 = df["feature_1"]
    encoder_feat_1 = LabelBinarizer()
    encoder_feat_1.fit(test_feat_1)
    test_feat_1_1hot = encoder_feat_1.transform(test_feat_1)
    X = np.concatenate((X,test_feat_1_1hot),axis=1)

    test_feat_2 = df["feature_2"]
    encoder_feat_2 = LabelBinarizer()
    encoder_feat_2.fit(test_feat_2)
    test_feat_2_1hot = encoder_feat_2.transform(test_feat_2)
    X = np.concatenate((X,test_feat_2_1hot),axis=1)

    test_feat_3 = df["feature_3"]
    encoder_feat_3 = LabelBinarizer()
    encoder_feat_3.fit(test_feat_3)
    test_feat_3_1hot = encoder_feat_3.transform(test_feat_3)
    X = np.concatenate((X,test_feat_3_1hot),axis=1)
    return X
```
