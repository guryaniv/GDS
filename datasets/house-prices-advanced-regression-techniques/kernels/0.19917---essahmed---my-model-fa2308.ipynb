```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor

main_file_path = '../input/train.csv'
#read and store data
main_data = pd.read_csv(main_file_path)
#predictors
train_y = main_data.SalePrice
list_cols = ['LotArea','YearBuilt' ,'1stFlrSF' ,'2ndFlrSF','FullBath', 'BedroomAbvGr']
train_X = main_data[list_cols]
#fitting
main_data = RandomForestRegressor()
main_data.fit(train_X, train_y)

#read test data
test_data = pd.read_csv('../input/test.csv')
test_X = test_data[list_cols]
Prdct_prices = main_data.predict(test_X)
print(Prdct_prices)

#submission
submit = pd.DataFrame({'ID': test_data.Id, 'SalePrice': Prdct_prices})
submit.to_csv('Submission file.csv', index = False)
```
