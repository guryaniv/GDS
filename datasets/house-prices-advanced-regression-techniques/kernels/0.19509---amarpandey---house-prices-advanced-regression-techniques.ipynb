```python
iowa_train_file_path = '../input/train.csv'
iowa_data = pd.read_csv(iowa_train_file_path)

iowa_test_file_path = '../input/test.csv'
iowa_data_test = pd.read_csv(iowa_test_file_path)
```

```python
data_without_missing_values = iowa_data.dropna(axis=1)
data_without_missing_values.columns
```

```python
predictor_columns = [
    'LotArea', 
    'YearBuilt', 
    '1stFlrSF', 
    '2ndFlrSF', 
    'FullBath', 
    'BedroomAbvGr', 
    'TotRmsAbvGrd'
]

y = iowa_data.SalePrice
X = iowa_data[predictor_columns]

test_X = iowa_data_test[predictor_columns]
```

```python
from sklearn.ensemble import RandomForestRegressor

forest_model = RandomForestRegressor(n_estimators = 100)

# fit model
forest_model.fit(X, y)

# Predict values using model
forest_sales_price_prediction = forest_model.predict(test_X)

forest_sales_price_prediction_pd = pd.DataFrame({'Id': iowa_data_test.Id, 'SalePrice': forest_sales_price_prediction})

forest_sales_price_prediction_pd.head()
```

```python
forest_sales_price_prediction_pd.to_csv('submission-latest.csv', index=False)
```
