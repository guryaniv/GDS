```python
#Let's First start by finding top 20 poor countries 
lowestGNI1960 = document.query("IndicatorCode == 'NY.GNP.PCAP.CD' & CountryName != list & Year == 1962").sort_values(by = 'Value', ascending = True)[:20]
fig = plt.subplots()
plot =sns.barplot(x='Value', y = "CountryName", palette='husl', data=lowestGNI1960)
plt.xlabel("Average Income [$]")
plt.ylabel("Country")
plt.title('The 20 Countries with Lowest Average Income in 1962')
```

```python

#so nepal was at the bottom of the pack with Rwanda, Malawi, china and Botswana
#Now lets see who are bottom 20 in 2014 
lowestGNI2014=document.query("IndicatorCode=='NY.GNP.PCAP.CD' & CountryName != list & Year == 2014").sort_values(by = 'Value', ascending = True)[:20]
fig = plt.subplots()
plot =sns.barplot(x='Value', y = "CountryName", palette='husl', data=lowestGNI2014)
plt.xlabel("Average Income [$]")
plt.ylabel("Country")
plt.title('The 20 Countries with Lowest Average Income in 2014')
```

```python
# so Nepal made some progrss from 1962 to 2014 but still at the bottom 20. How do china and nepal compare in 2014 ?
gni_nc2014 = lowestGNI2014=document.query("IndicatorCode=='NY.GNP.PCAP.CD' & (CountryName == 'Nepal' | CountryName == 'China' | CountryName== 'Rwanda' | CountryName== 'India')  & Year == 2014" ).sort_values(by = 'Value', ascending = True)[:20]

fig = plt.subplots()
plot =sns.barplot(x='Value', y = "CountryName", palette='husl', data=gni_nc2014)
plt.xlabel("Average Income [$]")
plt.ylabel("Country")
plt.title('Average Income in 2014')
```

That is amazing. Look how nepal and china used to be at the bottom of the pack in 1962. Its amazing to see how china has made  such progress in last 50 yrs. Now lets's see howthe GDP per capita has growed  over the year for these countries  ? 

```python
#Let's see how GDP
gdp_2014 = document.query("IndicatorCode=='NY.GDP.PCAP.KD.ZG' & (CountryName == 'Nepal' | CountryName == 'China' | CountryName== 'Rwanda' | CountryName== 'India')  & Year == 2014" ).sort_values(by = 'Value', ascending = True)[:4]
fig = plt.subplots()
plot =sns.barplot(x='Value', y = "CountryName", palette='husl', data=gdp_2014)
plt.xlabel("GDP per capita growth [annual [%]]")
plt.ylabel("Country")
plt.title('Annual GDP per capita growth in 2014')
```

```python
gdp_1962 = document.query("IndicatorCode=='NY.GDP.PCAP.KD.ZG' & (CountryName == 'Nepal' | CountryName == 'China' | CountryName== 'Rwanda' | CountryName== 'India')  & Year == 1962" ).sort_values(by = 'Value', ascending = False)[:4]
fig = plt.subplots()
plot =sns.barplot(x='Value', y = "CountryName", palette='husl', data=gdp_1962)
plt.xlabel("GDP per capita growth [annual [%]]")
plt.ylabel("Country")
plt.title('Annual GDP per capita growth in 1962')
```

```python
#let's see how GDP per capita has growed over the years in these countries 

gdpgrowth = document.query("IndicatorCode=='NY.GDP.PCAP.KD.ZG'")
gdp_growth_nepal = gdpgrowth.query("IndicatorCode=='NY.GDP.PCAP.KD.ZG' & CountryName=='Nepal' ")
gdp_growth_india = gdpgrowth.query("IndicatorCode=='NY.GDP.PCAP.KD.ZG'& CountryName=='India' ")
gdp_growth_china = gdpgrowth.query("IndicatorCode=='NY.GDP.PCAP.KD.ZG' & CountryName=='China' ")
gdp_growth_rwanda = gdpgrowth.query("IndicatorCode=='NY.GDP.PCAP.KD.ZG' & CountryName=='Rwanda' ")

fig = plt.subplots()
plt.plot(gdp_growth_nepal.Year, gdp_growth_nepal.Value, "r", label="Nepal" )
plt.plot(gdp_growth_india.Year, gdp_growth_india.Value, "b" , label="India")
plt.plot(gdp_growth_china.Year, gdp_growth_china.Value, "g" , label="China")
plt.plot(gdp_growth_rwanda.Year, gdp_growth_rwanda.Value, "cyan", label="Rwanda" )
plt.ylabel("GDP per capita growth [annual [%]]")
plt.xlabel("Year")
plt.title('Annual GDP per capita growth')
plt.legend()
```

```python
#China vs Nepal-- how do their average incomes compare overtime?¶
fig11, ax21 = plt.subplots(figsize = [15,8])
labels_cGNP = []
for key, group in document.query("IndicatorCode == 'NY.GNP.PCAP.CD' & CountryName == ['China', 'India', 'Nepal', 'Rwanda']").groupby(['CountryName']):
    ax21 = group.plot(ax=ax21, kind = "line", x = "Year", y = "Value", title = "Comparing average incomes- China vs Nepal")
    labels_cGNP.append(key)

lines, _ = ax21.get_legend_handles_labels()
ax21.legend(lines, labels_cGNP, loc = 'best')
```

```python
#China vs India vs Nepal vs Rwanda - how their average incomes compare overtime
fig12, axs12 = plt.subplots(figsize = [15,8])
labels_cross3pt2 = []
for key, group in document.query("IndicatorCode == 'NY.GNP.PCAP.CD' & CountryName == ['China', 'India', 'Nepal', 'Rwanda']").groupby(['CountryName']):
    axs12 = group.plot(ax = axs12, kind = "line", x = "Year", y = "Value", title = "Comparing average income- China vs India vs Nepal vs Rwanda")
    labels_cross3pt2.append(key)

lines,_ = axs12.get_legend_handles_labels()
axs12.legend(lines, labels_cross3pt2, loc = 'best')
```

```python

income_share = document.query("CountryName == ['Nepal', 'China', 'India', 'Rwanda'] & Year == 2014 ").groupby("IndicatorCode")
N = 4
print(income_share)
i1 = income_share.get_group('SI.DST.FRST.20')['Value']
i2 = income_share.get_group('SI.DST.02ND.20')['Value']
#print(i1)

i3 = income_share.get_group('SI.DST.03RD.20')['Value']
i4 = income_share.get_group('SI.DST.04TH.20')['Value']
i5 = income_share.get_group('SI.DST.05TH.20')['Value']

f, ax_1 = plt.subplots(1, figsize = (15,8))
ind = np.arange(N)
width = 0.35
p1 = ax_1.bar(ind, i1, width, color = '#404040')
p2 = ax_1.bar(ind, i2, width, color = '#bababa', bottom = i1)
p3 = ax_1.bar(ind, i3, width, color = '#ffffff', bottom = [i+j for i,j in zip(i1,i2)])
p4 = ax_1.bar(ind, i4, width, color = '#f4a582', bottom = [i+j+k for i,j,k in zip(i1,i2,i3)])
p5 = ax_1.bar(ind, i5, width, color = '#ca0020', bottom = [i+j+k+l for i,j,k,l in zip(i1,i2,i3,i4)])
plt.ylabel('Percent', fontsize = 14)
plt.xlabel('Country Name', fontsize = 14)
plt.xticks(ind + (width/2), ('Nepal', 'China', 'India', 'Rwanda'))
plt.title('Examining wealth distributions- Nepal, China, India, and Rwanda', fontsize = 14)
plt.legend((p1[0],p2[0],p3[0],p4[0],p5[0]),('Income Share Lowest 20%', 'Income Share Second 20%', 'Income Share Third 20%', 'Income Share Fourth 20%', 'Income Share Highest 20%'), loc = 'upper right', bbox_to_anchor=(1.3, 0.9))
axes = plt.gca()
axes.set_ylim([0,100]) 
```
