```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor

# file path to ... ?
main_file_path = '../input/train.csv'
# load a CSV file
data = pd.read_csv(main_file_path)

# prediction target saved as "y"
y=data.SalePrice

# predictors saved as "x"
pred=['LotArea','YearBuilt','1stFlrSF','2ndFlrSF','FullBath','BedroomAbvGr','TotRmsAbvGrd']
x=data[pred]

# split data into training and validation data sets
xTrain,xVal,yTrain,yVal = train_test_split(x,y,random_state=0)

# create model and fit data to training data set
model=DecisionTreeRegressor(max_leaf_nodes=50,random_state=0)
model.fit(xTrain,yTrain)

# predict data from val data set
predPrices=model.predict(xVal)
error=mean_absolute_error(yVal,predPrices)

# print data
print("Error:")
print(error)
print("Predition examples:")
print([yVal[1:5],predPrices[1:5]])

# function for evaluating model with various numbers of leaves in tree
def get_mae(max_leaf_nodes, predictors_train, predictors_val, targ_train, targ_val):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(predictors_train, targ_train)
    preds_val = model.predict(predictors_val)
    mae = mean_absolute_error(targ_val, preds_val)
    return(mae)

# loop through data to optimize no of leaves
for max_leaf_nodes in [5, 50, 500, 5000]:
    maeModel=get_mae(max_leaf_nodes,xTrain,xVal,yTrain,yVal)
    print("Max leaf nodes: %d \t\t MAE: %d" %(max_leaf_nodes,maeModel))
    
# use random forest and compare results
modelF=RandomForestRegressor()
modelF.fit(xTrain,yTrain)
predPricesF=modelF.predict(xVal)
errorF=mean_absolute_error(yVal,predPricesF)
print("MAE decision tree: %d \t MAE random forest: %d" %(error,errorF))


```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor

# file path to ... ?
main_file_path = '../input/train.csv'
# load a CSV file
data = pd.read_csv(main_file_path)

# prediction target saved as "y"
y=data.SalePrice

# predictors saved as "x"
pred=['LotArea','YearBuilt','1stFlrSF','2ndFlrSF','FullBath','BedroomAbvGr','TotRmsAbvGrd']
x=data[pred]

# use random forest to create model
modelF=RandomForestRegressor()
modelF.fit(x,y)

# pull in test data and select predictor columns
test = pd.read_csv('../input/test.csv')
testX = test[pred]

# use model to predict house prices & check data
predPrices=modelF.predict(testX)
print(predPrices)

# create submission
submission = pd.DataFrame({'Id': test.Id, 'SalePrice': predPrices})
submission.to_csv('submission1.csv', index=False)


```
