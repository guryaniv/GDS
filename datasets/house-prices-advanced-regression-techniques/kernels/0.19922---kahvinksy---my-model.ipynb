```python
import pandas as pd

main_file_path = '../input/train.csv'
data = pd.read_csv(main_file_path)
print('hello world')
```

```python
# save filepath to variable for easier access
train_file_path = '../input/train.csv'
# read the data and store data in DataFrame titled melbourne_data
train_data = pd.read_csv(train_file_path) 
# print a summary of the data in Melbourne data
print(train_data.describe())
```

```python
# store the series of prices separately as melbourne_price_data.
train_price_data = train_data.SalePrice
# the head command returns the top few lines of data.
print(train_price_data.head())
```

```python
columns_of_interest = ['LotArea', 'TotRmsAbvGrd']
two_columns_of_data = train_data[columns_of_interest]
two_columns_of_data.describe()
```

```python
y = train_data.SalePrice
```

```python
train_predictors = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X = train_data[train_predictors]
```

```python
from sklearn.tree import DecisionTreeRegressor

# Define model
training_model = DecisionTreeRegressor()

# Fit model
training_model.fit(X, y)
```

```python
print("Making predictions for the following 5 houses:")
print(X.head())
print("The predictions are")
print(training_model.predict(X.head()))
```

```python
from sklearn.model_selection import train_test_split
train_X, val_X, train_y, val_y = train_test_split(X, y,random_state = 0)
training_model = DecisionTreeRegressor()
# Fit model
training_model.fit(train_X, train_y)
```

```python
from sklearn.metrics import mean_absolute_error



val_predictions = training_model.predict(val_X)
print(mean_absolute_error(val_y, val_predictions))
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

forest_model = RandomForestRegressor()
forest_model.fit(train_X, train_y)
train_preds = forest_model.predict(val_X)
print(mean_absolute_error(val_y, train_preds))
```

```python
# Read the test data
test = pd.read_csv('../input/test.csv')
# Treat the test data in the same way as training data. In this case, pull same columns.
test_X = test[train_predictors]
# Use the model to make predictions
predicted_prices = forest_model.predict(test_X)
# We will look at the predicted prices to ensure we have something sensible.
print(predicted_prices)
```

```python
my_submission = pd.DataFrame({'Id': test.Id, 'SalePrice': predicted_prices})
# you could use any filename. We choose submission here
my_submission.to_csv('submission.csv', index=False)
```
