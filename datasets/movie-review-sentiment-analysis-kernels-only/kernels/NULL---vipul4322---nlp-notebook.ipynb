# **Importing the libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import nltk
import string
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
%matplotlib inline
```

# Importing the datasets

```python
train_dataset = pd.read_csv('../input/train.tsv',sep="\t")
test_dataset=pd.read_csv('../input/test.tsv',sep="\t")
```

## Importing Stopwords from nltk which will help in cleaning the text

```python
from nltk.corpus import stopwords
stopwords.words('english')
```

```python
test_dataset.shape
```

```python
train_dataset.shape
```

```python
corpus=[]
```

## Running a for loop for cleaning the text
### Things happening in the for loop is

1.   Removing all the stopwords
2.   Converting all the texts into lowercase
3.   Applying stemming operation
4.   Making a list corpus






```python
for i in range(0,156060):
    review=re.sub('[^a-zA-Z]',' ',train_dataset['Phrase'][i])
    review=review.lower()
    review=review.split()
    ps=PorterStemmer()
    review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
test_corpus=[]
```

## Performing the same operation as above but for test dataset

```python
for j in range(0,66292):
    review=re.sub('[^a-zA-Z]',' ',test_dataset['Phrase'][j])
    review=review.lower()
    review=review.split()
    ps1=PorterStemmer()
    review=[ps1.stem(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    test_corpus.append(review)
```

```python
len(corpus)
```

```python
len(test_corpus)
```

## Creating a sparse matrix so that we can apply our classification model

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=1000)
cv1=CountVectorizer(max_features=1000)
```

```python
X=cv.fit_transform(corpus).toarray()
```

```python
Y_new=cv1.fit_transform(test_corpus).toarray()
```

```python
Y_new.shape
```

```python
y=train_dataset.iloc[:,3].values
```

```python
from sklearn.cross_validation import train_test_split
```

## Seperating the dataset into train and test for training the model ang checking the accuracy

```python
x_train, x_test, y_train, y_test=train_test_split(X,y,test_size=0.02,random_state=0)
```

```python
x_test.shape
```

# Using classification model and fitting in to the training dataset
## Here we are using Naive Bayes Classification model

```python
from sklearn.naive_bayes import GaussianNB
classifier=GaussianNB()
classifier.fit(x_train, y_train)
```

## Predicting to check the accuracy

```python
y_pred=classifier.predict(x_test)
```

```python
y_pred.shape
```

## Creating the confusion matrix for calculating the incorrect prediction

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
```
