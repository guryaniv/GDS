# **Linear Regression**

```python

#Initialize Linear Regression to a variable reg
reg=LinearRegression()
```

```python
labels=data['price']
```

```python
conv_dates = [1 if values == 2014 else 0 for values in data.date ]
data['date']=conv_dates
```

```python
#drop columns not used in training
train1 = data.drop(['id', 'price','condition','yr_built','date','long'],axis=1)
```

```python
train1.head()
```

```python
#75%, 25% train, test split
x_train , x_test , y_train , y_test = train_test_split(train1 , labels , test_size = 0.3,random_state =5)
```

```python
train1.info()
```

```python
#Fitting our data training set
reg.fit(x_train,y_train)
```

```python
#Testing our accuracy!
reg.score(x_test,y_test)
```
