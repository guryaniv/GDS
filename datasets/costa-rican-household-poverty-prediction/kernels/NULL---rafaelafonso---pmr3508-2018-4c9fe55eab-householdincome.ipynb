Agora, vamos particularizar determinadas variáveis que desejamos levar em conta no classificador. Além delas, também a variável target.

```python
XYhouse = household[["v2a1","rooms","v14a","v18q","tamviv","escolari","age","female","estadocivil3","Target"]]
```

```python
XYhouse.shape
```

Com essa matriz, retiramos as linhas com missing data. Cerca de 7 mil amostras foram descartadas.

```python
nHouse = XYhouse.dropna()
```

```python
nHouse.shape
```

```python
household.shape
```

Separamos a variável de classe do dataset.

```python
Xhouse = nHouse[["v2a1","rooms","v14a","v18q","tamviv","escolari","age","female","estadocivil3"]]
Yhouse = nHouse.Target
```

Por fim, montamos o classificador kNN com k=45 (valor que obteve melhor acurácia) e realizamos a validação cruzada.

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=45)
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(knn, Xhouse, Yhouse, cv=10)
```

```python
scores
```

A média dos scores da validação cruzada chegou a cerca de 70,56% de acurácia para as variáveis e parâmetro k selecionados.

```python
sum(scores) / len(scores)
```
