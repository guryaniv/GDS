```python
from sklearn.metrics import log_loss
from itertools import combinations
```

```python
y = pd.read_csv('../input/msk-redefining-cancer-treatment/stage1_solution_filtered.csv')
my_solution = pd.read_csv('../input/my-personalized-medicine-stage-1-predictions/temp.csv')
my_solution = my_solution.loc[my_solution.ID.isin(y.ID),:]
y = pd.Series(y.iloc[:, 1:].values.argmax(1) + 1)
```

```python
for num_classes in range(1,10):
    for comb in combinations(range(1,10), num_classes):
        loss = log_loss(y_true = y[y.isin(comb)], y_pred = my_solution.iloc[y.isin(comb).tolist(), 1:], labels = range(1,10))
        print('for combination of classes {} , got log_loss value {}'.format(comb, loss))
```
