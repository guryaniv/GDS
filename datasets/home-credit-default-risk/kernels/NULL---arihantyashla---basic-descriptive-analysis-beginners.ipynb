```python
#  As a beginners we are trying to do some analysis to answer some frequently asked questions
## Libraries used 
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## Analysis of application_train dataset 
application_train = pd.read_csv('../input/application_train.csv',encoding = 'unicode_escape')
bureau = pd.read_csv('../input/bureau.csv',encoding = 'unicode_escape')
```

```python
### Question 1 : How many people repay loan. '1' is defaulter and '0' is non-defaulter
plt.figure(figsize=(3,3))
plt.show(sns.countplot(x = "TARGET", data = application_train))
```

```python
### Question 2 : Family Status.
plt.figure(figsize=(10,3))
plt.show(sns.countplot(x = "NAME_FAMILY_STATUS", hue = "TARGET", data = application_train))
aa = pd.crosstab(application_train['NAME_FAMILY_STATUS'], application_train['TARGET'])
lst = ("non_default", "default")
aa.columns = lst
aa["default"]
```

```python
### Question 3 : Which Income type and Family type Default the Most.
#### Filter the defaulters into a new DF
application_train2 = application_train.loc[application_train['TARGET'] == 1]
plt.figure(figsize = (15,4))
plt.show(sns.countplot(x = "NAME_FAMILY_STATUS", hue = "NAME_INCOME_TYPE" , data=application_train2, ))
## Married and Working people default the most.
```

```python
### Question 4: Check for Missing Values 
def missing_val(data):
    total_data = data.isnull().sum().sort_values(ascending = False)
    percent_data = (data.isnull().sum()/data.isnull().count()*100).sort_values(ascending = False)
    print(pd.concat([total_data, percent_data], axis=1, keys=['Total', 'Percentage']))
    
missing_val(application_train)
```

```python
### Question 5: Based upon marital status, count of defaulters and repayers
plt.figure(figsize=(6,15))
plt.subplot(311)
plt.title("Total Repayers")
application_train[application_train["TARGET"]==0]["NAME_FAMILY_STATUS"].value_counts().plot.pie(autopct = "%1.0f%%",
                                                                                                startangle=120,colors = sns.color_palette("Set1"),
                                                                                                wedgeprops={"linewidth":2,"edgecolor":"white"},shadow =True)

circ = plt.Circle((0,0),.7,color="white")
plt.gca().add_artist(circ)
plt.subplot(312)
plt.title("Total Defaulters")
application_train[application_train["TARGET"]==1]["NAME_FAMILY_STATUS"].value_counts().plot.pie(autopct = "%1.0f%%",
                                                    startangle=120,colors = sns.color_palette("Set1"),
                                              wedgeprops={"linewidth":2,"edgecolor":"white"},shadow =True)
circ = plt.Circle((0,0),.7,color="white")
plt.gca().add_artist(circ)
plt.tight_layout()
#lt.ylabel("")
```

```python
### Question 6: to know which income type has more credit active 
join_df= pd.merge(application_train,bureau, on="SK_ID_CURR", how="outer")
ax = sns.countplot(y="NAME_INCOME_TYPE", hue="CREDIT_ACTIVE", data=join_df)
```
