```python
import os

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from pandas_summary import DataFrameSummary
from sklearn import metrics
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from time import time

from fastai.imports import *
from fastai.structured import *
```

```python
data_raw = pd.read_csv('../input/TrainAndValid.csv', low_memory=False, parse_dates=["saledate"])
data_raw.head()
```

```python
train_cats(data_raw)
```

```python
add_datepart(data_raw, 'saledate')
```

```python
data_raw.UsageBand = data_raw.UsageBand.astype('category')
data_raw.UsageBand = data_raw.UsageBand.cat.codes
```

```python
data_raw['SalePrice'] = np.log(data_raw['SalePrice'])
data_raw['SalePrice'].head()
```

```python
p_data, y, nas = proc_df(data_raw, 'SalePrice')
```

```python
model = RandomForestRegressor(n_jobs=-1)
t0 = time.time()
model.fit(p_data, y)
print('Training time:', round(time.time() - t0, 3), 's')
model.score(p_data, y)
```
