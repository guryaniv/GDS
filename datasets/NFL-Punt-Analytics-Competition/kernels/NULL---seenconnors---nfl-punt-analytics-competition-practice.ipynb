# Proposed Solution 2: Fair Catch Rule

Our proposed fair catch rule would advance the ball 8 yards on a fair catch.  The punt returner would still be forced to catch the ball and then from the spot of the catch, the ball would be advanced 8 yards to start the drive.  This rule makes sense from a player safety, entertainment, and spirit of the game perspective. 

## *Player Safety*

*After successfully completing a fair catch, the punt returner’s team will begin their offensive drive with a 1st and 10 eight yards ahead of the spot where the fair catch was made. For example, if Deion Sanders of the Atlanta Falcons were to successfully fair catch a punt at his team’s **32 yard line**, Atlanta would start the offensive drive with a 1st and 10 at their own **40 yard line***. 

*The ball will not be advanced if the player attempting the fair catch muffs the fair catch and he and or team later recovers the ball.*

*The punt returner must be given a reasonable opportunity to catch the punt while it is in the air, an interference penalty will be awarded, and the returning team will start the offensive drive 15 yards from the spot of the foul. Furthermore, if the referee deems the hit on the returner egregious, the referee has the authority to eject the tackler from the game.*

By incentivising the punt returner to fair catch, our goal is to greatly reduce the number of concussion that occur.  Fair catches greatly reduce the chances that a concussion will occur on a punt play.  [Concussions occur in .1% of punt plays when the ball if fair caught.  Where as, concussion occur in .7% of punt plays when the ball is not fair caught](https://docs.google.com/spreadsheets/d/1PfLQQhwnABIDlPnxK4NYsyzi51Lx-w_CMXWMKmr7lmc/edit?usp=sharing)

![Image][7]
[7]: http://i65.tinypic.com/opoe14.png




Once this rule is in place, punting strategy may change slightly. One of the ways it could change is that the added 8 yards from a fair catch would incentivise punters to kick the ball out of bounds to avoid the extra 8 yards. 

But we believe this will make punts safer. [There were 640 punts which landed out of bounds in the 2016 and 2017 seasons. Out of the 640 punt plays, there were zero recorded concussions on those plays](https://docs.google.com/spreadsheets/d/1KjGhsxXYR4I-rvc0q9XpFJDVN5z1PjbI0ZrzY_nxSJk/edit?usp=sharing).

Similarly,  if the punter is trying to punt the ball out of bounds and he mishits and the ball stays inbounds, we still believe that punts will become safer. [We found the chance of a concussion occuring to be greatly reduced if the punt is within 15 yards of either sideline](https://docs.google.com/spreadsheets/d/179nfwQ9z9TwXuGGaPbKprhSkIk26AMN4uH86Rw_CRs0/edit?usp=sharing).

Another way that punting strategy could change is that coaches might want to instruct the gunners to get down the field and down the ball before the returner has an opportunity to catch it to avoid the yards, which could lead to more potential speed, contact, and therefore more concussions.

But we believe with the way that we wrote the rule and with the potential for ejection, gunner speed and contact will probably decrease. Basically, with that part of the rule, we are trying to get rid of plays that look like this:

## *Spirit of the Game*

The game of football is sacred and steeped in tradition. The sport is dangerous in nature, so to make the game safer can easily change the foundation of the sport.  We kept that in mind when attempting to reduce severe and irreversible harm to the players that comes in the form of concussions. 

[The average punt return distance in the NFL and FBS over the last three seasons, 2015, 2016, and 2017](https://docs.google.com/spreadsheets/d/1kOVLb4b02JnWHaU1Hgb_nWWdLOK-Y4wR3EG6PwJA_gs/edit?usp=sharing), was just over 8.4 yards for both levels.  This lead us to arriving at 8 yards for the distance after a fair catch.  This is why we believe that that adding an extra 8 yards for a far catch will not only make the game safer, but will also keep the game of football the same.  We believe that 8 yards will not impact the flow of games.

Lastly, we also believe this rule will lead to slightly more scoring, which will make the game more exciting, but it will also preserve the flow of the game.
