```python
X = df.drop(['id', 'date', 'price'], axis=1)
y = df[['price']]

model = LinearRegression()

scores = cross_val_score(model, X, y, cv=10)
print("Cross-validation scores: {}".format(scores))
print("Average cross-validation score: {:.2f}".format(scores.mean()))
```
