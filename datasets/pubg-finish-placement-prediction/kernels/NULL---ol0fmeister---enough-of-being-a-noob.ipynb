**The main purpose of this notebook will be to analyze the effective strategies needed to win a battle royale match, and for that purpose we will mostly be using the seaborn pairplots as i find them pretty easy to implement and understand. So, lets get started by importing the necessary libraries and the datasets provided to us.**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
print("Libraries Imported!!")
```

**Importing the data. We shall only be using the training data for the analysis purpose, the test data would be used later during modelling.**

```python
data = pd.read_csv("../input/train.csv")
data.head(10)
```

**For the purpose of analysing the effective tactics, we will break down the data into 4 different categories:**
1. ** First would be the assault strategies, like directly decimating or killing your enemies.**
2. ** Secondly, we shall analyze how downing or injuring your opponents help your team.**
3. ** Then we shall see how self help strategies effect your winning chances.**
4. ** Lastly, we shall see whether it is better to remain static/camping most of the time or moving about the map. So let's get started.**

Analyzing assault strategies:

```python
plot_1 = sns.pairplot(data.iloc[:10000], vars=['winPoints', 'winPlacePerc', 'assists', 'killPoints', 'kills', 'killStreaks', 'longestKill'], diag_kind = 'kde',
                     kind = 'reg')
```

**From the winPoints(elo/mmr) vs winPlacePerc plot we can see that a lot of players with high elo/mmr(1700-1900) or atleast decent mmr's(1500-1600) are quite low placed on the winPlacePerc line, my first guess would be that these are probably solo queuers, who surely has decent gunskills, but probably fall due to the lack of communication or lack of information regarding where the enemies are, so if you want to increase your win percent, do squad up a lot!, this statement however is based only on the basis of our sample population not the entire one.**

**The winPoints vs killPoints curve clearly indicates the expected fact that highly skilled players usually end up getting the highest kill points/kill based elos**

**The more kills you get the more your winPlacePerc increases, getting more than 3-4 assists for your team drastically increases your winPlacePerc; with the threshold being at 0.4 instead of 0 for our sample population.**

**Also if we inspect the winPlacePerc vs longestKill graph we will notice that the maybe it is a bit safe to engage your enemy from as far out as possible, considering the win percent goes on increasing the further your killshots were from the target, maybe that's why 8x scope is bae.**

Analyzing Injure/Downing strategies:

```python
plot_2 = sns.pairplot(data.iloc[:10000], vars=['winPoints', 'winPlacePerc', 'DBNOs', 'damageDealt'], diag_kind = 'kde')
```

**The curves here follow an expected trend, as you deal more damage or knock out more opponents, you and your team's chances of winning increases as can be seen from the winPlacePerc vs DBNOs/damageDealt curves, also you get more elo as well for knocking out opponents and doing extensive damage, not sure why the curve for winPoints tapers towards a constant value though with increase in damage and DBNO counts.**

Analyzing the effect of health boost, first aids etc!

```python
plot_3 = sns.pairplot(data.iloc[:10000], vars=["winPoints", "winPlacePerc", 'boosts', 'heals', 'revives'], diag_kind = 'kde', kind = 'reg')
```

**The more boosts you use, the more your winPoints and winPlacePerc,  keep healing yourself occasionally as well as that increases your damage taking capacity thereby increasing your chances of winning. Can't say much about how reviving your teammates help, as from the curve it does seem like there is a slight positive correlation between them, but most of the times an enemy might use your downed teammate as a bait, so bear caution!**

Is it really worth camping?????

```python
plot_4 = sns.pairplot(data.iloc[:10000], vars=["winPoints", "winPlacePerc", 'rideDistance', 'swimDistance', 'walkDistance'], 
                      diag_kind = 'kde')
```

**The more you ride, swim, walk...in other words the more you move around, the more your chances of survival increases => more your winning chances increase**
