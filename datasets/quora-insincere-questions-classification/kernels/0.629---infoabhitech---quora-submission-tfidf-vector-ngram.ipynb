```python
df = pd.read_csv("../input/train.csv")
```

```python
df1 = pd.read_csv("../input/test.csv")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vector = TfidfVectorizer(ngram_range=(1,4)).fit(df['question_text'])
```

```python
X_train_vectorized = vector.transform(df['question_text'])
```

```python
y_train = df['target']
```

```python
X_test = df1['question_text']
```

```python
vector_test = TfidfVectorizer(ngram_range=(1,4)).fit(df1['question_text'])
```

```python
X_test_vectorized = vector_test.transform(df1['question_text'])
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(C=6)
```

```python
model.fit(X_train_vectorized, y_train)
```

```python
predictions = model.predict(vector.transform(X_test))
```

```python
preds = predictions.astype(int)
```

```python
prediction = pd.Series(preds,name="prediction")
```

```python
qid = df1['qid']
```

```python
submission_df = pd.concat([qid, prediction], axis=1)
```

```python
submission_df.to_csv("submission.csv", columns = submission_df.columns, index=False)
```
