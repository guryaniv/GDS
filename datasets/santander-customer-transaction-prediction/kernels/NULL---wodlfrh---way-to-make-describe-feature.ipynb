```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
print(os.listdir("../input"))
```

```python
%%time
train_df = pd.read_csv('../input/train.csv')
test_df = pd.read_csv('../input/test.csv')
```

```python
train_df.head()
```

```python
train_df.describe()
```

```python
%%time
train_trans = train_df.iloc[:,2:].T
```

```python
train_trans.head()
```

```python
%%time
train_stat = train_trans.describe()
```

```python
train_stat
```

```python
%%time
train_stat = train_stat.iloc[1:,:].T
```

```python
train_stat.tail()
```

```python
train_stat.shape, train_df.shape
```

```python
train_data = pd.merge(train_df,train_stat,left_index=True, right_index=True)
```

```python
train_data
```

```python
test_df.head()
```

```python
%%time
test_trans = test_df.iloc[:,1:].T
```

```python
%%time
test_stat = test_trans.describe()
```

```python
test_stat = test_stat.iloc[1:,:].T
```

```python
test_data = pd.merge(test_df,test_stat,left_index=True, right_index=True)
```

```python
train_data.to_csv('train_data_stat.csv', index=False)
test_data.to_csv('test_data_stat.csv', index=False)
```
