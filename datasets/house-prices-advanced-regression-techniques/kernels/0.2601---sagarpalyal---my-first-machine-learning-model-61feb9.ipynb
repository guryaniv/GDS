```python
import pandas as pd

main_file_path = '../input/train.csv'
data = pd.read_csv(main_file_path)
print(data.describe())
```

```python
#Print collumns
print(data.columns)
```

```python
#Select single column and print
price_data = data.SalePrice
print(price_data.head())
```

```python
#Select two columns and print describe command
columns_of_interest = ['SaleType', 'SaleCondition']
two_columns_of_data = data[columns_of_interest]
two_columns_of_data.describe()
```

```python
#Select target variale to pridict
y = data.SalePrice
#Create list of the names of predictor
data_predictors = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
#Select new data frame of predictor data and save it with variable name X
X = data[data_predictors]
```

```python
#Create a DecisionTreeRegressorModel and save it to a variable 'my_model'
from sklearn.tree import DecisionTreeRegressor
#Define model
my_model = DecisionTreeRegressor()
#Fit model
my_model.fit(X, y)
```

```python
#Make few predictions with model predict command and print out the predictions
print(my_model.predict(X.head()))
```

```python
#Split data into train and validation parts
from sklearn.model_selection import train_test_split
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
#fit the model with training data
my_model.fit(train_X, train_y)
```

```python
#Make prediction with validation predictor
val_predictions = my_model.predict(val_X)
```

```python
#Calculate MAE between predictins and actual target values for validation data
from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(val_y, val_predictions))
```

```python
# Define utility function to get MAE
def get_mae(max_leaf_nodes, predictors_train, predictors_val, targ_train, targ_val):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(predictors_train, targ_train)
    preds_val = model.predict(predictors_val)
    mae = mean_absolute_error(targ_val, preds_val)
    return(mae)
```

```python
# compare MAE with differing values of max_leaf_nodes
for max_leaf_nodes in [5, 50, 500, 5000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
```

```python
#Run the random forest regressor on your data
from sklearn.ensemble import RandomForestRegressor
forest_model = RandomForestRegressor()
forest_model.fit(train_X, train_y)
val_predictors_forest = forest_model.predict(val_X)
print(mean_absolute_error(val_y, val_predictors_forest))
```
