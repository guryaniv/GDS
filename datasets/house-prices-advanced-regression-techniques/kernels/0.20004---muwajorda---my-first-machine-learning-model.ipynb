```python
#Load the Modules
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor

#load the train dataset
main_file_path = '../input/train.csv'
data = pd.read_csv(main_file_path)
#view all columns
print(data.columns)
```

```python
#select a single  column with '.'
data_price= data.SalePrice
data_price.head()
```

```python
#select two columns
columns_to_see=['SalePrice','OverallQual']
columns_data= data[columns_to_see]
columns_data.describe()
```

```python
#Decision Trees
#target
y=data_price

data_predictors=['LotArea', 'OverallQual', 'YearBuilt', 'TotRmsAbvGrd']
#predictors
X=data[data_predictors]

#fit the model
from sklearn.tree import DecisionTreeRegressor
data_model = DecisionTreeRegressor()
data_model.fit(X,y)

#predict
print("Making predictions :")
print(X.head())
print("The predictions are")
print(data_model.predict(X.head()))
```

```python
#Test and Training
#import models
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

#get random test and training datasets 
train_X, val_X, train_y, val_y = train_test_split(X, y,random_state = 0)
# Define model
Second_model = DecisionTreeRegressor()
# Define model
Second_model.fit(train_X, train_y)
#Validations
val_predictions=Second_model.predict(val_X)
print(mean_absolute_error(val_y,val_predictions))







```

```python
#MAE comparison
def get_mae(max_leaf_nodes, predictors_train, predictors_val, targ_train, targ_val):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(predictors_train, targ_train)
    preds_val = model.predict(predictors_val)
    mae = mean_absolute_error(targ_val, preds_val)
    return(mae)

#Compare MAE with differing values of max_leaf_nodes
for max_leaf_nodes in [5, 50, 500, 5000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))

#The optimal number of leaves is 50
```

```python
#Random Forest
forest_model = RandomForestRegressor()
forest_model.fit(train_X, train_y)
melb_preds = forest_model.predict(val_X)
print(mean_absolute_error(val_y, melb_preds))
```

```python
##Now run the model on the test data set
# Read the test data
test = pd.read_csv('../input/test.csv')
# Treat the test data in the same way as training data. In this case, pull same columns.
test_X = test[data_predictors]
# Use the model to make predictions
predicted_prices = forest_model.predict(test_X)
# We will look at the predicted prices to ensure we have something sensible.
print(predicted_prices)
```

```python
#submit output to a csv file as a dataframe of two columns; test.id and prices
my_submission = pd.DataFrame({'Id': test.Id, 'SalePrice': predicted_prices})
# you could use any filename. We choose submission here
my_submission.to_csv('Jordana_submission2.csv', index=False)
```
