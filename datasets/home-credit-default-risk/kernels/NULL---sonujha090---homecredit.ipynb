## Missing Values

```python
(df.isnull().sum()/len(df)).sort_values(ascending = False)
```

```python
for c in df.columns:
    if df[c].dtype == 'object':
        df[c].fillna(df[c].mode()[0],inplace=True)
    if df[c].dtype == 'float':
        df[c].fillna(df[c].median(),inplace=True)
    if df[c].dtype == 'int64':
        df[c].fillna(df[c].median(),inplace=True)
```

 ## Treating Categorical Values

```python
df.select_dtypes('object').apply(pd.Series.nunique, axis = 0)
```

```python
le = LabelEncoder()
for c in df.columns:
    if df[c].dtype == 'object':
        if len(list(df[c].unique())) <= 2:
            df[c] = le.fit_transform(df[c])
```

### One Hot Encoding

```python
print(df.shape)
df = pd.get_dummies(df)
print(df.shape)
```

### Training and Validation set

```python
from sklearn.model_selection import train_test_split
X_train,X_valid,y_train,y_valid = train_test_split(X,y,test_size=48700)
X_train.shape,X_valid.shape,y_train.shape,y_valid.shape
```

```python
from sklearn.linear_model import LogisticRegression
m = LogisticRegression(C = .0001)
%time m.fit(X_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix,roc_auc_score,classification_report,accuracy_score
y_pred = m.predict(X_valid)
cm = confusion_matrix(y_valid,y_pred)
print(cm)
print(classification_report(y_valid,y_pred))
print(roc_auc_score(y_valid,y_pred))
print(accuracy_score(y_valid,y_pred))
```
