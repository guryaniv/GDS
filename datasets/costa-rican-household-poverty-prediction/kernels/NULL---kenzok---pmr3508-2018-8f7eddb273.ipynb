Importando as bibliotecas:

```python
import pandas as pd
import sklearn
```

Carregando a base de testes:

```python
household = pd.read_csv("../input/train.csv",
        sep=r'\s*,\s*',
        engine='python',
        na_values="NaN")
```

```python
household
```

Listando colunas x nº de missing data na coluna em ordem decrescente:

```python
household.isnull().sum(axis = 0).sort_values(ascending = False)
```

Podemos ver que as colunas "rez_esc", "v18q1", "v2a1" apresentam muitos dados faltantes, dessa forma, optou-se por removê-las.
Assim como a coluna 'id':

```python
Dhousehold = household.drop(columns = ['rez_esc', 'v18q1', 'v2a1', 'Id'])
```

Removendo linhas com missing data:

```python
nDhousehold = Dhousehold.dropna()
nDhousehold.shape
```

Carregando base de testes e fazendo o mesmo:

```python
testhousehold = pd.read_csv("../input/test.csv",
        sep=r'\s*,\s*',
        engine='python',
        na_values="NaN")
```

```python
Dtesthousehold = testhousehold.drop(columns = ['rez_esc', 'v18q1', 'v2a1', 'Id'])
nDtesthousehold = Dtesthousehold.dropna()
```

```python
nDtesthousehold.head()
```

Importando bibliotecas:

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from sklearn import preprocessing
import numpy as np
```

```python
numnDhousehold = nDhousehold.apply(preprocessing.LabelEncoder().fit_transform)
numnDtesthousehold = nDtesthousehold.apply(preprocessing.LabelEncoder().fit_transform)
```

Testes utilizando a todas as colunas:

```python
Xhousehold = numnDhousehold.iloc[:,0:138]
```

```python
Yhousehold = nDhousehold.Target
```

```python
knn = KNeighborsClassifier(n_neighbors = 3)
```

```python
scores = cross_val_score(knn, Xhousehold, Yhousehold, cv = 10)
scores
```

```python
np.average(scores)
```

```python
Xhousehold = Xhousehold.drop(columns = ['age', 'SQBescolari', 'SQBage', 'SQBhogar_total', 'SQBedjefe', 'SQBhogar_nin', 'SQBovercrowding', 'SQBdependency', 'SQBmeaned'])
```

```python
Xhousehold.shape
```

```python
scores = cross_val_score(knn, Xhousehold, Yhousehold, cv = 10)
scores
```

```python
np.average(scores)
```

Aparentemente, a validação cruzada sem a remoção dessas colunas obteve um resultado melhor

Agora, testando diferentes valores de K

```python
Xhousehold = numnDhousehold.iloc[:,0:138]
```

```python
knn = KNeighborsClassifier(n_neighbors = 10)
```

```python
scores = cross_val_score(knn, Xhousehold, Yhousehold, cv = 10)
scores
```

```python
np.average(scores)
```

```python
knn = KNeighborsClassifier(n_neighbors = 20)
scores = cross_val_score(knn, Xhousehold, Yhousehold, cv = 10)
np.average(scores)
```

```python
knn = KNeighborsClassifier(n_neighbors = 30)
scores = cross_val_score(knn, Xhousehold, Yhousehold, cv = 10)
np.average(scores)
```

```python
knn = KNeighborsClassifier(n_neighbors = 40)
scores = cross_val_score(knn, Xhousehold, Yhousehold, cv = 10)
np.average(scores)
```

```python
knn = KNeighborsClassifier(n_neighbors = 45)
scores = cross_val_score(knn, Xhousehold, Yhousehold, cv = 10)
np.average(scores)
```

Aumentando ainda mais o valor de K, a média do score também aumenta, porém a taxas menores, dessa forma, escolheu-se K = 40 como base

```python
knn = KNeighborsClassifier(n_neighbors = 40)
```

```python
knn.fit(Xhousehold, Yhousehold)
```
