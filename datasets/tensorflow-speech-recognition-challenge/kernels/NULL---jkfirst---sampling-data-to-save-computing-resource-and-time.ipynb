```python
import numpy as np
import os
from subprocess import call

# original data directory
dataroot = '../input/train/'
audroot = dataroot + 'audio/'

# sample directory
sampleroot = '../input/sample/'
sample_audroot = sampleroot + 'audio/'

# extract commands by removing '_background_noise_'
commands = os.listdir(audroot)
commands.remove('_background_noise_')
commands = [cmd + '/' for cmd in commands]

# get test and validation file list
testFile = dataroot + 'testing_list.txt'
validFile = dataroot + 'validation_list.txt'
```

In the testing_list.txt, and validation_list.txt, audio files that should not be used for training are listed. I read audio file list from the text file and will use it to get rid of those files from training dataset.

```python
# get a list of files contained in testing_list.txt and validation_list.txt on dictionary
def get_audiolist(listFileName):
    # fill out dictionary with audio files
    audlist = []
    f = open(listFileName, 'r')
    for oneline in f:
        onepath = oneline[:-1]
        audlist.append(onepath)
    return audlist

# get test and validation audio file list
test_flist = get_audiolist(testFile)
valid_flist = get_audiolist(validFile)
```

After that, read all the audio file names under ../input/train/audio/ and concatenate those with each commands ex) 'yes/' + 1111.wav => yes/1111.wav

```python
# get train file list
train_flist = []
for cmd in commands:
    onecmd_flist = os.listdir(audroot + cmd)
    onecmd_flist = [cmd + cmdfile for cmdfile in onecmd_flist]
    train_flist += onecmd_flist
```

Since I have a list of audio files under ../input/train/audio/, I want to get rid of audio files for test and valid dataset.

```python
# remove duplicate from test and validation dataset
train_flist = [tf for tf in train_flist if tf not in test_flist and tf not in valid_flist]
```
