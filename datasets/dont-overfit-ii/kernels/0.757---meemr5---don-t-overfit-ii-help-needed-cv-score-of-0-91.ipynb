```python
import pandas as pd
%matplotlib inline
```

```python
x_train = pd.read_csv("../input/processed-data/New-Xtrain.csv")
y_train = pd.read_csv("../input/dont-overfit-ii/train.csv").target
```

```python
x_test = pd.read_csv("../input/processed-data/New-Xtest.csv")
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
import warnings
warnings.filterwarnings("ignore")
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import BaggingClassifier
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import learning_curve,KFold,cross_val_score

LR = LogisticRegression(class_weight='balanced',solver='liblinear',random_state=90)
n_estimators = range(10,220,20)
kf = KFold(n_splits=5,shuffle=True,random_state=30)
cv_mean = []
cv_std = []
for n in n_estimators:
    clf = BaggingClassifier(base_estimator=LR,max_features=0.8,n_estimators=n,random_state=92)
    cv_scores = cross_val_score(estimator=clf,X=x_train,y=y_train,scoring='roc_auc',cv=kf,n_jobs=-1)
    print("cv_scores for:",n)
    print(cv_scores)
    print("std:",cv_scores.std())
    cv_mean.append(cv_scores.mean())
    cv_std.append(cv_scores.std())
print("\n",pd.Series(cv_mean,index=n_estimators))
print("\n",pd.Series(cv_std,index=n_estimators))
```

```python
pd.Series(cv_mean).max()
```

```python
clf = BaggingClassifier(base_estimator=LR,max_features=0.8,n_estimators=130)
clf.fit(x_train,y_train)
```

```python
submission = pd.read_csv("../input/dont-overfit-ii/sample_submission.csv")
submission['target'] = clf.predict_proba(x_test)
submission.to_csv("submission.csv",index = False)
```
