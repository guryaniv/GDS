cat = train_data.filter(regex='cat')
bin = train_data.filter(regex='bin')

    bin[col] = bin[col].astype('category',copy=False)

#concatenating categorical and binary columns with the non catgeorical columns#

#imputing missing values#


for col in new_df.columns:
    new_df[col].fillna(value=new_df[col].mode()[0], inplace=True)
