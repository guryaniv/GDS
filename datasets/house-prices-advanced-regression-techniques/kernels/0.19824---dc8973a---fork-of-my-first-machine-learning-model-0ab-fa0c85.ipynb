```python
import pandas as pd

main_file_path = '../input/train.csv'
data = pd.read_csv(main_file_path)
data.head()
```

```python
y = data.SalePrice
```

```python
features = [
'LotArea',
'YearBuilt',
'1stFlrSF',
'2ndFlrSF',
'FullBath',
'BedroomAbvGr',
'TotRmsAbvGrd']

X = data[features]
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error


train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
model = DecisionTreeRegressor()
model.fit(train_X, train_y)

val_predictions = model.predict(val_X)
print(mean_absolute_error(val_y, val_predictions))
```

```python
def get_mae(max_leaf_nodes, predictors_train, predictors_val, targ_train, targ_val):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(predictors_train, targ_train)
    preds_val = model.predict(predictors_val)
    mae = mean_absolute_error(targ_val, preds_val)
    return(mae)

for max_leaf_nodes in [ 20,30, 40,50 , 60,70, 80, 90, 100]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print(f'Max leaf nodes: {max_leaf_nodes} \t\t Mean Absolute Error: {my_mae}')
```

```python
from sklearn.ensemble import RandomForestRegressor
forest_model = RandomForestRegressor()
forest_model.fit(train_X, train_y)
preds_new = forest_model.predict(val_X)
print(mean_absolute_error(val_y, preds_new))
```

```python
test = pd.read_csv('../input/test.csv')
test_X = test[features]

test_prediction = forest_model.predict(test_X)
print(type(test_prediction))
```

```python
my_submission = pd.DataFrame({'Id': test.Id, 'SalePrice': test_prediction})
# you could use any filename. We choose submission here
my_submission.to_csv('submission.csv', index=False)
```
