Load the training data and exam the columns.

```python
train = pd.read_csv('../input/train.csv')
train.head()
```

```python
train.columns
```

First try a simple model with just a few features that make sense intuitively.

```python
my_features = ['LotArea','YearBuilt','1stFlrSF','BedroomAbvGr']
X = train[my_features]
y = train.SalePrice
```

Split data into training and test/validation sets

```python
from sklearn.model_selection import train_test_split
Xtrain,Xtest,ytrain,ytest=train_test_split(X, y, random_state=1)
```

Fit a random forest regressor model.

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(Xtrain,ytrain)

```

```python
test = pd.read_csv('../input/test.csv')
test_features = test[my_features]
preds = rf.predict(test_features)
my_submission = pd.DataFrame({'Id':test.Id,'SalePrice':preds})
```
