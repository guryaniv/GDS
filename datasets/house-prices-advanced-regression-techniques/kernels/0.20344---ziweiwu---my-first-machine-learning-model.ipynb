```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.ensemble import RandomForestRegressor


# load data
main_file_path = '../input/house-prices-advanced-regression-techniques/train.csv' # this is the path to the Iowa data that you will use
data = pd.read_csv(main_file_path)

# explore data
print(data.shape)
print(data.describe())
print(data.columns)

melbourne_price = data.SalePrice
print(melbourne_price.head())

columns_of_interest = ['LandSlope', 'Alley']
two_columns_of_data = data[columns_of_interest]
print(two_columns_of_data.describe())

# define X and Y 
Y = data.SalePrice
features = [
    'LotArea',
    'YearBuilt',
    '1stFlrSF',
    '2ndFlrSF',
    'FullBath',
    'BedroomAbvGr',
    'TotRmsAbvGrd']
X = data[features]

# split data into train and test
train_X, test_X, train_Y, test_Y = train_test_split(X, Y, test_size= 0.4, random_state = 0)

# find the optimal parameter
def get_mae(max_leaf_nodes, predictors_train, predictors_val, targ_train, targ_val):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(predictors_train, targ_train)
    preds_val = model.predict(predictors_val)
    mae = mean_absolute_error(targ_val, preds_val)
    return(mae)

def get_mae(max_leaf_nodes, predictors_train, predictors_val, targ_train, targ_val):
    model = RandomForestRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(predictors_train, targ_train)
    preds_val = model.predict(predictors_val)
    mae = mean_absolute_error(targ_val, preds_val)
    return(mae)

# compare MAE with differing values of max_leaf_nodes
for max_leaf_nodes in [5, 50, 100, 200, 300, 500,1000,2000, 5000]:
    my_mae = get_mae(max_leaf_nodes, train_X, test_X, train_Y, test_Y)
    print("Decision Tree Max leaf nodes: %d  \t\t Mean Absolute Error:  %f" %(max_leaf_nodes, my_mae))

    
for max_leaf_nodes in [5, 50, 100, 200, 300, 500,1000,2000, 5000]:
    my_mae = get_mae(max_leaf_nodes, train_X, test_X, train_Y, test_Y)
    print("Random Forest Max leaf nodes: %d  \t\t Mean Absolute Error:  %f" %(max_leaf_nodes, my_mae))


# define the optimal model
melbourne_model = RandomForestRegressor(max_leaf_nodes = 50)
# fit model
main_file_path = '../input/house-prices-advanced-regression-techniques/test.csv' # this is the path to the Iowa data that you will use
test = pd.read_csv(main_file_path)
melbourne_model.fit(train_X, train_Y)
Predictions = melbourne_model.predict(test[features])

# submit the results
my_submission = pd.DataFrame({'Id': test.Id, 'SalePrice': Predictions})
my_submission.to_csv('submission.csv', index=False)
```
