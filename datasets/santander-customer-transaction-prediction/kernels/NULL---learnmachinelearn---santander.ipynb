```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('bmh')
```

```python
df = pd.read_csv('../input/train.csv')
```

```python
df.head()
```

```python
X_train = df.iloc[:, 2:].values.astype('float64')
y_train = df['target'].values
```

```python
from sklearn.preprocessing import StandardScaler

X_train = pd.DataFrame(StandardScaler().fit_transform(X_train))
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(X_train.corr())
plt.colorbar()
plt.title('Correlation Matrix Plot of the Features', fontsize=20, y=1.05);
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import QuantileTransformer


pipeline = make_pipeline(QuantileTransformer(output_distribution='normal'), GaussianNB())
pipeline.fit(X_train, y_train)
```

```python
from sklearn.metrics import roc_curve, auc

plt.figure(figsize=(10, 10))
fpr, tpr, thr = roc_curve(y_train, pipeline.predict_proba(X_train)[:,1])
plt.plot(fpr, tpr)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operator Characteristic Plot', fontsize=20, y=1.05)
auc(fpr, tpr)
```

```python
test = pd.read_csv('../input/test.csv')
```

```python
X_test = test.iloc[:, 1:].values.astype('float64')
submission = pd.read_csv('../input/sample_submission.csv')
submission['target'] = pipeline.predict_proba(X_test)[:,1]
submission.to_csv('submission.csv', index=False)
```
