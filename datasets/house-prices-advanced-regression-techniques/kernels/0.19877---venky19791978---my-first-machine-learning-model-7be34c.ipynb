```python
import pandas as pd

lowa_file_path = '../input/train.csv'
lowa_data = pd.read_csv(lowa_file_path) 
print(lowa_data.columns)
```

```python
# store the series of prices separately as melbourne_price_data.
lowa_price_data = lowa_data.SalePrice
# the head command returns the top few lines of data.
print(lowa_price_data.head())
```

```python
columns_of_interest = ['LotFrontage', 'LotArea']
two_columns_of_data = lowa_data[columns_of_interest]
```

```python
two_columns_of_data.describe()
```

```python
y = lowa_data.SalePrice
```

```python
predictor_cols = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']

X = lowa_data[predictor_cols]
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
iowa_model = DecisionTreeRegressor()
```

```python
iowa_model.fit(X, y)
```

```python
print(iowa_model.predict(X.head()))
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_X, val_X, train_y, val_y = train_test_split(X, y,random_state = 0)
```

```python
iowa_model.fit(train_X, train_y)
```

```python
val_pred = iowa_model.predict(val_X)
```

```python
from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(val_y, val_pred))
```

```python
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return(mae)
```

```python
# compare MAE with differing values of max_leaf_nodes
for max_leaf_nodes in [5, 50, 500, 5000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Max leaf nodes: %d  \t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
```

```python
from sklearn.ensemble import RandomForestRegressor

forest_model = RandomForestRegressor()
forest_model.fit(train_X, train_y)
val_preds = forest_model.predict(val_X)
print(mean_absolute_error(val_y, val_preds))
```

```python
# Read the test data
test = pd.read_csv('../input/test.csv')
# Treat the test data in the same way as training data. In this case, pull same columns.
test_X = test[predictor_cols]
# Use the model to make predictions
predicted_prices = forest_model.predict(test_X)
# We will look at the predicted prices to ensure we have something sensible.
print(predicted_prices)
```

```python
my_submission = pd.DataFrame({'Id': test.Id, 'SalePrice': predicted_prices})
# you could use any filename. We choose submission here
my_submission.to_csv('mymodel.csv', index=False)
```
