```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
import os
print(os.listdir('../input/house-prices-advanced-regression-techniques'))
main_file_path = '../input/house-prices-advanced-regression-techniques/train.csv' # this is the path to the Iowa data that you will use
data = pd.read_csv(main_file_path)

# Run this code block with the control-enter keys on your keyboard. Or click the blue botton on the left
model= RandomForestRegressor()
print(data.columns)
train_y= data.SalePrice
predictor= ['LotArea','2ndFlrSF', 'BedroomAbvGr','TotRmsAbvGrd']
X=data[predictor]
model.fit(X,train_y)

#test 
test_data_file = '../input/house-prices-advanced-regression-techniques/test.csv'
test = pd.read_csv(test_data_file)
print(test.columns)
test_x= test[predictor]
predicted_y = model.predict(test_x)
print(predicted_y)
```

```python
my_submission = pd.DataFrame({'Id': test.Id, 'SalePrice': predicted_y})
# you could use any filename. We choose submission here
my_submission.to_csv('submission.csv', index=False)
```
