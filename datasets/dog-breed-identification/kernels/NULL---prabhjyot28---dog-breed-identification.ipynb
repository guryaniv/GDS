# Intro

Here we are going to use a pretrained Resnt50 model, one of the best model in computer vision. We will use TensorFlow and Keras to implement this model to detect the breeds of dog. We don't have to train this model, because it's already trained on ImageNet data. 

Note - Here we can't build a model from scatch because we have very small dataset of 120 images for training, so we must have to use some pretrained model with all the weights already defined.

## Creating model with pretrained weight file.

```python
from tensorflow.python.keras.applications import ResNet50

my_model = ResNet50(weights='../input/resnet50/resnet50_weights_tf_dim_ordering_tf_kernels.h5')
```

## Reading and Pre-processing images for modelling

```python
import numpy as np
from tensorflow.python.keras.applications.resnet50 import preprocess_input
from tensorflow.python.keras.preprocessing.image import load_img, img_to_array
from os.path import join

image_dir = '../input/dog-breed-identification/train/'
img_paths = [join(image_dir, filename) for filename in 
                           ['0246f44bb123ce3f91c939861eb97fb7.jpg',
                            '84728e78632c0910a69d33f82e62638c.jpg',
                            '8825e914555803f4c67b26593c9d5aff.jpg',
                            '91a5e8db15bccfb6cfa2df5e8b95ec03.jpg']]

image_size = 224

def read_and_prep_images(img_paths, img_height=image_size, img_width=image_size):
    imgs = [load_img(img_path, target_size=(img_height, img_width)) for img_path in img_paths]
    img_array = np.array([img_to_array(img) for img in imgs])
    output = preprocess_input(img_array)
    return(output)

```

## Making Predictions


```python
test_data = read_and_prep_images(img_paths)
preds = my_model.predict(test_data)
```

## Visualizing Predictions

```python
from learntools.deep_learning.decode_predictions import decode_predictions
from IPython.display import Image, display

most_likely_labels = decode_predictions(preds, top=3)

for i, img_path in enumerate(img_paths):
    display(Image(img_path))
    print(most_likely_labels[i])
```
