**Function to find number of missing values**

```python
def show_missing(data):
    missing = data.columns[data.isnull().any()].tolist() 
    return missing
```

**No of Missing values in train_data**

```python
train_data[show_missing(train_data)].isnull().sum() 
```

**No of Missing values in test_data**

```python
test_data[show_missing(test_data)].isnull().sum()
```

**No of missing values in previous applicants**

```python
previous_application[show_missing(previous_application)].isnull().sum()
```

**No of Missing values in point of sale cash balance**

```python
POS_CASH_balance[show_missing(POS_CASH_balance)].isnull().sum()
```

**No of missing values in bureau**

```python
bureau[show_missing(bureau)].isnull().sum()
```
