```python
data_kills = train.copy()
data_kills = data_kills[data_kills['kills']==0]
```

```python
print("{} players ({:.4f}%) have won without a single kill!".format(len(data_kills[data_kills['winPlacePerc']==1]), 100*len(data_kills[data_kills['winPlacePerc']==1])/len(train)))
```

```python
data_nodamage = train[train['damageDealt'] == 0].copy()
```

```python
print("{} players ({:.4f}%) have won without dealing damage!".format(len(data_nodamage[data_nodamage['winPlacePerc']==1]), 100*len(data_nodamage[data_nodamage['winPlacePerc']==1])/len(train)))
```
