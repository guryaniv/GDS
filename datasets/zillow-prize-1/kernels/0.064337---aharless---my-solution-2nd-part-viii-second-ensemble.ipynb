This is the code that combined the results of parts I, VII, III, and IV, and VI, now with some adjustments to make it run on Kaggle, and the option to correct for the bug in the XGBoost model.

```python
ACTUALLY_DO_SOMETHING = True
CORRECT_FOR_BUG = True
```

```python
from tqdm import tqdm
```

```python
import gzip
```

```python
ff = 1.0  # Global fudge factor, based on individual FFs guesstimated from LB probes

weights = [ 0.292, 0.165, 0.147, 0.200, 0.196 ]  
                               
```

```python
print(sum(weights))
```

```python
if CORRECT_FOR_BUG:
    weights[1] /= 10
```

```python
fnames = ('output_dataset2017a.csv',           # 0 StackNet
          'finalxgb20171015_145310.csv',       # 1 My XGBoost tuned with Q4 CV, Oct31 seas
          '2017Q4.csv',                        # 2 Genetic
          'catboost_opt4.csv',                 # 3 CatBoost
          'simplad2017.csv')                   # 4 SimpLAD
output="withSimpLADNoFudge.csv"   # file to generate submissions to
```

```python
dir = "../input/my-zillow-predictions/"
if ACTUALLY_DO_SOMETHING:
    fs=open(output,"w")
    result = []
    ids = []
    first = True
    for w, n in zip(weights, fnames):
        print ("Processing file: ", n)
        f = open(dir+n, 'r')
        for iline, line in tqdm(enumerate(f)):
            if iline==0:
                if first:
                    fs.write(line)
            else:
                i = iline-1
                splits=line.replace("\n","").split(",")
                if first:
                    ids.append(splits[0])
                else:
                    assert(ids[i]==splits[0])
                preds=[]
                for j in range (1,7):
                    preds.append(float(splits[j])*ff*w)
                if first:
                    result.append(preds)
                else:
                    for j in range(0,6):
                        result[i][j] += preds[j]
        f.close()
        first = False        
```

```python
if ACTUALLY_DO_SOMETHING:
    for id, r in zip(ids,result):
        fs.write(id)
        for j in range(6):
            fs.write( "," +str(round(r[j],6) ))
        fs.write("\n")
```

```python
if ACTUALLY_DO_SOMETHING:
    fs.close()
```

```python
if ACTUALLY_DO_SOMETHING:
    f_in = open(output, 'rb')
    f_out = gzip.open(output+'.gz', 'wb')
    f_out.writelines(f_in)
    f_out.close()
    f_in.close()
```
