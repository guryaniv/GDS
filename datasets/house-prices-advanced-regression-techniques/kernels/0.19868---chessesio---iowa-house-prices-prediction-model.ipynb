```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

train_file_path = '../input/house-prices-advanced-regression-techniques/train.csv' # this is the path to the Iowa data that you will use
train_data = pd.read_csv(train_file_path)

predictors = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
train_X = train_data[predictors]
train_y = train_data.SalePrice

test_file_path = '../input/house-prices-advanced-regression-techniques/test.csv'
test_data = pd.read_csv(test_file_path)

val_X = test_data[predictors]
# val_y = test_data.SalePrice

forest_model = RandomForestRegressor()

forest_model.fit(train_X, train_y)
predicted_prices = forest_model.predict(val_X)
# print(mean_absolute_error(val_y, predictions))
```

```python
submission = pd.DataFrame({'Id': test_data.Id, 'SalePrice': predicted_prices})
```

```python
submission.to_csv('iowa_sale_price_prediction.csv', index=False)
```
