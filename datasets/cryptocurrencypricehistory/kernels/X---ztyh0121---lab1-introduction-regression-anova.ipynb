```python
x = np.arange(1, 20, 0.5)
x
```

```python
w = 1 + x / 2
1/w
```

```python
np.random.normal(0, 1, len(x))
```

```python
y = x + w * np.random.normal(0, 1, len(x))
y
```

```python
type(x)
```

```python
type(y)
```

```python
x.shape
```

```python
y.shape
```

```python
x = x.reshape(len(x), 1)
y = y.reshape(len(y), 1)
```

```python
lm2 = linear_model.LinearRegression()
lm2.fit(x, y)
```

```python
print(lm2.coef_)
```

```python
print(lm2.intercept_)
```

```python
print(lm2.score(x, y))
```

```python
plt.scatter(x, y)
plt.plot(x, lm2.predict(x))
```

```python
type(x)
```

```python
import statsmodels.api as sm
lowess = sm.nonparametric.lowess(y.flatten(), x.flatten(), frac=.3)
lowess_x = list(zip(*lowess))[0]
lowess_y = list(zip(*lowess))[1]
help(sm.nonparametric.lowess)
```

```python
lm3 = linear_model.LinearRegression()
lm3.fit(x, y,sample_weight=1/w**2)
```

```python
plt.scatter(x, y)
plt.plot(x, lm2.predict(x))
plt.plot(x, lm3.predict(x))
plt.plot(lowess_x, lowess_y)
```

```python
michel = pd.read_csv('../input/michelson/michelson.csv')
```

```python
from statsmodels.formula.api import ols
import statsmodels
model1 = ols('Speed ~ C(Run)+C(Expt)', data=michel).fit()
aov_table1 = statsmodels.stats.anova.anova_lm(model1, typ=2)
print(aov_table1)
```

```python
model2 = ols('Speed ~ C(Expt)', data=michel).fit()
aov_table2 = statsmodels.stats.anova.anova_lm(model2, typ=2)
print(aov_table2)
```
