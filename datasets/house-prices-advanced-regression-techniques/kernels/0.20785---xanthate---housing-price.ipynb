```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
# Read the data
train_file_path = '../input/train.csv'
train = pd.read_csv(train_file_path)
train.describe()
```

```python
# target value
train_y = train.SalePrice
predictor_cols = ['LotArea', 'YearBuilt', '1stFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']

from sklearn.grid_search import GridSearchCV
estimators = list(range(10, 100))
#min_sample_leaf = list(range(10, 100, 5))
param_grid = dict(n_estimators=estimators)
print(param_grid)
```

```python
# instansiate the grid
rf = RandomForestRegressor()
grid = GridSearchCV(rf, param_grid, cv=10, scoring='neg_mean_squared_error')
# training set
train_X = train[predictor_cols]
#model = RandomForestRegressor()
grid.fit(train_X, train_y)
```

```python
print(grid.best_score_)
print(grid.best_params_)
```

```python
model = RandomForestRegressor(n_estimators=26)
# Read the test data
model.fit(train_X, train_y)
test_file_path = '../input/test.csv'
test = pd.read_csv(test_file_path)
# Select the same features as train data
test_X = test[predictor_cols]
# Make predictions
predict_prices = model.predict(test_X)
print(predict_prices)
```

**Prepare for submission**

```python
submission_file = pd.DataFrame({'Id': test.Id, 'SalePrice': predict_prices})
submission_file.to_csv('submission.csv', index=False)
```
