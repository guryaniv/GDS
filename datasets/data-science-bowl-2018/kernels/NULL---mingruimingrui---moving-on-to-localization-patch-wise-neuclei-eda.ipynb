But before any of that, let's fetch the images and masks shall we?

```python
import os
from functools import reduce

from tqdm import tqdm

import numpy as np
import pandas as pd
from PIL import Image

import matplotlib.pyplot as plt
```

```python
# first get the images and the relavent masks
train_folder = '../input/stage1_train/'
image_ids = os.listdir(train_folder)
data = []

for i_id in tqdm(image_ids, desc='Getting images and masks'):
    img_files = [train_folder + i_id + '/images/' + f for f in os.listdir(train_folder + i_id + '/images/')]
    mask_files = [train_folder + i_id + '/masks/' + f for f in os.listdir(train_folder + i_id + '/masks/')]
    
    assert len(img_files) == 1, 'Is this an error? There should only be 1 img to each id'
    
    new_data = {
        'img': np.array(Image.open(img_files[0])),
        'masks': [np.array(Image.open(m)) for m in mask_files]
    }
    
    data.append(new_data)

nb_masks = reduce(lambda x, y: x + len(y['masks']), data, 0)

print('Fetching of data complete')
print('Total of {} images and {} masks'.format(len(data), nb_masks))
```

A very decently sized dataset! 670 labelled images is a **lot** and there are over 2000 labelled neucleis.
