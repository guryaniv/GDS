## Peek at the Data:
First, let's import some basic libraries and see what data we have in our input folder:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
print(os.listdir("../input"))
```

We have three separate CSV files, containing the codes of the different breeds, different colors and Malaysia states asscoiated. Let's load this data and have a look at it.
First, the breeds encodings:

```python
breeds = pd.read_csv("../input/breed_labels.csv")
print("{}\n\n{} ".format(breeds.head(2), breeds.tail(2)))
print("\nIndex {} is BreedId={}: {}".format(240, breeds.BreedID[240], breeds.BreedName[240]))
```

We have total 307 different breeds with aproppriate encodings, with Type=1 referring to dogs and Type=2 to cats.
<br>Pay attention that BreedID=307 is a "Mixed Breed" with index of 240...
<br>What about the color encodings:

```python
colors = pd.read_csv("../input/color_labels.csv")
print("{}".format(colors.head(7)))
```

Total of seven different colors. Seems ok here. No "Mixed Color" entry!
<br>Let's look on the states:

```python
states = pd.read_csv("../input/state_labels.csv")
print("{}\n\n{} ".format(states.head(2), states.tail(2)))
```

Well, no surprises here as well...
<br>Now, let's load the actual training and test data:

```python
train = pd.read_csv('../input/train/train.csv')
test = pd.read_csv('../input/test/test.csv')
print("Train data has {} entries with {} columns".format(train.shape[0], train.shape[1]))
print("Test data has {} entries with {} columns".format(test.shape[0], test.shape[1]))
```

```python
train.head(1)
```

Let's explore the adoption rates by looking at cats and dogs separately:

```python
plt.figure(figsize=(8,6))
sns.countplot(x='Type', hue='Gender', data=train)
plt.ylabel('Number of Occurrences', fontsize=12)
plt.xlabel('Type', fontsize=12)
plt.grid()
```

```python
plt.figure(figsize=(10,5))
sns.violinplot(x='Type', y='AdoptionSpeed', hue="Gender", data=train)
plt.xlabel('Type: Cat or Dog', fontsize=12)
plt.ylabel('Adoption Speed', fontsize=12)
plt.grid()
plt.show()
```
