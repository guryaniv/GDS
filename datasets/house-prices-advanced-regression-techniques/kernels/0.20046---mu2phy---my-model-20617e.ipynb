```python
import pandas as pd

main_file_path = '../input/train.csv'
data = pd.read_csv(main_file_path)
print(data.describe())
```

```python
Iowa_price_data = data.SalePrice

print(Iowa_price_data.head())
```

```python
columns_of_interest = ["YearBuilt","YrSold"]
two_columns_of_data = data[columns_of_interest]

two_columns_of_data.describe()
```

```python
y=data.SalePrice
```

```python
Iowa_pred = ["LotArea",
             "YearBuilt",
             "1stFlrSF",
             "2ndFlrSF",
             "FullBath",
             "BedroomAbvGr",
             "TotRmsAbvGrd"]

X = data[Iowa_pred]

X.head()
```

```python
from sklearn.tree import DecisionTreeRegressor

Iowa_model = DecisionTreeRegressor()

Iowa_model.fit(X,y)
```

```python
print("The predictions are")
print(Iowa_model.predict(X.head()))
```

```python
from sklearn.metrics import mean_absolute_error

predicted_prices = Iowa_model.predict(X)
mean_absolute_error(y,predicted_prices)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 0)
Iowa_model = DecisionTreeRegressor()
Iowa_model.fit(X_train, y_train)
```

```python
valid_pred = Iowa_model.predict(X_test)
print(mean_absolute_error(y_test, valid_pred))
```

```python
# utility function to help compare MAE scores 
#from different values for max_leaf_nodes

from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor

def get_mae(max_leaf_nodes, predictors_train, predictors_val, targ_train, targ_val):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(predictors_train, targ_train)
    preds_val = model.predict(predictors_val)
    mae = mean_absolute_error(targ_val, preds_val)
    return(mae)
```

```python
for max_leaf_nodes in [5,50,500,5000]:
    my_mae = get_mae(max_leaf_nodes, X_train, X_test, y_train, y_test)
    print("Max leaf nodes: %d \t\t Mean Absolute Error: %d" %(max_leaf_nodes,my_mae))
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

forest_model = RandomForestRegressor()
forest_model.fit(X_train, y_train)
Iowa_pred = forest_model.predict(X_test)
print(mean_absolute_error(y_test, Iowa_pred))
```
