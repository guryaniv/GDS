```python
from fastai.vision import *
from fastai.metrics import error_rate
from fastai import *
from pathlib import Path
```

```python
bs = 16

path = Path('../input/')
path_img = path/'dogs-vs-cats-redux-kernels-edition/train'

fnames = get_image_files(path_img)
fnames[:5]
```

```python
np.random.seed(2)
pat = re.compile(r'(cat|dog).\d+.jpg$')
```

```python
data = ImageDataBunch.from_name_re(path_img, fnames, pat, ds_tfms=get_transforms(), size=224, bs=bs
                                  ).normalize(imagenet_stats)
```

```python
data.show_batch(rows=3, figsize=(7,6))
```

```python
# Fix to enable Resnet to live on Kaggle - creates a writable location for the models
cache_dir = os.path.expanduser(os.path.join('~', '.torch'))
if not os.path.exists(cache_dir):
    os.makedirs(cache_dir)
   # print("directory created :" .cache_dir)
models_dir = os.path.join(cache_dir, 'models')
if not os.path.exists(models_dir):
    os.makedirs(models_dir)
  #  print("directory created :" . cache_dir)

#copying model to writable location
#cd /kaggle/working
shutil.copy("/kaggle/input/resnet101/resnet101.pth", "/tmp/.torch/models/resnet101-5d3b4d8f.pth")
```

```python
print(data.classes)
print(data.c)
```

```python
TMP_PATH = "/tmp/tmp"
MODEL_PATH = "/tmp/model/"

learn = create_cnn(data, models.resnet101, metrics=error_rate, model_dir=MODEL_PATH)
```

```python
learn.model
```

```python
learn.fit_one_cycle(2)
```

```python
interp = ClassificationInterpretation.from_learner(learn)

losses,idxs = interp.top_losses()

len(data.valid_ds)==len(losses)==len(idxs)
```

```python
interp.plot_top_losses(9, figsize=(15,11))
```

```python
interp.plot_confusion_matrix(figsize=(12,12), dpi=60)
```
