```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor

train_X, val_X, train_y, val_y = train_test_split(X, y,random_state = 0)

tree_model = DecisionTreeRegressor()

tree_model.fit(train_X,train_y)

val_predictions = tree_model.predict(val_X)
print(mean_absolute_error(val_y,val_predictions))
```

```python
from sklearn.metrics import mean_absolute_error as MAE
from sklearn.tree import DecisionTreeRegressor as DTR

def get_mae(max_leaf_nodes,predictors_train,predictors_val,targ_train,targ_val):
    model = DTR(max_leaf_nodes = max_leaf_nodes,random_state=0)
    model.fit(predictors_train,targ_train)
    preds_val = model.predict(predictors_val)
    mae = MAE(targ_val,preds_val)
    return(mae)
```

```python
for max_l_n in[5,34,35,36,40,50,60,]:
    my_mae = get_mae(max_l_n,train_X,val_X,train_y,val_y)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_l_n, my_mae))
```
