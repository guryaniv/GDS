```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load in 

import os
from pathlib import Path

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

from skimage.data import imread

import time
import cv2


# Any results you write to the current directory are saved as output.
```

```python
input_path = Path('../input')
train_path = input_path / 'train'
test_path = input_path / 'test'
```

```python
cameras = os.listdir(train_path)

train_images = []
for camera in cameras:
    for fname in sorted(os.listdir(train_path / camera)):
        train_images.append((camera, fname))

train = pd.DataFrame(train_images, columns=['camera', 'fname'])
print(train.shape)
```

```python
test_images = []
for fname in sorted(os.listdir(test_path)):
    test_images.append(fname)

test = pd.DataFrame(test_images, columns=['fname'])
print(test.shape)
```

# Lets see training images

```python
random_images = train.sample(1000)
import matplotlib.pyplot as plt
cnt = 0
for i, r in random_images[:].iterrows():
    if(r['camera'] != 'iPhone-4s'):
        continue
    if cnt == 50:
        break
    cnt+=1
    x = cv2.imread("../input/train/" + 'iPhone-4s' + '/' + r['fname'])
    x = cv2.cvtColor(x, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(8,8))
    plt.imshow(x)
    plt.title(r['fname'])
    plt.show()
```
