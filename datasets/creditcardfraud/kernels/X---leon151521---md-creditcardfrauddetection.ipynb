```python
import pandas as pd
import numpy as np
import seaborn as sns

dataset = pd.read_csv("../input/creditcard.csv")
dataset.head()
```

```python
X = dataset.iloc[:,1:29].values
y = dataset.iloc[:,30].values

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 101)
```

```python
#Feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

# Fitting Logistic Regression
from sklearn.linear_model import LogisticRegression
lr_classifier = LogisticRegression(random_state=42)
lr_classifier.fit(X_train,y_train)
y_lr_pred = lr_classifier.predict(X_test)

# Fitting K-NN to the training set
from sklearn.neighbors import KNeighborsClassifier
knn_classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
knn_classifier.fit(X_train, y_train)
y_knn_pred = knn_classifier.predict(X_test)

#Fitting RandomForest to the training set
from sklearn.ensemble import RandomForestClassifier
rc_classifier = RandomForestClassifier(n_estimators = 5, random_state = 42)
rc_classifier.fit(X_train, y_train)
y_rf_pred = rc_classifier.predict(X_test)

# Fitting DecisionTree to the training set
from sklearn.tree import DecisionTreeClassifier
dt_classifier = DecisionTreeClassifier(max_features='sqrt', random_state = 0)
dt_classifier.fit(X_train, y_train)
y_dt_pred = dt_classifier.predict(X_test)
```

```python
# Analyzing model fitment based on metrics
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
lr_cm = confusion_matrix(y_test, y_lr_pred)
knn_cm = confusion_matrix(y_test, y_knn_pred)
rf_cm = confusion_matrix(y_test, y_rf_pred)
dt_cm = confusion_matrix(y_test, y_dt_pred)

#Accurancy scores
lr_ac = accuracy_score(y_test, y_lr_pred).
knn_ac = accuracy_score(y_test, y_knn_pred)
rf_ac = accuracy_score(y_test, y_rf_pred)
dt_ac = accuracy_score(y_test, y_dt_pred)

score_dict ={"Decision Tree": round(dt_ac*100 , 2), "KNN_Score": round(knn_ac*100 , 2), "Logistic Regression": round(lr_ac*100 , 2), "Random Forest Score": round(rf_ac*100 , 2)}

print(score_dict)
```
