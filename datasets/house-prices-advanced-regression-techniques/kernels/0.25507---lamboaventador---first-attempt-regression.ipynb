```python
train = pd.read_csv('../input/train.csv')
sample_submission = pd.read_csv('../input/sample_submission.csv')
test = pd.read_csv('../input/test.csv')
train.columns
```

```python
train_price = train.SalePrice
print(train_price.head())
```

```python
columns_of_interest = ['LotArea', 'LotFrontage']
two_columns_of_data = train[columns_of_interest]
two_columns_of_data.describe()
```

```python
y = train.SalePrice
price_predictors = ['1stFlrSF', 'LotArea', '2ndFlrSF', 'YearBuilt', 
                        'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
x = train[price_predictors]
```

```python
from sklearn.tree import DecisionTreeRegressor

price_model = DecisionTreeRegressor()
price_model.fit(x,y)
```

```python
print("Making predictions for the following 5 houses:")
print(x.head())
print("The predictions are")
print(price_model.predict(x))
```

```python
from sklearn.metrics import mean_absolute_error

predicted_home_prices =price_model.predict(x)
mean_absolute_error(y, predicted_home_prices)
```

```python

price_predictors = ['1stFlrSF', 'LotArea', '2ndFlrSF', 'YearBuilt', 
                        'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
price_model = DecisionTreeRegressor()
price_model.fit(x,y)
X = test[price_predictors]
Y = price_model.predict(X)
```

```python
submission = pd.DataFrame({"Id": test["Id"], "SalePrice": Y})
```
