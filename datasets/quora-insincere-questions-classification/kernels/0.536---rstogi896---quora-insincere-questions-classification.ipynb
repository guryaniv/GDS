```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
train=pd.read_csv('../input/train.csv')
```

```python
test=pd.read_csv('../input/test.csv')
test.shape
```

```python
test.head()
```

```python
test_id=test['qid']
```

```python
sample_submission=pd.read_csv('../input/sample_submission.csv')
```

```python
sample_submission.head()
```

```python
from nltk.corpus import stopwords
```

```python
stop = stopwords.words('english')
```

```python
tfidf_vect = TfidfVectorizer(sublinear_tf=True,ngram_range=(1,2),min_df=5, norm='l2', encoding='latin-1',stop_words='english')
tfidf_vect.fit_transform(train['question_text'].values.tolist() + test['question_text'].values.tolist())
```

```python
train_tfidf = tfidf_vect.transform(train['question_text'].values.tolist())
test_tfidf = tfidf_vect.transform(test['question_text'].values.tolist())
```

```python
from sklearn.linear_model import LogisticRegression
clf= LogisticRegression(C=500.0)
```

```python
y_train=train['target']
y_train.shape
```

```python
clf.fit(train_tfidf,y_train)
```

```python
y_pred=clf.predict(test_tfidf)
```

```python
y_pred
```

```python
sum(y_pred)
```

```python
y_pred=y_pred.tolist()
```

```python
columns = ['prediction']
sub = pd.DataFrame(data=y_pred, columns=columns)
sub['qid'] = test_id
sub = sub[['qid','prediction']]
```

```python
sub.head()
```

```python
sub.to_csv('submission.csv',index=False)
```
