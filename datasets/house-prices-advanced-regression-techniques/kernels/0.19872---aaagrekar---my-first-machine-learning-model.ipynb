```python
y = data.SalePrice
```

```python
iowa_predictors = ['LotArea','YearBuilt','1stFlrSF','2ndFlrSF','FullBath','BedroomAbvGr','TotRmsAbvGrd']
X = data[iowa_predictors]
```

```python
from sklearn.tree import DecisionTreeRegressor

# define the model
iowa_model = DecisionTreeRegressor()

# fit the model
iowa_model.fit(X,y)
```

```python
print("Making predictions for the following 5 houses:")
print(X.head())
print("The predictions are")
print(iowa_model.predict(X.head()))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
# train test split
train_X, val_X, train_y, val_y = train_test_split(X,y, random_state = 0)

#define model
iowa_model = DecisionTreeRegressor()

#fit model
iowa_model.fit(train_X, train_y)

#predict
val_predictions = iowa_model.predict(val_X)

print("The mean absolute error for the predictionson the validation set is ", mean_absolute_error(val_y,val_predictions))
```

```python
def get_mae(max_leaf_nodes, predictors_train, predictors_val, targ_train, targ_val):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(predictors_train, targ_train)
    preds_val = model.predict(predictors_val)
    mae = mean_absolute_error(targ_val, preds_val)
    return(mae)
```

```python
for max_leaf_nodes in [5, 10, 50, 100, 200, 300, 400, 500, 600, 700, 800, 900, 1000, 2000, 3000, 4000, 5000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
```

```python
from sklearn.ensemble import RandomForestRegressor

iowa_rf_model = RandomForestRegressor()
iowa_rf_model.fit(train_X, train_y)
val_rf_predictions = iowa_rf_model.predict(val_X)
print("The mean absolute error for the Random Forest Regressor is ", mean_absolute_error(val_y,val_rf_predictions))
```
