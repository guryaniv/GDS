### Naïve Bayes Baseline

Let's start with some assumptions: features are independent and normal distributed. This lead us to simple model - Naive Bayes Classifier with gaussian distributed features.

```python
import pandas as pd
import numpy as np
```

```python
df_train = pd.read_csv('../input/train.csv', index_col='id')
df_test = pd.read_csv('../input/test.csv', index_col='id')
```

```python
target = 'target'

X_train = df_train.drop([target], axis=1).values
y_train = df_train[target]
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import cross_val_score
```

```python
cv = StratifiedKFold(10)

clf = GaussianNB()

scores = cross_val_score(estimator=clf, X=X_train, y=y_train, cv=cv, scoring='roc_auc')
```

```python
scores
```

```python
np.mean(scores)
```

This is optimistic score; on public leaderbord we will receive only 0.61

```python
X_test = df_test.values
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
pd.value_counts(y_pred, normalize=True)
```
