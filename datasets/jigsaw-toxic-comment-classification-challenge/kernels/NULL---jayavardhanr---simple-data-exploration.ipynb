```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
from wordcloud import WordCloud

import warnings
warnings.filterwarnings('ignore')
```

```python
train_df=pd.read_csv('../input/train.csv')
```

```python
train_df.head(20)
```

```python
train_df.shape
```

```python
test_df=pd.read_csv('../input/test.csv')
```

We can clearly see, the distributi[](http://)on of training data is skewed. We have lesser number of positive case( where a comment is labeled 1 for some toxic category)

We create a New Label 'Toxicity label' which is equal to 1 if any type of toxicity was true for the comment

```python
#if any toxic label was set to 1
train_df['toxicity_label']=(train_df==1).any(axis=1)
```

```python
is_toxic = train_df['toxicity_label'].value_counts()
plt.figure(figsize=(10,5))
sns.barplot(is_toxic.index, is_toxic.values, alpha=0.8)
plt.ylabel('Number of Occurrences', fontsize=12)
plt.xlabel('Toxicity label', fontsize=12)
plt.show()
```

**Observation**
1. Toxic , Obscene, Insult are highly correlated relative to each other
2. Main Toxicity Label is highly correlated to 'toxic' label

```python
colormap = plt.cm.viridis
plt.figure(figsize=(14,12))
plt.title('Pearson Correlation plot of toxic behaviors', y=1.05, size=15)
sns.heatmap(train_df[['toxicity_label','toxic','severe_toxic','obscene','threat','insult','identity_hate']].astype(float).corr(),linewidths=0.1,
            vmax=1.0, square=True, cmap=colormap, linecolor='white',
            annot=True)
```

**Occurence Distribution of each type of Toxicity**
1. For every category, we can see the positive examples are lesser than negative examples
