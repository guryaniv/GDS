```python
properties0 = pd.read_csv("../input/properties_2016.csv")
properties0.shape
```

```python
merge0 = pd.merge(training0, properties0, on='parcelid', how='left')
merge0.head()
```

```python
training0['transaction_month'] = training0['transactiondate'].dt.month
# training0['transaction_year'] = training0['transactiondate'].dt.year
# after checking we realized all the data are from year 2016
training0.head()
```

```python
count_month = training0['transaction_month'].value_counts()
plt.figure(figsize=(12,6))
sns.barplot(count_month.index, count_month.values, alpha=0.8, color=color[3])
plt.xticks(rotation='vertical')
plt.xlabel('Month of transaction', fontsize=12)
plt.ylabel('Number of Occurrences', fontsize=12)
plt.show()
```
