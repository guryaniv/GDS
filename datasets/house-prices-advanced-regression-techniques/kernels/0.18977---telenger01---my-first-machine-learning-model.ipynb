```python
#Original, the one I started with 
import pandas as pd

main_file_path = '../input/train.csv' # this is the path to the Iowa data that you will use
test_file_path = '../input/test.csv'
data = pd.read_csv(main_file_path)
y = data.SalePrice
columns_of_interest = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X = data[columns_of_interest]

#Decision Tree Model
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor.fit(X,y)
#predicting price
predicted_price = regressor.predict(X)
#calculating Absolute error
from sklearn.metrics import mean_absolute_error
error_before_spliting = mean_absolute_error(y, predicted_price)

#Spliting data into training set and test/validation set
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X, y,random_state = 0)
regressor1 = DecisionTreeRegressor()
regressor1.fit(train_X,train_y)

# get predicted prices on validation data
test_predicted_price = regressor1.predict(test_X)
error_after_spliting = mean_absolute_error(test_y, test_predicted_price)

#print("Making predictions for the following 5 houses:")
#print (X.head())
#print("The Error without splitting is")
#print(error_before_spliting)
#print("The Error after splitting is")
#print(error_after_spliting)


#compare MAE scores from different values for max_leaf_nodes to find the optimal number of leaves
def get_mae(max_leaf_nodes, predictors_train, predictors_val, targ_train, targ_val):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(predictors_train, targ_train)
    preds_val = model.predict(predictors_val)
    mae = mean_absolute_error(targ_val, preds_val)
    return(mae)
for max_leaf_nodes in [5, 50, 500, 5000, 10000]:
    my_mae = get_mae(max_leaf_nodes, train_X, test_X, train_y, test_y)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
    
    #Result of above loop shows that 50 is optimal number of leaves
```
