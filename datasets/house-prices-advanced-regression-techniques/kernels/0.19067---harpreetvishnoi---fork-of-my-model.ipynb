```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor

main_file_path = '../input/train.csv'
data = pd.read_csv(main_file_path)

y=data.SalePrice
X=data[['LotArea','YearBuilt','1stFlrSF','2ndFlrSF','FullBath','BedroomAbvGr','TotRmsAbvGrd']]

forest_model = RandomForestRegressor()
forest_model.fit(X,y)
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor

main_file_path = '../input/test.csv'
data = pd.read_csv(main_file_path)

X=data[['LotArea','YearBuilt','1stFlrSF','2ndFlrSF','FullBath','BedroomAbvGr','TotRmsAbvGrd']]
SalePrice = forest_model.predict(X)

my_submission = pd.DataFrame({'Id': data.Id, 'SalePrice': SalePrice})
#print(my_submission)
# you could use any filename. We choose submission here
my_submission.to_csv('submission007.csv', index=False)
```
