```python
#https://www.kaggle.com/opanichev/lightgbm-regressor
#https://www.kaggle.com/aantonova/797-lgbm-and-bayesian-optimization
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
print('Training and making predictions')
params = {
    'boosting_type': 'gbdt',
    'objective': 'regression_l2',
    'metric': 'rmse',
   # 'max_depth': 6, 
    'learning_rate': 0.001,
    'verbose': 0, 
    'subsample': 0.8,
    #'min_child_weight': 4,
    #'colsample_bytree': 0.4,
    'num_leaves': 240,
    #'feature_fraction':0.8,
    #'early_stopping_round': 20
}


n_estimators = 2000

n_iters = 10
preds_buf = []
err_buf = []
for i in range(n_iters): 
    x_train, x_valid, y_train, y_valid = train_test_split(X, y, test_size=0.30, random_state=i)
    d_train = lgb.Dataset(x_train, label=y_train)
    d_valid = lgb.Dataset(x_valid, label=y_valid)
    watchlist = [d_valid]

    model = lgb.train(params=params, train_set=d_train, num_boost_round=n_estimators, valid_sets=watchlist, verbose_eval=1)
    #model = lgb.train(params=lgbm_params, train_set=d_train, valid_sets=watchlist, verbose_eval=1)
    preds = model.predict(x_valid)
    print(preds.min())
    preds[preds < 0] = median_revenue
    print(preds.min())
    #err = rmse(np.exp(y_valid), preds)
    err = mean_squared_error(y_valid, preds) ** 0.5
    #err = y_valid - preds 
    print("rmse is: ", err)
    print('********************************************************')
    
    err_buf.append(err)
    preds = model.predict(X_test)
    
    preds[preds < 0] = median_revenue
    preds_buf.append(preds)


print("Mean RMSE =  {}   +/-   {} ".format(np.mean(err_buf), np.std(err_buf)))

# Average predictions
preds = np.mean(preds_buf, axis=0)

# Prepare submission
subm = pd.DataFrame()
subm['id'] = test.id.values
subm['revenue'] = preds
subm.to_csv('submission.csv', index=False)
```
