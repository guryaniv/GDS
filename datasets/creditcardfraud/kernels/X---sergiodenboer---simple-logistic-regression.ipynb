```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score, confusion_matrix
```

```python
SEED=113
models = {}
```

```python
dataset = pd.read_csv("creditcard.csv")
```

```python
features = list(dataset.columns.values)
features.remove('Class')
```

```python
count_classes = pd.value_counts(dataset['Class'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(dataset[features], dataset['Class'], test_size=0.33, random_state=SEED)
```

The first attempt will be a Logistic regression where one choose to ignore the regularization parameter (hence the huge C value in the dfinition) The regularization is not interesting due to the low number of variables in the dataset


```python
models['logistic'] = LogisticRegression(random_state=SEED, C=10E6)
```

```python
models['logistic'].fit(X_train, y_train)
```

```python
print confusion_matrix(y_test, models['logistic'].predict(X_test), labels=(0,1))
print roc_auc_score(y_test,models['logistic'].predict(X_test))
```

The model does not to seem to perform good when predicting 1's (second row) although it yields an good auc and a great accuracy, This baseline cuttoff value is set to 0.5.
