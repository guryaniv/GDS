```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

train_file_path = '../input/house-prices-advanced-regression-techniques/train.csv' # this is the path to the Iowa data that you will use
test_file_path = '../input/house-prices-advanced-regression-techniques/test.csv'
# Read train data
train = pd.read_csv(train_file_path)

# Read test data
test = pd.read_csv(test_file_path)

# target prediction
train_y = train.SalePrice

# training predictors data
predictors = ['LotArea','OverallQual','YearBuilt','TotRmsAbvGrd']
train_X = train[predictors]
test_X = test[predictors]

# define the model
iowa_model = RandomForestRegressor()

# fit the model
iowa_model.fit(train_X,train_y)

# predicted prices
predicted_prices = iowa_model.predict(test_X)
print(str("The predicted prices are: ") + str("\n") + str(predicted_prices))

# Submission file
submission_file = pd.DataFrame({'Id': test.Id, 'SalePrice' : predicted_prices})
submission_file.to_csv('submission.csv',index= False)






```
