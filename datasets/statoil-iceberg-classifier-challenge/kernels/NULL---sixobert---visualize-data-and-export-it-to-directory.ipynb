## Load data

## Class distribution

Here you can see how many images do we have in each classe

```python
iceberg_image = train_data.loc[train_data['is_iceberg'] == 1]
non_iceberg_image = train_data.loc[train_data['is_iceberg'] == 0]
class_examples = {'Non_Iceberg_examples':len(non_iceberg_image), 'Iceberg_examples':len(iceberg_image) }
print("Iceberg image examples ", class_examples['Iceberg_examples'])
print("Non_Iceberg image examples ", class_examples['Non_Iceberg_examples'])
plt.figure(figsize=(6,6))
plt.bar(range(len(class_examples)), class_examples.values(), align='center')
plt.xticks(range(len(class_examples)), class_examples.keys())
```
