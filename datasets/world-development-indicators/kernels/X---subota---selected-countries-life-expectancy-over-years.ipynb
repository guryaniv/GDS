# Life Expectancy
Over the years of 1960-2012
Over six selected countries

Some conclusions:
## The World is Progressing
The world gained about 18 extra years of life expectancy since 1960
## USSR had little progress
Started 15 points ahead of the world average in 1960, finished in zero difference in 2012..
## West world steadily progresses
Started equal to USSR in 1960, ended up 12 points ahead.
Development rate is about the world average growth.

```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load in 

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt

# Input data files are available in the "../input/" directory.
# For example, running this (by clicking run or pressing Shift+Enter) will list the files in the input directory

from subprocess import check_output
# print(check_output(["ls", "../input"]).decode("utf8"))

# Any results you write to the current directory are saved as output.
ind = pd.read_csv('../input/Indicators.csv')
countries = pd.read_csv('../input/Country.csv')
```

```python
life_expectancy = ind[ ind.IndicatorCode == 'SP.DYN.LE00.IN' ]

plt.figure(1).set_size_inches(12,8)
styles = ['orange', 'b', 'g', 'r', 'magenta', 'black', 'olive', 'yellow', 'cyan', ]
for c in ['JPN', 'DEU', 'UKR', 'RUS', 'WLD', 'NGA']:
    d = life_expectancy[ life_expectancy.CountryCode == c ]
    cpountryName = countries[countries.CountryCode == c].iloc[0].ShortName
    plt.plot(d.Year, d.Value, styles.pop(0), label = cpountryName)
plt.ylabel('Life expectancy, years')
plt.xlabel('Year')
plt.legend(loc='lower left', shadow=True)
plt.grid()
plt.show()
```
