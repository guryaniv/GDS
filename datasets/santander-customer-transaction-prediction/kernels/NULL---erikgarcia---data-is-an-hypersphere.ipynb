```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv("../input/train.csv").values
test = pd.read_csv("../input/test.csv").values[:,1:]
labels = train[:,1]
train = train[:,2:]
```

Scale the data:

```python
data = np.concatenate((train,test),axis=0)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data = scaler.fit_transform(data)
train = data[:200000]
test = data[200000:]
```

```python
l_0 = labels==0
l_1 = labels==1
```

Mean value of each row:

```python
aux = np.mean(train,axis=1)
sns.kdeplot(aux[l_0], label='Train : 0')
sns.kdeplot(aux[l_1], label='Train : 1')
```

Percentiles:

```python
for q in [0,10,25,50,75,90,100]:
    fig, ax = plt.subplots(1, 1, figsize=(10, 4))
    aux = np.percentile(train,q,axis=1)
    sns.kdeplot(aux[l_0],ax=ax, label='Train : 0')
    sns.kdeplot(aux[l_1],ax=ax, label='Train : 1')

    ax.set_title('percentile: {}'.format(q))
    plt.show()
```

We can see that 1 target values have a bigger deviation than the 0 target values so lets calculate the euclidean distance of each row:

```python
aux = np.linalg.norm(train,axis=1)
sns.kdeplot(aux[l_0], label='Train : 0')
sns.kdeplot(aux[l_1], label='Train : 1')
```

Data is an hypersphere of 200 dimensions where the 0 values are closer to the center of the sphere. 

We can see the proyections of that hypersphere in a  lower dimension just picking a number of columns and we will see that this property holds.

Distance of 5 random proyections in a space of 100 dimensions (pick randomly 100 columns)

```python
for i in range(5):
    fig, ax = plt.subplots(1, 1, figsize=(10, 4))
    ind = np.random.randint(200,size=100)
    aux = np.linalg.norm(train[:,ind],axis=1)
    sns.kdeplot(aux[l_0], label='Target == 0')
    sns.kdeplot(aux[l_1], label='Target == 1')

    plt.show()
```
