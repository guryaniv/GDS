```python
import pandas as pd
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split


#loading input file by using read_csv
main_file_path = '../input/train.csv'
input_data = pd.read_csv(main_file_path)
mutiple_columns=['LotArea','1stFlrSF','2ndFlrSF','FullBath','YearBuilt','BedroomAbvGr','TotRmsAbvGrd']
x=input_data[mutiple_columns]
y=input_data.SalePrice

#splitting the input data for training and validation of model
train_data_x,validation_data_x,train_data_y,validation_data_y =train_test_split(x,y,random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor

#defining random forest regressor model
ensemble_model =RandomForestRegressor()

#fitting the model with training data 
ensemble_model.fit(train_data_x,train_data_y)

#predicting the error of the model by validation data 
random_prediction = ensemble_model.predict(validation_data_x)
mean_abs_error = mean_absolute_error(random_prediction,validation_data_y)
print(mean_abs_error)
```

```python
#importing test data
test_data = pd.read_csv('../input/test.csv')

test_input =test_data[mutiple_columns]
test_prediction=ensemble_model.predict(test_input)

#printing the predicted test data output 
print(test_prediction)

```

```python
#preparing the kernel for submission

submission_data=pd.DataFrame({"ID":test_data.Id ,"SalePrice": test_prediction})
submission_data.to_csv("mlsubmission.csv",index=False)
```
