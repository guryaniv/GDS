```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pyarrow.parquet as pq
import os

# Library for STL decomposition
from statsmodels.tsa.seasonal import seasonal_decompose
```

## Load Data

```python
train = pq.read_pandas('../input/train.parquet').to_pandas()
```

```python
train.info()
```

```python
meta_train = pd.read_csv('../input/metadata_train.csv')
```

```python
meta_train.info()
```

The data is imbalanced, and there are not many entries.

```python
fig = sns.countplot(meta_train.target)
```

## STL Transformation

```python
grouped_measures = meta_train.groupby('id_measurement').signal_id.apply(list).apply(lambda l: [str(x) for x in l])
grouped_measures = pd.DataFrame(dict(group_id=grouped_measures.values, target=meta_train.groupby('id_measurement').target.min()))
```

```python
grouped_measures.head()
```

#### Positive example

```python
n_pos = (grouped_measures.target==0).sum()

for i in np.random.choice(range(n_pos), 5, replace=False):
    print('Positive sample n°{}'.format(i))
    ex = train[grouped_measures[grouped_measures.target==0].group_id.values[i]]
    result = seasonal_decompose(ex, model='additive', freq=10000)
    fig = result.plot()
    fig.set_size_inches(11,8)
    plt.show()
    print()
```

#### Negative example

```python
n_neg = (grouped_measures.target==1).sum()

for i in np.random.choice(range(n_neg), 5, replace=False):
    print('Negative sample n°{}'.format(i))
    ex = train[grouped_measures[grouped_measures.target==1].group_id.values[i]]
    result = seasonal_decompose(ex, model='additive', freq=10000)
    fig = result.plot()
    fig.set_size_inches(11,8)
    plt.show()
```
