```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.ensemble import RandomForestRegressor
```

```python
main_file_path = '../input/train.csv'
data = pd.read_csv(main_file_path)
print('hello world')
```

```python
data.describe()
```

```python
data.columns 
```

```python
house_sale_prices = data['SalePrice']
house_sale_prices.head()
```

```python
two_cols = ['LotArea', 'SalePrice']
two_cols_df = data[two_cols]
two_cols_df.describe()
```

```python
y = data['SalePrice']
```

```python
predictors = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X = data[predictors]
```

```python
iowa_model = DecisionTreeRegressor()
iowa_model.fit(X, y)
```

```python
iowa_model.predict(X.head())
```

```python
y.head()
```

```python
train_x, val_x, train_y, val_y = train_test_split(X, y, random_state = 0)
```

```python
new_iowa_model = DecisionTreeRegressor()
new_iowa_model.fit(train_x, train_y)
```

```python
val_predictions = new_iowa_model.predict(val_x)
```

```python
mean_absolute_error(val_y, val_predictions)
```

```python
def get_mae(max_leaf_nodes, train_x, train_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_x, train_y)
    val_pred = model.predict(val_x)
    return mean_absolute_error(val_y, val_pred)
```

```python
for leaf_nodes in [20,50, 100, 300, 500, 1000]:
    mae = get_mae(leaf_nodes, train_x, train_y)
    print('leaf nodes = ', leaf_nodes, ';mae = ',mae)
```

```python
random_forest_model = RandomForestRegressor()
random_forest_model.fit(train_x, train_y)
random_forest_pred = random_forest_model.predict(val_x)
mean_absolute_error(val_y, random_forest_pred)
```

```python
test_file_path =  '../input/test.csv'
test_data = pd.read_csv(test_file_path)
```

```python
test_x = test_data[predictors]
final_predictions = random_forest_model.predict(test_x)
```

```python
my_submission = pd.DataFrame({'Id': test_data.Id, 'SalePrice': final_predictions})
# you could use any filename. We choose submission here
my_submission.to_csv('submission.csv', index=False)
```
