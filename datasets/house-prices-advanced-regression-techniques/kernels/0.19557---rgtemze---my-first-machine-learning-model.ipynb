```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split


main_file_path = '../input/train.csv'
data = pd.read_csv(main_file_path)
test = pd.read_csv('../input/test.csv')
y = data.SalePrice
prediction_cols = ['LotArea','YearBuilt','1stFlrSF','2ndFlrSF','FullBath','BedroomAbvGr','TotRmsAbvGrd']
X = data[prediction_cols]

test_X = test[prediction_cols]

forest_model = RandomForestRegressor()
forest_model.fit(X, y)
predictions = forest_model.predict(test_X)
print(predictions)
my_submission = pd.DataFrame({'Id': test.Id, 'SalePrice': predictions})
# you could use any filename. We choose submission here
my_submission.to_csv('submission.csv', index=False)
```
