# Parameter
df: the dataframe  
col: columns that you want to impute  
n: number of iteration, the higher n, the more likely the imputed value obey the real distribution

```python
#from sklearn.ensemble import RandomForestClassifier
#from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from xgboost import XGBClassifier
def Imputation(df, col, n=10):
    Y = df[col]
    X = pd.get_dummies(df.drop(col,axis=1))
    datatype = Y.dtype
    if datatype == 'float64':
        model = XGBRegressor()
    elif datatype == 'object' or 'int64':
        model = XGBClassifier()
    obs = pd.isnull(df[col]) != True
    mis = pd.isnull(df[col]) == True
    model.fit(X[obs],Y[obs])
    df.loc[mis,col] = model.predict(X.loc[mis])
    for i in range(n):
        Y = df[col]
        #print(X)
        #print(Y)
        #print(model)
        model.fit(X,Y)
        df[mis][col] = model.predict(X[mis])
    return df
```
