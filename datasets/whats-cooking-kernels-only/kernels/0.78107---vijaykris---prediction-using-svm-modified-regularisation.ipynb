```python
from sklearn import svm
from sklearn.svm import SVC
from sklearn.metrics import f1_score
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split

train=pd.read_json('../input/train.json' )
test=pd.read_json('../input/test.json')

train_n = [", ".join(ingredients) for ingredients in train['ingredients']]
test_n=[", ".join(ingredients) for ingredients in test['ingredients']]

X_train, X_val, y_train, y_val = train_test_split(train_n, train['cuisine'], random_state=0)
vect=TfidfVectorizer().fit(X_train)
X_train_vect=vect.transform(X_train)
#considering svm model
svmmodel=svm.SVC(kernel='linear', C=1.2)
svmmodel.fit(X_train_vect, y_train)

predictions=svmmodel.predict(vect.transform(X_val))
print(f1_score(predictions, y_val, average='micro'))
test_predict=svmmodel.predict(vect.transform(test_n))

test['cuisine']=test_predict
test = test.sort_values('id' , ascending=True)

test[['id' , 'cuisine' ]].to_csv("submission.csv", index=False)
```
