Kasutatav andmestik on loodud leheküljelt [WineEnthusiast](http://www.winemag.com/?s=&drink_type=wine) kogutud veini hinnangutest. Andmestik koosneb 130 000 veinist, mis on saanud maitsja poolt vähemalt 80 punktise hinnangu. Andmestikus on välja toodud järgnev informatsioon: riik, lühikirjeldus, punktisumma, hind, provints, regioon, maitsja nimi, veini nimi, veini sort, veinikelder.

Võrdlen erinevate riikide veinide hinda ja kvaliteeti ning uurin milline veinisort on maitsjate lemmik. 


```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load in 

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

%matplotlib inline
pd.set_option('display.max_rows', 20)
# Input data files are available in the "../input/" directory.
# For example, running this (by clicking run or pressing Shift+Enter) will list the files in the input directory

df = pd.read_csv("../input/winemag-data-130k-v2.csv", sep=",", decimal=",", engine='python')

# Any results you write to the current directory are saved as output.
```

* Eemaldades andmestikust väga kallid veinid ( < 1% andmetest). Näeme, et keskmine veinipudeli hind jääb vahemikku 10-20 dollarit

```python
df[df.price <= 150].price.hist(bins=15,grid=False,rwidth=0.9)
```

* Keskmine veinile antud hinnang jääb aga vahemikku 88-89.

```python
df.points.hist(bins=20,grid=False,rwidth=0.9)
```

* Konkurentsitult kõige kõrgema keskmise hinnaga, 85 dollarit pudel, on Šveitsi veinid. Kõrge hind ei väljendu aga maitses, mis on vaid veidi üle keskmise.
* Veel üheks outlier'iks scatterplotis on Inglismaa veinid, mis on nii pudelihinna kui ka hinnangu poolest esireas. 
* Kõige parema hinna / kvaliteedi suhtega veinid pärinevad aga Indiast, kust pärit veinide keskmiseks punktisummaks on 90 ning pudelihinnaks vaid 13 dollarit.
* Kõige halvema keskmise hinnanguga veinid pärinevad Peruust.

```python
mean_points = df.groupby('country', as_index=False)['points'].mean()
mean_price = df.groupby('country', as_index=False)['price'].mean()
df2 = pd.merge(mean_points,mean_price,on="country")

df2.plot.scatter("points", "price",s=50,alpha=0.75);
```

```python
mean_points.sort_values(by="points").sort_values(by='points',ascending=False).reset_index(drop=True)
```

```python
mean_price.sort_values(by="price").sort_values(by='price',ascending=False).reset_index(drop=True)
```

* Veinisortide hulgast, mida on hinnatud vähemalt viiel korral, on kõige parema keskmise punktisummaga Bual ja Muscadelle.
* Üheks parima hinna/kvaliteedi suhtega veinisordiks on aga Tannat-Cabernet Franc, mille pudeli hind jääb 20 dollari kanti.
* Kõige halvemini hinnatud veinisordiks on Tempranillo Blanco

```python
count = df["variety"].value_counts()
df[df['variety'].isin(count[count >= 5].index)].groupby('variety', as_index=False)['points','price'].mean().sort_values(by='points',ascending=False).reset_index(drop=True)
```
