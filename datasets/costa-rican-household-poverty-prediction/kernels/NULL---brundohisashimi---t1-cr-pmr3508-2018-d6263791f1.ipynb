```python
def agrupa(dat,st,st2):
    dat.insert(2, st, 0)
    for i in range(1,len(dat[st])):
        for j in range(len(st2)):
            m=st+str(st2[j])
            if dat.loc[i,(m)] == '1':
                dat.loc[i,(st)] = j
    for x in range(len(st2)):
        k=st+str(st2[x])
        dat = dat.drop([k],axis=1)
    return dat
```
