```python
import pandas as pd
from sklearn.metrics import mean_absolute_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

RandomForest=RandomForestRegressor()
main_file_path = '../input/train.csv' # this is the path to the Iowa data that you will use
Iowa_data = pd.read_csv(main_file_path)
test = pd.read_csv('../input/test.csv')


x_columns=['LotArea','YearBuilt','1stFlrSF','2ndFlrSF','FullBath','BedroomAbvGr','TotRmsAbvGrd'];
y=Iowa_data.SalePrice;
####print(test.Id)
x=Iowa_data[x_columns]
test_X = test[x_columns]
RandomForest.fit(x,y);
predicted_prices=RandomForest.predict(test_X);
my_submission=pd.DataFrame({'Id':test.Id,'SalePrice': predicted_prices});
# you could use any filename. We choose submission here
my_submission.to_csv('submission.csv', index=False)
```
