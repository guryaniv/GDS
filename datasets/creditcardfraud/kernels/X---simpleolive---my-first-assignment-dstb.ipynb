**SCOPE**

There are three main concepts here we need to focus:
* global scope : defined in the main body of the script
* local scope : defined in a specific function
* built-in scope : names in predefined built in scope such as print, len

Let's take a look at some examples about the scope concept.

```python
# difference between local and global scopes

a = 5
def f():
    a = 10
    return a
print(a)                     # global scope
print(f())                  # local scope
```

```python
# Let's see what happens when there is no local scope

a = 10
def f():
    b = a**2               # there is no local scope a
    return b
print(f())                 # it uses global scope a

# The script first searches local scope, then global scopes. 
# If nothing found in these then script searches for built in scopes. 
```

```python
# Let's see how we can see the built-in scopes:

import builtins
dir(builtins)
```

**NESTED FUNCTION**

Nested function can be defined as a function with more than one layer. By layer we mean that this kid of function is like a function inside a function.

There is a LEGB rule that is search local scope, enclosing function, global and built in scopes, respectively.

```python
# nested function

def divide():
    """this function returns the one tenth of the value"""
    def subtract():
        """this function returns the subtraction of 10 from the value"""
        c = 200
        d = c-10
        return d
    return subtract()/10
print(divide())
```

**DEFAULT AND FLEXIBLE ARGUMENTS**

* Default argument : this argument is fixed in the function
   
   Ex: f(x,y=1):
   
   """ here y=1 is the default argument in the function and we do not need to write this argument when running the function.
   
* Flexible argument: this argument is like an open ath for the variables so that we can enter one or more variables to the     function.
   
   Ex: f(x,*args) 
  
  """ *args can be one or more """
  
  Ex: f(** kwargs)
  
  """ ** kwargs is a dictionary """

```python
# Default arguments

def g(m,n=5,o=25):
    p = m+n+o
    return p
print(g(1))

# If we would like to change the default arguments, we can simply enter the variables to the function

print(g(1,20,40))
```

```python
# flexible arguments

def h(*args):
    for i in args:
        print(i)
h(1)
print("")
h(1,2,3,4)

# flexible arguments **kwargs which is dictionary

def fg(**kwargs):
    """print keys and values of the dictionary"""
    for key, value in kwargs.items():
        print(key,":",value)
fg(atom1 = 'hydrogen', atom2 = 'oxygen', molecule = 'water', molecule_weight = 18)

```

**LIST COMPREHENSION**

**One of the most important topic of this kernel**


List comprehension is usually used for data analysis. 

By using list comprehension we can define loops in a single line. When using for loops it takes us longer to define the loop.

For instance we want to transform the num1=[1,2,3] into num2=[2,3,4]. We can do it by using for loop. However itis unnecessarily long. We can do it by list comprehensionin a single line.

```python
# example of list comprehension
num1 = [1,2,3]
num2 = [i+1 for i in num1]
print(num2)
```

**[i + 1 for i in num1 ]** : list of comprehension 

**i +1**                            : list comprehension syntax 

**for i in num1**            : for loop syntax 

**i**                                 : iterator

**num1 **                          : iterable object

```python
# Conditionals in iterables
# We can also define different rules for each element in the object as follows:
num3 = [10,20,30]
num4 = [i/10 if i == 20 else i-10 if i < 15 else i+5 for i in num3]
print(num4)
```
