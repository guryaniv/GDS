```python
dtypes = {
        'Id'                : 'uint32',
        'groupId'           : 'uint32',
        'matchId'           : 'uint16',
        'assists'           : 'uint8',
        'boosts'            : 'uint8',
        'damageDealt'       : 'float16',
        'DBNOs'             : 'uint8',
        'headshotKills'     : 'uint8', 
        'heals'             : 'uint8',    
        'killPlace'         : 'uint8',    
        'killPoints'        : 'uint8',    
        'kills'             : 'uint8',    
        'killStreaks'       : 'uint8',    
        'longestKill'       : 'float16',    
        'maxPlace'          : 'uint8',    
        'numGroups'         : 'uint8',    
        'revives'           : 'uint8',    
        'rideDistance'      : 'float16',    
        'roadKills'         : 'uint8',    
        'swimDistance'      : 'float16',    
        'teamKills'         : 'uint8',    
        'vehicleDestroys'   : 'uint8',    
        'walkDistance'      : 'float16',    
        'weaponsAcquired'   : 'uint8',    
        'winPoints'         : 'uint8', 
        'winPlacePerc'      : 'float16' 
}
```

```python
df=pd.read_csv('../input/train.csv', dtype=dtypes)
```

```python
df1=pd.read_csv('../input/test.csv', dtype=dtypes)
```

```python
df1.head()
```

```python
X=df.iloc[0:100000, 0:-1].values
```

```python
X1=df1.iloc[:, :].values
```

```python
Y=df.iloc[0:100000,-1].values
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X = sc_X.fit_transform(X)
```

```python
X1=sc_X.transform(X1)
```

```python
# Applying PCA
from sklearn.decomposition import PCA
pca=PCA(n_components=5)
X = pca.fit_transform(X)
explained_variance = pca.explained_variance_ratio_
```

```python
X1=pca.transform(X1)
```

```python
explained_variance
```

```python
X
```

```python
# Fitting SVR to the dataset
from sklearn.svm import SVR
regressor = SVR(kernel = 'rbf')
regressor.fit(X, Y)
```

```python
Y_pred=regressor.predict(X1)
```

```python
Y_pred.to_csv('csv_to_submit.csv', index = False)
```

```python
my_submission = pd.DataFrame({'Id': df1.Id, 'winPlacePerc': Y_pred})
```
