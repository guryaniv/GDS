## Dog Breed classification with fast ai

The fastai library simplifies training fast and accurate neural nets using modern best practices. This library is built on top of Pytorch. You can more about fastai by visiting this [link](https://docs.fast.ai/index.html).  


```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
from fastai.vision import *
from fastai.metrics import error_rate, accuracy

print(os.listdir("../input"))
```

```python
%reload_ext autoreload
%autoreload 2
%matplotlib inline
```

```python
PATH = "../input/"
MODEL_PATH = "/tmp/model/"
```

```python
df = pd.read_csv('../input/labels.csv')
df.head(10)
```

```python
print(df['breed'].value_counts().sort_values(ascending=False))
```

```python
plt.figure(figsize=(12,8))
plt.hist(df['breed'].value_counts().sort_values(ascending=False))
plt.show()
```

```python
bs = 64
```

### Get the Data

```python
tfms = get_transforms(max_rotate=25); len(tfms)
```

```python
data = ImageDataBunch.from_csv(PATH, folder='train', test='test', suffix='.jpg', ds_tfms=tfms,
                               csv_labels='labels.csv', fn_col=0, label_col=1, 
                               size=128, bs=bs).normalize(imagenet_stats)
```

```python
data.show_batch(rows=4, figsize=(12,8))
```

```python
print(data.classes)
len(data.classes),data.c
```

### Training with Resnet 34

```python
learn = create_cnn(data, models.resnet34, metrics=accuracy, model_dir=MODEL_PATH)
```

```python
learn.model
```

```python
learn.lr_find()
learn.recorder.plot()
```

```python
learn.fit_one_cycle(5, max_lr=slice(3e-2,1e-2))
```

```python
learn.save('/tmp/model/stage-1-34')
```

```python
interp = ClassificationInterpretation.from_learner(learn)
losses,idxs = interp.top_losses()
len(data.valid_ds)==len(losses)==len(idxs)
```

```python
interp.plot_top_losses(9, figsize=(15,11))
```

```python
interp.most_confused(min_val=2)
```

```python
#learn.load('/tmp/model/stage-1')
learn.unfreeze()
```

```python
learn.lr_find()
learn.recorder.plot()
```

```python
learn.fit_one_cycle(2, max_lr=slice(1e-6,1e-4))
```

```python
learn.save('/tmp/model/stage-2-34')
```

### Training with Rest Net 50

```python
data = ImageDataBunch.from_csv(PATH, folder='train', test='test', suffix='.jpg', ds_tfms=tfms,
                               csv_labels='labels.csv', fn_col=0, label_col=1, 
                               size=224, bs=16).normalize(imagenet_stats)
```

```python
data.show_batch(rows=4, figsize=(12,8))
```

```python
learn = create_cnn(data, models.resnet50, metrics=accuracy, model_dir=MODEL_PATH)
```

```python
learn.lr_find()
learn.recorder.plot()
```

```python
learn.fit_one_cycle(3, max_lr=slice(1e-3,1e-2))
```

```python
learn.save('/tmp/model/stage-1-50')
```

```python
learn.unfreeze()
```

```python
learn.lr_find()
learn.recorder.plot()
```

```python
learn.fit_one_cycle(3, max_lr=slice(1e-5,1e-4))
```

```python
learn.save('/tmp/model/stage-2-50')
```

```python
predictions = learn.get_preds(ds_type=DatasetType.Test)
```

```python
predictions[0][0]
```

```python
sample_submission = pd.read_csv('../input/sample_submission.csv')
sample_submission.head()
```

```python
submission = sample_submission.copy()
for i in range(len(submission)):
    submission.iloc[i, 1:] = predictions[0][i].tolist()
submission.head()
```
