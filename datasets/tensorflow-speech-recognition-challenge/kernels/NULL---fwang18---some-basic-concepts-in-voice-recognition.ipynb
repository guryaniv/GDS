 # Some basic concepts in voice recognition

## Basic concepts:

FFT: time domain-> frequency domain
Get spectrogram: time domain->fft to different span->rotate 90 degrees->translate into grey scale->spectrogram


Mel-frequency filter: 
point: human ears concentrate on low frequency components


Cepstral Analysis
Spectrum can be split  into spectral envelope (log H[k]) and spectral details (log E[k]) in log form by FFT
Then perform IFFT: x[k]=h[k]+e[k]
h[k] is what we want
