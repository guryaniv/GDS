```python
def describe_columns(data, label_column = None):
    '''
        Returns information for each column in the dataset
    '''
    columns = pd.DataFrame(columns=["data_type","unique_values","unique_values_count", "NA's","is_label","discard_column","need_label_encoding","need_one_hot_encoding"], index=data.columns)

    for index,column in enumerate(data.columns):
        columns.loc[column,"data_type"] = data[column].dtype
        columns.loc[column,"unique_values"] = data[column].unique()
        columns.loc[column,"unique_values_count"] = len(columns.loc[column,"unique_values"])
        columns.loc[column,"NA's"] = data[column].isna().sum()
        columns.loc[column,"is_label"] = column == label_column
        columns.loc[column,"discard_column"] = columns.loc[column,"data_type"] == "object" and data.shape[0] == columns.loc[column,"unique_values_count"]
        columns.loc[column,"need_label_encoding"] = columns.loc[column,"data_type"] == "object" and columns.loc[column,"discard_column"] == False
        columns.loc[column,"need_one_hot_encoding"] = columns.loc[column,"data_type"] == "object" and columns.loc[column,"unique_values_count"] > 2 and columns.loc[column,"discard_column"] == False

    return columns.sort_values(by="unique_values_count")
```
