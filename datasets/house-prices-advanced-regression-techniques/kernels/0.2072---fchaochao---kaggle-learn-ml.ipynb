```python
import pandas as pd

main_file_path = '../input/house-prices-advanced-regression-techniques/train.csv' # this is the path to the Iowa data that you will use
original_data = pd.read_csv(main_file_path)
data = original_data.copy()
# Run this code block with the control-enter keys on your keyboard. Or click the blue botton on the left
print('Some output from running this cell')
#print(data.describe())
print(data.columns)
```

```python
# 1. choose columns
# choose a single column
data_price = data.SalePrice
data_price.head()
# choose multiple columns
good_column = ["Fence","SalePrice","ScreenPorch"]
data_good_columns = data[good_column]
data_good_columns.head()
data_good_columns.describe()
```

```python
# 2. Model1:DecisionTreeRegreesor
# (1)Creat datasets
y = data.SalePrice
predictors = ["LotArea","YearBuilt","1stFlrSF","2ndFlrSF","FullBath","BedroomAbvGr","TotRmsAbvGrd"]
X = data[predictors]

# (2)Train model
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
import numpy as np
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state = 40,test_size = 0.25)
my_model = DecisionTreeRegressor()
my_model.fit(X_train,y_train)
y_predict = my_model.predict(X_test)

# (3) Evaluate the model
from sklearn.metrics import mean_absolute_error
model_accuracy = mean_absolute_error(y_test,y_predict)
print("How good is your model? The mean absolute error is {:3f}".format(model_accuracy))

# (4). Choose the best parameters
def get_best_para (max_leaf_nodes,X_train,X_test,y_train,y_test,accuracy_metric):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes,random_state = 40)
    model.fit(X_train,y_train)
    y_predict = model.predict(X_test)
    accuracy = accuracy_metric(y_test,y_predict)
    print("Accuracy is {:.2f}".format(accuracy))

for nodes in [5,50,500,5000]:
    accuracy_list = []
    a = get_best_para(nodes,X_train,X_test,y_train,y_test,mean_absolute_error)
    accuracy_list.append(a)
    
```

```python
# Model2: RandomForestRegressor
from sklearn.ensemble import RandomForestRegressor
rf_model = RandomForestRegressor(max_leaf_nodes = 50,random_state =40)
rf_model.fit(X_train,y_train)
y_predict = rf_model.predict(X_test)
print("The MAE is {:.2f}".format(mean_absolute_error(y_test,y_predict)))
```

```python
# Use Model2 to output the final result
test_path = '../input/house-prices-advanced-regression-techniques/test.csv'
test_final = pd.read_csv(test_path)
test_X = test_final[predictors]
predicted_prices = rf_model.predict(test_X)
# Form your submission
my_submission = pd.DataFrame({'Id':test_final.Id,"SalePrice":predicted_prices})
my_submission.to_csv('submission1.csv',index = False)
```
