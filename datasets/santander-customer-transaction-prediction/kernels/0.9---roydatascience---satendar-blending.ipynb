```python
import numpy as np
import pandas as pd
import os

from scipy.stats import rankdata

LABELS = ["target"]
```

```python
!ls ../input/
```

```python
predict_list = []

predict_list.append(pd.read_csv("../input/blendsatander/submission (v1).csv")[LABELS].values)
predict_list.append(pd.read_csv("../input/blendsatander/submission (v2).csv")[LABELS].values)
predict_list.append(pd.read_csv("../input/blendsatander/submission_ashish.csv")[LABELS].values)
predict_list.append(pd.read_csv("../input/blendsatander/submission_lgb.csv")[LABELS].values)
```

```python
print("Rank averaging on ", len(predict_list), " files")
predictions = np.zeros_like(predict_list[0])
for predict in predict_list:
    for i in range(1):
        predictions[:, i] = np.add(predictions[:, i], rankdata(predict[:, i])/predictions.shape[0])  
predictions /= len(predict_list)

submission = pd.read_csv('../input/santander-customer-transaction-prediction/sample_submission.csv')
submission[LABELS] = predictions
submission.to_csv('super_blend.csv', index=False)
```
