```python
diamond_data = pd.read_csv("../input/diamonds.csv")
```

```python
diamond_data.loc[1,'depth'] = None
diamond_data  =diamond_data.drop(["Unnamed: 0"],axis=1)
diamond_data.head()
```

```python
diamond_data.isnull().sum()
```

```python
diamond_data.dtypes
```

```python
one_hot_encoders_diamond_data =  pd.get_dummies(diamond_data)
```

```python
one_hot_encoders_diamond_data.head()
```

```python
from sklearn.preprocessing import Imputer
my_imputer = Imputer()
cols = one_hot_encoders_diamond_data.columns
imputed_encoded_diamond_data = my_imputer.fit_transform(one_hot_encoders_diamond_data)
```

```python
type(imputed_encoded_diamond_data)
#print(imputed_encoded_diamond_data[0])
diamond_clean_data = pd.DataFrame(imputed_encoded_diamond_data,columns= cols)
```

```python
diamond_clean_data.head()

from sklearn import preprocessing
#diamond_clean_data = preprocessing.scale(diamond_clean_data)

#diamond_clean_data = pd.DataFrame(diamond_clean_data,columns= cols)
diamond_clean_data.head()
```

```python
x = diamond_clean_data.drop("price",axis=1)

y = diamond_clean_data.price
```

```python
from sklearn.model_selection import train_test_split
train_x, val_x, train_y, val_y = train_test_split(x, y,random_state = 0)
```

```python
train_x.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score


regr = linear_model.LinearRegression()

regr.fit(train_x,train_y)
```

```python
y_pred = regr.predict(val_x)
```

```python
# plt.scatter(val_x.index[1:20], y_pred[1:20],  color='black')
# plt.scatter(val_x.index[1:20], y_pred[1:20], color='blue', linewidth=3)

# plt.xticks(())
# plt.yticks(())

# plt.show()
```

```python
from sklearn.metrics import mean_absolute_error
print("accuracy: "+ str(regr.score(val_x,val_y)*100) + "%")
print(mean_absolute_error(val_y,y_pred))
print(mean_squared_error(val_y,y_pred))

print('Coefficients: \n', regr.coef_)

```
