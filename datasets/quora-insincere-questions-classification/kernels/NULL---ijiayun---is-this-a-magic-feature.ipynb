```python
from collections import Counter
from nltk import word_tokenize
```

```python
train=pd.read_csv('../input/train.csv')
```

```python
def split_word(x):
    x=str(x)
    x=word_tokenize(x)
    return x
```

```python
train['question_text']=train['question_text'].apply(lambda x: split_word(x))
```

```python
text=train['question_text'].values
word_counts=Counter()
for s in text:
    word_counts.update(s)
```

```python
def fre_count(x):
    all_word=0
    all_fre=0
    for word in x:
        all_word+=1
        all_fre+=word_counts[word]
    return all_word/all_fre
```

```python
train['fre']=train['question_text'].apply(lambda x: fre_count(x))
```

```python
train=train.sort_values(by='fre',ascending=True)
train[0:50]
```

```python
train=train.sort_values(by='fre',ascending=False)
train[0:50]
```
