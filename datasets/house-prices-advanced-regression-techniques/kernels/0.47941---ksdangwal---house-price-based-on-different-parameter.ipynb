# 1. OverallQual

```python
trainData[['OverallQual', 'SalePrice']].head()
```

```python
avgSalesPrice_overallQual = trainData.groupby(['OverallQual']).mean()['SalePrice'].reset_index()

# Create a new column in a dataframe
avgSalesPrice_overallQual['avgSalesPriceQuality'] = avgSalesPrice_overallQual['SalePrice']

# Print dataframe
avgSalesPrice_overallQual
```

#### SalePrice Prediction based on OverallQual (Overall Quality)

##### Merge two dataframe based on inner join

```python
# Merge two column based on inner join
combinedQual = pd.merge(trainData, avgSalesPrice_overallQual, how='inner', on = 'OverallQual')

# Drop any column form the dataframe
combinedQual.drop(columns=['SalePrice_y'])
combinedQual.head()
```

```python
submissionData.head()
```

```python
submissionData['SalePrice'] = combinedQual['avgSalesPriceQuality'].values[0]

# find how many rows & columns in dataframe
submissionData.shape
```

# Submit Prediction based on Overall Quality
