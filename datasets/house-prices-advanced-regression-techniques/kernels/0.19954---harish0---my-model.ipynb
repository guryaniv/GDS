```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn import preprocessing




melbourne_file_path = '../input/train.csv'
data = pd.read_csv(melbourne_file_path) 
data=data.dropna(subset=['GarageCars'])
print(data.columns)
melbourne_predictors = ['BedroomAbvGr', 'TotRmsAbvGrd','FullBath', 'LotArea', 'YearBuilt'
                        , 'GrLivArea','Neighborhood','MSSubClass','OverallQual','OverallCond','YearBuilt',
                        'YearRemodAdd','GarageCars']

neighbor=(LabelEncoder().fit_transform((data['Neighborhood'])))
zone=LabelEncoder().fit_transform(data['MSZoning'])
subclass=LabelEncoder().fit_transform(data['MSSubClass'])                                  
print("Neighbourhood : ",neighbor)
data['Neighborhood']=pd.DataFrame(neighbor)
data['MSZoning']=pd.DataFrame(zone)
data['MSSubClass']=pd.DataFrame(subclass)
data['HouseStyle']=pd.DataFrame(LabelEncoder().fit_transform(data['HouseStyle']))
data['SaleCondition']=pd.DataFrame(LabelEncoder().fit_transform(data['SaleCondition']))

#data['OverallQual']=
#OverallCond
#YearBuilt
#YearRemodAdd

print("NEIGHBOURHOOD :::::", data['Neighborhood'].head(5))
print("zoning :::::", data['MSZoning'].head(5))
print("Subclass :::::", data['MSSubClass'].head(5))


X = data[melbourne_predictors]
y = data['SalePrice']


train_X, val_X, train_y, val_y = train_test_split(X, y,random_state = 0)
forest_model = RandomForestRegressor()
forest_model.fit(train_X, train_y)
melb_preds = forest_model.predict(val_X)
print(mean_absolute_error(val_y, melb_preds))

test = pd.read_csv('../input/test.csv')

#test=test.dropna()
print("NEEEEEEEEEEEEEEEEEEEEIGH", test.shape)
#test= test.drop(test['MSZoning'].str.contains("<")==False)
#print("ASSFSFDSDFSDAFSADFSDF",test['MSZoning'].head(5))

neighbor=LabelEncoder().fit_transform(test['Neighborhood'])
#zone=LabelEncoder().fit_transform(test['MSZoning'])
subclass=LabelEncoder().fit_transform(test['MSSubClass']) 
print("Neighbourhood : ",neighbor)
test['Neighborhood']=pd.DataFrame(LabelEncoder().fit_transform(test['Neighborhood']))
test['MSZoning'] = pd.DataFrame(zone)
test['MSSubClass']=pd.DataFrame(subclass)
test['HouseStyle']=pd.DataFrame(LabelEncoder().fit_transform(test['HouseStyle']))
#test['SaleCondition']=pd.DataFrame(LabelEncoder().fit_transform(test['SaleCondition']))

test_X = test[melbourne_predictors]


predicted_prices = forest_model.predict(test_X)

print(predicted_prices)

my_submission = pd.DataFrame({'Id': test.Id, 'SalePrice': predicted_prices})
my_submission.to_csv('submission.csv', index=False)
```
