```python
import pandas as pd

main_file_path = '../input/train.csv'
data = pd.read_csv(main_file_path)
print('hello world')
```

```python
print(data.describe())
```

```python
print(data.columns)
```

```python
data_price = data.SalePrice
print(data_price.head())
```

```python
columns_of_interest = ['LotArea', 'SalePrice']
area_price_view = data[columns_of_interest]
print(area_price_view.describe())
```

```python
y = data_price

predictors = ['LotArea','YearBuilt','1stFlrSF','2ndFlrSF','FullBath','BedroomAbvGr','TotRmsAbvGrd']
X = data[predictors]

from sklearn.tree import DecisionTreeRegressor

# Define the model
model = DecisionTreeRegressor()

# Fit the model
model.fit(X, y)
```

```python
print("Making predictions for the following 5 houses:")
print(X.head())
print("The predictions are")
print(model.predict(X.head()))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)

# Define and fit the model on training data
model = DecisionTreeRegressor()
model.fit(train_X, train_y)

predicted_home_prices = model.predict(val_X)
mean_absolute_error(val_y, predicted_home_prices)
```

```python
def get_mae(max_leaf_nodes, predictors_train, predictors_val, target_train, target_val):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(predictors_train, target_train)
    return mean_absolute_error(target_val, model.predict(predictors_val))

for max_leaf_nodes in [5, 50, 500, 5000]:
    mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Max leaf nodes: {}  \t\t Mean Absolute Error: {}".format(max_leaf_nodes, mae))
```

```python
from sklearn.ensemble import RandomForestRegressor

forest_model = RandomForestRegressor()
forest_model.fit(train_X, train_y)
predicted_home_prices = forest_model.predict(val_X)
print(mean_absolute_error(val_y, predicted_home_prices))
```

```python
# Read the test data
test = pd.read_csv('../input/test.csv')
# Treat the test data in the same way as training data. In this case, pull same columns.
test_X = test[predictors]
# Use the model to make predictions
predicted_home_prices = forest_model.predict(test_X)
# We will look at the predicted prices to ensure we have something sensible.
print(predicted_home_prices)
```

```python
my_submission = pd.DataFrame({'Id': test.Id, 'SalePrice': predicted_home_prices})
# you could use any filename. We choose submission here
my_submission.to_csv('submission.csv', index=False)
```
