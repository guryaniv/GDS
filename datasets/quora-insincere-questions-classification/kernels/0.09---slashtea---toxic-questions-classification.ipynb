#### We have few columns no initial EDA will be fast. Now what we see here is that we have class imbalance so we will need to stratify.

#### What we will need to do next is to do some feature extraction using TfidfVectorizer.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# TfidfVectorizer instance.
tfidf = TfidfVectorizer()

# Feature vector and target variable.
X = train_data['question_text']
y = train_data['target'].values

X_vec = tfidf.fit_transform(X)
```

```python
X_vec.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_vec, y, stratify=y)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

clf = MultinomialNB()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

print(accuracy_score(y_test, y_pred) *100, 2)
```
