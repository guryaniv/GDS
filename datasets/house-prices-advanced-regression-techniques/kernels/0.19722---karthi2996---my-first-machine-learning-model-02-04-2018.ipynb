```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

train_data = RandomForestRegressor()

main_file_path = '../input/train.csv'
data = pd.read_csv(main_file_path)

y=data.SalePrice
predict_data = ["LotArea", "YearBuilt", "1stFlrSF", "2ndFlrSF", "FullBath", "BedroomAbvGr", "TotRmsAbvGrd"]
X=data[predict_data]

train_data.fit(X, y)

test_file_path = '../input/test.csv'
test_data = pd.read_csv(test_file_path)

predict = test_data[predict_data]
anticipated_prices = train_data.predict(predict)

submit = pd.DataFrame({'Id':test_data.Id, 'SalePrice':anticipated_prices})
submit.to_csv('submit.csv', index=False)
```
