This kernel will show you how to combine all of the ETFs data into a single dataframe. This same code will technically work for the stocks data, but expect that to take a long time as there is a lot more data there.

```python
import os
import pandas as pd
```

```python
# kernels let us navigate through the zipfile as if it were a directory
os.chdir('../input/Data/ETFs/')
```

```python
# the data is initially stored in many small csv files
os.listdir()[:5]
```

```python
data = []
csvs = [x for x in os.listdir() if x.endswith('.txt')]
# trying to read a file of size zero will throw an error, so skip them
csvs = [x for x in csvs if os.path.getsize(x) > 0]
for csv in csvs:
    df = pd.read_csv(csv)
    df['ticker'] = csv.replace('.txt', '')
    data.append(df)
data = pd.concat(data, ignore_index=True)
data.reset_index(inplace=True, drop=True)
```
