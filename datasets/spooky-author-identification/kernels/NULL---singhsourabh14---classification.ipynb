```python
df = pd.read_csv('../input/train.csv')
```

```python
df.head()
```

```python
import nltk
```

```python
dc = df.copy()
```

```python
import re
corpus = []
```

```python
for i in df['text']:
    x = re.sub(r'\W', ' ', i)
    x = x.lower()
    x = re.sub(r'\s+[a-z]\s+', ' ', x)
    x = re.sub(r'^[a-z]\s+', ' ', x)
    x = re.sub(r'$\s+', '', x)
    x = re.sub(r'\s+', ' ', x)
    corpus.append(x)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
l = LabelEncoder()
```

```python
l.fit(df.author.unique())
```

```python
y = l.transform(dc.author)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfv = TfidfVectorizer(min_df=3, analyzer='word', stop_words='english')
```

```python
X = tfv.fit_transform(corpus).toarray()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
trainX, testX, trainY, testY = train_test_split(X,y, test_size=.3)
```

```python
clf = LogisticRegression(C=1.0, max_iter=200)
```

```python
clf.fit(trainX,trainY)
```

```python
pred = clf.predict(testX)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(pred,testY)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(pred,testY)
```

```python
pred = clf.predict_proba(testX)
```
