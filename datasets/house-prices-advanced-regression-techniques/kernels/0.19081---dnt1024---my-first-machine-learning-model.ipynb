```python
import pandas as pd

main_file_path = '../input/train.csv'
data = pd.read_csv(main_file_path)
#print(data.describe())
print(data.columns)
```

```python
data_price = data.SalePrice
print(data_price.head())
```

```python
columns_of_interest = ['LotArea', 'LotFrontage']
two_columns_of_data = data[columns_of_interest]
two_columns_of_data.describe()
```

```python
y = data_price
list_predictors = ['LotArea','YearBuilt','1stFlrSF','2ndFlrSF','FullBath','BedroomAbvGr','TotRmsAbvGrd']
X = data[list_predictors]
```

```python
from sklearn.tree import DecisionTreeRegressor

# Define model
my_model = DecisionTreeRegressor()

# Fit model
my_model.fit(X, y)
```

```python
print("Making predictions for the following houses:")
print(X.head())
print("The Predictions are:")
print(my_model.predict(X.head()))
```

```python
# measuering the model 's quolity
from sklearn.metrics import mean_absolute_error

price_of_prediction = my_model.predict(X)
mean_absolute_error(y,price_of_prediction)
```

```python
from sklearn.model_selection import train_test_split

train_X, test_X, train_y, test_y = r=train_test_split(X,y,random_state = 0)
print(train_X.head())
model1 = DecisionTreeRegressor()
model1.fit(train_X, train_y)

predict_value = model1.predict(test_X)
print(mean_absolute_error(test_y, predict_value))
```

```python
def get_mae(max_leaf_node,train_X,train_y,test_X,test_y):
    model = DecisionTreeRegressor(max_leaf_nodes= max_leaf_node,random_state=0)
    model.fit(train_X,train_y)
    predict_test = model.predict(test_X)
    mae = mean_absolute_error(test_y, predict_test)
    return mae
```

```python
# compare MAE with differing values of max_leaf_nodes
for max_leaf_node in [5, 50, 500, 5000]:
    my_mae = get_mae(max_leaf_node, train_X, train_y, test_X, test_y)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" % (max_leaf_node, my_mae))
```
