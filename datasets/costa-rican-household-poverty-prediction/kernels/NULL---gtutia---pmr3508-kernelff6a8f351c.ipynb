```python
%matplotlib inline

import matplotlib
import pandas as pd
house = pd.read_csv("../input/treino-costa-rica/train.csv",na_values='NaN')
house.shape
```

```python
house.head()
```

Vamos observar quantos dados faltantes existem na base.

```python
pd.options.display.max_rows = 150
pd.options.display.max_columns = 150
```

```python
house.isnull().sum()
```

É possível perceber que três colunas são responsáveis pela maioria dos dados faltantes: v2a1, v18q1 e rez_esc. Essas colunas possuem mais de 70% de seus dados não preenchidos. Assim, vamos optar pela exclusão destas colunas e em seguida retiraremos as linhas com dados faltantes das colunas restantes.

```python
house = house.drop(['v2a1','v18q1','rez_esc'],axis=1)
house_full = house.dropna()
house_full.shape
```

```python
house_full.head()
```

Vamos retirar os dados não numéricos e utilizar todos os dados restantes para construir o classificador kNN. Em seguida, vamos iterar o valor de k para atingir o que dará maior acurácia. Feito essas etapas começaremos a analisar os dados para aumentar aidna mais a acurácia do classificador

```python
house_full_num = house_full.drop(['Id','idhogar','dependency','edjefe','edjefa'],axis=1)

house_full_num.shape
```

```python
Xhouse = house_full_num.drop(["Target"],axis=1)
Yhouse = house_full_num.Target

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
from sklearn.model_selection import cross_val_score
scores = cross_val_score(knn,Xhouse,Yhouse,cv=10)
scores
```

```python
scores.mean()
```

A acurácia média com todos os dados não numéricos e k = 1 foi de aproximadamente 51,57%

```python
knn2 = KNeighborsClassifier(n_neighbors=83)
scores2 = cross_val_score(knn2,Xhouse,Yhouse,cv=10)
scores2.mean()
```

O valor de k encontrado que resultava na melhor acurácia média foi o de k = 83, com 63,51% aproximadamente de acurácia média.

A partir de agora vamos analisar a influência de certos grupos de dados na construção de classificador. Na seção Data do desafio do Kaggle está a descrição sobre cada uma das colunas da base de dados. Alguns grupos de dados foram retirados do classificador e foi observado a variação na acurácia média do classificador.

Vamos iniciar retirar os últimos dados da base, que se tratam nada menos do que outros dados elevados ao quadrado. Os dados retirados foram: SQBescolari, SQBage, SQBhogar_total, SQBedjefe, SQBhogar_nin, SQBovercrowding, SQBdependency, SQBmeanede e agesq.

```python
Xhouse2 = house_full_num.drop(["Target","SQBescolari","SQBage","SQBhogar_total","SQBedjefe","SQBhogar_nin","SQBovercrowding","SQBdependency","SQBmeaned","agesq"],axis=1)
Yhouse2 = house_full_num.Target

from sklearn.model_selection import cross_val_score
scores3 = cross_val_score(knn2,Xhouse2,Yhouse2,cv=10)
scores3.mean()
```

A retirada desses dados melhorou a acurácia do classificador. Agora vamos retirar os dados que dizem respeito ao tipo de material que é constituido as paredes das casas. Foram retirados os seguintes dados: paredblolad,paredzocalo,paredpreb, pareddes, paredmad, paredzinc, paredfibras e paredother.

```python
Xhouse3 = house_full_num.drop(["Target","SQBescolari","SQBage","SQBhogar_total","SQBedjefe","SQBhogar_nin","SQBovercrowding","SQBdependency","SQBmeaned","agesq","paredblolad","paredzocalo","paredpreb","pareddes","paredmad","paredzinc","paredfibras","paredother"],axis=1)
Yhouse3 = house_full_num.Target

from sklearn.model_selection import cross_val_score
scores4 = cross_val_score(knn2,Xhouse3,Yhouse3,cv=10)
scores4.mean()
```

A retirada desses dados também melhorou a acurácia do classificador. Seguindo a linha da última modificação, iremos retirar os dados que falam sobre o tipo de material utilizado nos pisos das casas. Os seguintes dados foram retirados:pisomoscer, pisocemento, pisoother, pisonatur, pisonotiene e pisomadera.

```python
Xhouse3 = house_full_num.drop(["Target","SQBescolari","SQBage","SQBhogar_total","SQBedjefe","SQBhogar_nin","SQBovercrowding","SQBdependency","SQBmeaned","agesq","paredblolad","paredzocalo","paredpreb","pareddes","paredmad","paredzinc","paredfibras","paredother","pisomoscer","pisocemento","pisoother","pisonatur","pisomadera"],axis=1)
Yhouse3 = house_full_num.Target

from sklearn.model_selection import cross_val_score
scores4 = cross_val_score(knn2,Xhouse3,Yhouse3,cv=10)
scores4.mean()
```
