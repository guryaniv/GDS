```python
import pandas as pd

main_file_path = '../input/house-prices-advanced-regression-techniques/train.csv' # this is the path to the Iowa data that you will use
data = pd.read_csv(main_file_path)

# Run this code block with the control-enter keys on your keyboard. Or click the blue botton on the left
print(data.describe())
```

```python
print(data.columns)
```

```python
price = data.SalePrice
print(price.head())
```

```python
columnsOfData = ['Neighborhood', 'LotArea']
columns = data[columnsOfData]
columns.describe()
```

```python
y = data.SalePrice
```

```python
dataPredictors = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X = data[dataPredictors]
```

```python
from sklearn.tree import DecisionTreeRegressor

iowaModel = DecisionTreeRegressor()

iowaModel.fit(X,y)
```

```python
print("Making predictions for the following 5 houses:")
print(X.head())
print("The predictions are")
print(iowaModel.predict(X.head()))
```

```python
from sklearn.metrics import mean_absolute_error

predictedPrice = iowaModel.predict(X)
mean_absolute_error(y, predictedPrice)
```

```python
from sklearn.model_selection import train_test_split

trainX, valX, trainy, valy = train_test_split(X, y,random_state = 0)

iowaModel.fit(trainX, trainy)

predictedPrice = iowaModel.predict(valX)

print(mean_absolute_error(valy, predictedPrice))
```

```python
print(trainy.describe())
print(valy.describe())
```

```python
# Function for MAE calculation based on max_leaf_nodes
def get_mae(max_leaf_nodes, predictors_train, predictors_val, targ_train, targ_val):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(predictors_train, targ_train)
    preds_val = model.predict(predictors_val)
    mae = mean_absolute_error(targ_val, preds_val)
    return(mae)

# compare MAE with differing values of max_leaf_nodes
for max_leaf_nodes in [5, 50, 500, 5000]:
    my_mae = get_mae(max_leaf_nodes, trainX, valX, trainy, valy)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
    
```

```python
# Random forest

from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

forestModel = RandomForestRegressor()
forestModel.fit(trainX, trainy)
pricePredictions = forestModel.predict(valX)
print(mean_absolute_error(valy, pricePredictions))
```

```python
# Predicting on test data for submission to competition
test = pd.read_csv('../input/house-prices-advanced-regression-techniques/test.csv')

testX = test[dataPredictors]

testPricePrediction = forestModel.predict(testX)

print(testPricePrediction)
```

```python
# Preparing submission file
mySubmission = pd.DataFrame({'Id': test.Id, 'SalePrice': testPricePrediction})

# you could use any filename. We choose submission here
mySubmission.to_csv('submission.csv', index=False)
```
