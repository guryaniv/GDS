```python
#A few preparation of cryptocurrencies pricing data:
list_of_currency_files = check_output(["ls", "../input"]).decode("utf8").split('\n')
input_folder = '../input/'
list_of_currency_files.remove('')
```

Certainly it is not wise to process all those data once. We will try to analyze a few leading or top performing cryptocurrencies first. I used the tableau to visualize the market cap (the main benchmark) of each currency and select five currencies: Bitcoin, Litecoin, Ripple, Monero and Ethereum. 

In this case, we will dive into the Bitcoin first since that's the top performing cryptocurrency. We do have some extra data in case of Bitcoin in file bitcoin data set, but we will use the bitcoin price data.

First of all, let's visualize the pricing history of Bitcoin, Litecoin, Ripple, Monero and Ethereum:

```python
#Bitcoin
color = sns.color_palette()
bitcoin_price['Date_mpl'] = bitcoin_price['Date'].apply(lambda x: mdates.date2num(x))

print(bitcoin_price.head())
fig, ax = plt.subplots(figsize=(10,8))
sns.tsplot(bitcoin_price.Close.values, time=bitcoin_price.Date_mpl.values, alpha=0.8,color = color[2], ax=ax)
ax.xaxis.set_major_locator(mdates.AutoDateLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y.%m.%d'))
fig.autofmt_xdate()
plt.ylabel('Price in USD', fontsize=12)
plt.title("Closing price distribution of bitcoin", fontsize=15)
plt.show()
```
