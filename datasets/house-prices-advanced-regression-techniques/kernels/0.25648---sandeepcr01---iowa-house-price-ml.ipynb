The submission now considers only numerical values to predict the prices in IOWA.






```python
import pandas as pd
import numpy as np
main_file_path = '../input/train.csv'
data = pd.read_csv(main_file_path)


#Define the set of columns with which we will be training the data
print(data.head())
columns_data=['LotArea','YearBuilt','1stFlrSF','2ndFlrSF','FullBath','BedroomAbvGr','TotRmsAbvGrd']

X=data[columns_data]


#We will build a model using the Basic decision tree Regressor algorithm
from sklearn.tree import DecisionTreeRegressor

y=data.SalePrice
# Define model

# Define model
iowa_model = DecisionTreeRegressor(max_leaf_nodes=2)


# Fit model
iowa_model.fit(X, y)

from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

forest_model = RandomForestRegressor()
forest_model.fit(X,y)
randomforest = forest_model.predict(X)

predicted_data=iowa_model.predict(X)

#Checking the predicted values against the mean square error
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(y,randomforest))


test_file_path= '../input/test.csv'

testData= pd.read_csv(test_file_path)




X_Test=testData[columns_data]

X_Test=X_Test


predicted_data_Test=iowa_model.predict(X_Test)

predicted_data_RF=forest_model.predict(X_Test)
y=y.drop([1459])


print("the predicted values are")
print (predicted_data_Test)
print(mean_absolute_error(y,predicted_data_Test))
print(mean_absolute_error(y,predicted_data_RF))

my_submission = pd.DataFrame({'Id': testData.Id, 'SalePrice': predicted_data_Test})
my_submission.head()
my_submission.to_csv('IOWA_Predicted_Prices.csv',header=True, index=False)
```
