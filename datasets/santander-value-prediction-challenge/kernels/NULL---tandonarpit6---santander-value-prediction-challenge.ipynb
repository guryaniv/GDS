Trying to predict the financial transaction value of each customer based on the different columns- columns are all masked. This becomes more of a data analytics exercise than using data to make business sense. 

```python
import numpy as np 
import pandas as pd 

import os
print(os.listdir("../input"))
```

```python
train=pd.read_csv("../input/train.csv")
test=pd.read_csv("../input/test.csv")

train.head()
test.head()
```

```python
train.shape
```

```python
test.shape
```

data is completely anonymous- only a computing exercise. Training has 4459 rows and test has 49342 rows, looks a little incorrect. But train has 4993 while  test has 4992 (1 coumn for target might have been removed).  Since, there is no sense of data- let's use basic neural network to predict the target values.

```python
train_nonzero=train.astype(bool).sum(axis=0)
print(train_nonzero)
```

Let us remove the columns below a certain threshold. Tough to remove as we do not know the meaning of any column but still any value below 10% of all rows (10% of rows is 445), so can remove any column with value below 10% of row number- take 500 as a threshold.

```python
modified_train=train.loc[:,train_nonzero>500]
modified_train.shape
```

In modified_train dataframe took all rows from train file but removed columns with less than 500 non-zero entries. The new data frame has 332 columns, less than 10% in the  original file but still a large number. Let's build Keras model to train and perform predictions. Let's also set up training set (X), Target (Y) and align test set to the training set.

```python
Y=modified_train.loc[:,'target']
X=modified_train
X=X.drop(columns=['ID','target'])
```

```python
X_hat= test[test.columns.intersection(X.columns)]
X_hat.shape
```

```python
X
```

```python
X_hat
```

Let us set up the Keras model with X as feature set for training data, Y as target and X_hat as feature set for test data.

```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Activation
```

```python
model=Sequential()
model.add(Dense(128,input_dim=330,activation='relu'))
model.add(Dense(128,activation='relu'))
model.add(Dense(128,activation='relu'))
model.add(Dense(1,activation='relu'))
```

```python
model.summary()
```

```python
model.compile(optimizer='rmsprop',loss='mse')
```

```python
model.fit(X,Y,epochs=64,batch_size=32)
```

```python
Y_hat=model.predict(X_hat)
```

Preapring the submission file by adding columns id and prediction value from the neural network model.

```python
test['target']=Y_hat
```

```python
submission_file=test.loc[:,['ID','target']]
```
