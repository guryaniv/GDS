```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
####reading the data
```

```python
data=pd.read_csv(r'../input/creditcard.csv')
```

```python
###We will check if dataset is balanced or not
```

```python
from collections import Counter
print('Resampled dataset shape {}'.format(Counter(data['Class'])))
```

```python
###As we can see we have highly imbalanced dataset,we need to balance it before predicting.
###There are two ways,undersampling or oversampling.We will use SMOTE to oversample the fraud data
```

```python
from imblearn.over_sampling import SMOTE
from collections import Counter
sampler = SMOTE()
X_res, y_res=sampler.fit_sample(data.ix[:, data.columns != 'Class'],data.Class)
print('Resampled dataset shape {}'.format(Counter(y_res)))
```

```python
###Now we will shortlist the variables for classification using varianceThreshold method.I am starting with selecting all those variables which has atleast 80 percent variance
```

```python
from sklearn.feature_selection import VarianceThreshold
sel = VarianceThreshold(threshold=(.8 * (1 - .8)))
X_res=sel.fit_transform(X_res)
```

```python
####Now we will start with logistic regression,we will take all data till 28000 as test and remaining as train
```

```python
from sklearn.linear_model import LogisticRegression
logit=LogisticRegression()
logit.fit(X_res[:28000],y_res[:28000])
y_pred=logit.predict(X_res[28000:])
```

```python
#####Create confusion matrix to know TP,FP,TN and FN
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_res[28000:],y_pred)
```

```python
###create classification to check precision and recall of model
```

```python
from sklearn.metrics import classification_report
target_names = ['class 0', 'class 1']
print(classification_report(y_res[28000:],y_pred,target_names=target_names))
```

```python
####generating AUC to check accuracy of th model
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_res[28000:],logit.predict_proba(X_res[28000:])[:,1])
```
